        1.PIP安装
        首先呢，进入ubantu终端，在终端里，输入下面两条指令更新一下源

        $ sudo apt-get update  
        安装pip包
        $ sudo apt install python3-pip  #针对python3
        $ sudo apt install python-pip  #针对python2
        #建议可以都安装一下，安装时间比较长，耐心等待

        安装完成之后，对pip install 包 的下载的镜像源进行更改，可以使得下载速度提高很多
        在主目录下创建.pip文件，并在pip.conf中添加配置：

        $ mkdir ~/.pip  
        $ vim ~/.pip/pip.conf
        在pip.conf文件中添加
        [global]
        index-url = http://mirrors.aliyun.com/pypi/simple/
        [install]
        trusted-host=mirrors.aliyun.com

        2.配置python的虚拟环境
        python虚拟环境的配置可以使得当你在进行不同的项目时，使用不同的虚拟环境，因为不同的项目往往所需要的python包的版本不同，以免混淆。
        需要下载virtualenv与virtualenvwrapper两个包
        $ sudo pip install virtualenv
        $ sudo pip install virtualwrapper
        在主目录~/.bashrc文件中加上如下代码：
        1.打开vim ~/.bashrc
        2.添加：
        export VIRTUALENVWRAPPER_PYTHON =/usr/bin/python3  # 这个是指定python3的路径
        export WORKON_HOME = ~/.local/bin/virtualenvdir    # 这个是将来创建的虚拟环境的路径
        source ~/.local/bin/virtualenvwrapper.sh		   # 这个是ubantu19下载完virtualenvwrapper之后自动生成的路径
        
        这里WORKON_HOME的参数可以更改，你想要将虚拟环境放在什么路径，就填写什么，前提是这个路径必须存在。这个路径很重要，之后pycharm连接ubantu的时候要用到。
        完成之后大概是这个样子的
        
        对这个路径再做一下补充~/.local/bin/virtualenvdir，这个路径是我自己创建的，用来存放虚拟环境的相关文件。
        配置完成~/.bashrc这个文件之后，执行如下指令

        source ~/.bashrc
        如果执行之后报错，应该是路径问题，要么就是python的路径不正确，或者说virtualwrapper.sh的路径不正确，也可能是你用来存放虚拟环境文件的路径不存在。。。

        此时就可以使用virtualenv的一些指令了：
        deactivate # 退出当前虚拟环境
        workon [虚拟环境名称] # 使用某个虚拟环境
        rmvirtualenv [虚拟环境名称] # 删除某个虚拟环境
        lsvirtualenv # 列出所有虚拟环境
        mkvirtualenv [虚拟环境的名称] 创建一个虚拟环境
        $ mkvirtualenv remote1  # 创建一个python虚拟环境

        大概就是这样，之后你就进入到了remote1的python虚拟环境中，你在这个环境中随意的使用pip install都不会影响到其他虚拟环境中的包。
