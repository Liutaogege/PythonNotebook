# 短信验证码
- 短信验证法服务商有很多，我们选择使用阿里云平台。可以实现短信验证码发送，短信服务提醒，语音验证码，语音服务通知等
# 配置验证码方法如下：
- 1.登录阿里云：https://account.aliyun.com/
- 2.搜索短信服务，并跳转到该网页
- 3.首先点击用户头像有一个AccessKey管理，点击进去，会弹出一个页面，点击开始使用子用户AccessKey（子用户可以自己添加权限，比较安全）。点击进去以后，点击创建用户并输入相关信息后，勾选控制台访问和编程访问，并在下方设置定义密码
- 4.点击确定后跳转，点击下载csv文件，里面有AccessKeyID和AccessKeySercet，很重要，然后给该用户添加权限，点击添加权限后，会弹出一个页面，搜索短信将其两个相关的权限添加进去
- 5.然后回到短信服务页面的首页（即：https://dysms.console.aliyun.com/dysms.htm?#/overview ），点击快速学习来添加应用，模板
- 6.见下图，分别点击马上点击添加签名，添加模板

    - 然后分别按提示操作即可，申请的签名和模板需要1-2小时来审核。审核通过后才可使用
- 

- 下载通过后，把相应的签名和模板填入，点击下方的API Demo，选择为python的。把API接口代码复制到自己的的任意一个python package文件下的py文件中。最好是把它封装成一个函数，然后在views.py中引入使用，封装如下。
- 还要安装相关的阿里云API包。可以点击python SDK使用说明查看安装
- "cn-hangzhou"表示的是区域id：详情查看：https://developer.aliyun.com/endpoints

```python
from aliyunsdkcore.client import AcsClient
from aliyunsdkcore.request import CommonRequest
import json


def sms_send(telephone, code):
    client = AcsClient('LTAI4GFc5RyXtAHJyRjaQh6T', '0uB85yLGOXaos51fC3tFX9wVGSDe4Y', 'cn-hangzhou')
    telmplateparam = json.dumps({"code": code})
    request = CommonRequest()
    request.set_accept_format('json')
    request.set_domain('dysmsapi.aliyuncs.com')
    request.set_method('POST')
    request.set_protocol_type('https') # https | http
    request.set_version('2017-05-25')
    request.set_action_name('SendSms')

    request.add_query_param('RegionId', "cn-hangzhou")
    request.add_query_param('PhoneNumbers', telephone)
    request.add_query_param('SignName', "小饭桌")
    request.add_query_param('TemplateCode', "SMS_206538139")
    # request.add_query_param('TemplateParam', "{\"code\":\"2580\"}")
    request.add_query_param('TemplateParam', telmplateparam)
    response = client.do_action(request)
    return str(response, encoding='utf-8')
```
