# 前端工具的安装
## nvm包的安装
- nvm（Node Version Manager）是一个用来管理node版本的工具，使用它使用的主要是npm（Node Package Manager），npm的作用是为了能够方便的管理一些前端开发的包。
- nvm的安装步骤如下：
    - 1.setup包地址：https://github.com/coreybutler/nvm-windows/releases
    - 2.然后点击下一步即可
    - 3.安装完成后，打开cmd输入nvm，如果报错，则需要配置环境变量，在我的电脑-属性-高级系统设置=环境变量-系统环境变量-path下新建一个，把nvm所处的路径加入进去即可
    - Linux安装nvm网址：https://github.com/creationix/nvm,也要配置环境变量

## nvm常用命令
- 1.nvm install node:安装最新版的node.js,
- 2.nvm install [version];安装指定版本的node.js
- 3.nvm use [version]:使用某个版本的node
- 4.nvm list：列出当前安装了哪些版本的node
- 5.nvm uninstall [version]:卸载指定版本的node
- 6.nvm node_mirror [url]:设置nvm的镜像
    - nvm node_mirror https://npm.taobao.org/mirrors/node/
- 7.nvm npm_mirror [url]:设置npm 的镜像
    - nvm npm_mirror https://npm.taobao.org/mirrors/npm


## node安装
- 安装完nvm后，就可以通过nvm来安装node了。因为最新版的node.js的npm可能有很多坑，所以安装不要最新的
- 比如nvm install v10.14.0

## npm
- npm（Node Package Manager）在安装node的时候就会自动的安装， 前提是你需要设置当前的node的版本，nvm use 10.14.0，然后就可以使用npm了
### 安装包
- 安装包分为全局安装和本地安装，全局安装是安装在当前的node环境中，可以在cmd中当做命令使用。而本地安装时安装在当前项目中，只有当前这个项目才能使用，并且可以通过require引用，安装的方式只有-g参数的区别
    - 本地安装：npm install express
    - 全局安装：npm install express -g
    
### 

#### 本地安装
- 1.将安装包放在./node_modules下（运行npm命令时所在的目录），如果没有该目录，会在执行当前npm命令的目录下生成node_modules目录
- 2.可以通过require()来引入本地安装的包

#### 全局安装
- 1.将安装包放在/usr/local下或者你node的安装目录
- 2.可以直接在命令行中使用

#### 卸载包
- npm uninstall [package]

#### 更新包
- npm update [package]

#### 搜索包
- npm search [package]

#### 使用淘宝镜像
- npm install -g cnpm -registry=https://registry.npm.taobao.org

# 前端项目搭建
- 前端使用gulp来自动化开发流程。配置好gulp后，可以自动给我们处理好一些工作。比如写完css后，要压缩成.min.css，写完js后，要做混淆和压缩，图片压缩等，这些工作都可以让gulp来帮我们完成

# 安装gulp
## 1.创建本地包管理环境
- 使用npm init命令在本地生成一个package.json文件。package.json是用来记录你当前这个项目依赖了哪些包，以后别人拿到你这个项目后，不需要你的mode_modeles文件夹（因为node_modules中的包实在是太庞大了）.只需要执行npm install 命令，即会自动安装package.json下devDependencies中指定的依赖包
    
## 2.安装gulp
- gulp的安装只需要使用npm命令即可。但是因为gulp需要作为命令行的方式运行。因此需要在安装在系统级别的目录中。先全局安装
    - npm install gulp -g
- 因为本地需要使用require的方式gulp，因此也需要在本地安装一份。进入到项目的目录的地址下，先使用初始化命令：npm init,然后在使用如下命令
    - npm install gulp --save-dev
    
- 以上的--save-dev是将安装的包添加到package.json下的devDependencies依赖中。以后通过npm install即可自动安装。devDependencies这个是用来记录开发环境下使用的包。如果想要记录生产环境下使用的包。那么在安装包的时候使用npm install xx --save就会记录到package.json下的dependencies中。dependencies是专门用来记录生产环境下的依赖包的
- 安装完成以后再使用npm install

## 3.创建gulp任务
- 要使用gulp来流程化我们的开发工作，首先需要在项目的根目录下创建一个gulpfile.js文件，然后在gulpfile.js中填入以下代码：
        var gulp = require("gulp")
        gulp.task("greet", function() {
            console.log('hello world');
        }
- 对以上代码的解释：
    - 1.通过require语句引用已经安装的第三方依赖包，这个require只能是引用当前项目的，不能引用全局下的。require语法是node.js独有的，只能在node.js环境下使用
    - 2.gulp.task是用来创建一个任务，gulp.task的第一个参数是命令的名字，第二个参数是一个函数，就是执行这个命令的时候会做什么事情，都是写在这个里面的
    - 3.写完以上代码后，以后如果想要执行greet命令，那么只需要进入到项目所在的路径，然后终端使用gulp greet即可执行
    
## 4.创建处理css文件的任务
- gulp只是提供一个框架给我们，如果想要实现一些更加复杂的功能，比如css压缩，那么我们还需要安装一下gulp-cssnano插件，gulp相关的插件安装也是通过npm命令安装，安装方式跟其他包是一模一样的（gulp插件本身就是一个普通的包）
- 对css文件的处理，需要做的事情就是压缩，然后再将压缩后的文件放到指定目录下（不要和原来css文件重合了），这里我们使用gulp-cssnano来处理这个工作
    - npm install gulp-cssnano --save-dev

- 然后在gulpfile.js中写入以下代码：
        var gulp = require("gulp")
        var cssnano = require("gulp-cssnano")
        
        //定义一个处理css文件改动的任务
        gulp.task("css", (cb) => {
            gulp.src("./css/*.css")
            .pipe(cssnano())
            .pipe(gulp.dest("./dist/css/"));
            cb();
        });
        代码写完以后，命令行中执行gulp css即可实现压缩
- 对以上代码进行解释：
    - 1.gulp.task：创建一个css处理的任务
    - 2.gulp.src找到当前css目录下所有以.css结尾的css文件
    - 3.pipe:管道方法,将上一个方法的返回结果传给另外一个出来，比如以上的cssnano
    - 4.gulp.dest:将处理完后的文件存放到指定的目录下，不要放在和源文件相同的目录下，以免产生冲突，也不方便管理
    
## 5.修改压缩后的文件名
- 像以上任务，压缩完css文件后，最好是添加一个.min.css后缀，方便知道它是压缩后的文件。此时需要使用gulp-rename来修改，当然需要安装npm install gulp-rename --save-dev
        var gulp = require("gulp");
        var cssnano = require("gulp-cssnano");
        var rename = require("gulp-rename");
        gulp.task("css", (cb) => {
            gulp.src("./css/*.css")
            .pipe(cssnano())
            .pipe(rename({"suffix": ".min"}))
            .pipe(gulp.dest("./dist/css/"));
            cb();
        });
- 以上代码，增加了一行.pipe(rename({"suffix":".min"}))，这个就是使用rename方法，并且传递一个对象参数，指定修改名字的规则为添加一个.min后缀名，这个gulp-rename还有其他的指定文件名的方式。比如可以在文件名前加个前缀等。更多的rename文档参看：https://www.npmjs.com/package/gulp-rename

### 5.1自动添加css样式前缀
- 一些css的样式，不同的浏览器处理的时候是不一样的，比如box-shadow，是用来指定元素阴影的，但是一些老版本的浏览器，需要加相应的前缀，比如： 
         //firefox4.0
         -moz-box-shadow: 10px 10px 5px #888888;
         //Safari and Google chrome 10.0
         -webkit-box-shadow: 10px 10px 5px #888888;
         //firefox4.0+,Google chrome 10.0+.Opera10.5+and IE9
         box-shadow: 10px 10px 5px #888888;
         
## 6.创建处理js文件的任务
- 处理js文件，我们需要使用到gulp-uglify插件：安装命令：npm install gulp-uglify --save-dev
- 安装完成后，我们就可以对js文件进行处理了
        var gulp = require("gulp");
        var rename = require("gulp-rename");  //重命名
        var uglify = require("gulp-uglify");  //压缩js文件
        // 压缩js文件并添加后缀min
        gulp.task("script", (cb) => {
            gulp.src("./js/*.js")
            .pipe(uglify({"toplevel": true}))   //可通过在uglify中添加参数来达到更简洁的压缩效果
            .pipe(rename({"suffix": ".min"}))
            .pipe(gulp.dest("./dist/js/"));
            cb();
        });
- 这里就是增加了一个.pipe(uglify())的处理，对js文件进行压缩和丑化（修改变量名）等处理，更多关于gulp-uglify的教程参看：https://github.com/mishoo/UglifyJS2#minify-options


## 7.合并多个文件：
- 在网页开发中，为了加快网页的渲染速度，有时候会将多个文件压缩成一个文件，从而减少请求的次数，要拼接文件，需要使用guip-concat插件
    - 安装： npm install guip-concat --save-dev
- 比如我们现在有一个nav.js文件用来控制导航条的，有个index.js文件来控制首页整体内容的，那么可以使用以下代码将这两个文件合并成一个文件
        var guip = require("guip");
        var uglify = require("gulp-uglify");  //压缩js文件
        var concat = require("gulp-concat");
        gulp.task("vendorjs", (cb) => {
            gulp.src("./js/*.js")
            .pipe(concat("index.min.js"))  //合并后的文件名
            .pipe(uglify())  //尽管合并了文件，但没用真正压缩，所以需要使用uglify
            .pipe(gulp.dest("./dist/js/"));
            cb();
        });

## 8.压缩图片
- 图片是限制网站加载速度的一个主要原因，图片越大，从网站上下载所花费的时间越长。因此，对于一些图片可以采用无损压缩，即在不改变图片质量的基础上进行压缩，通过gulp-imagemin来实现：npm install gulp-imagemin --save-dev
- 压缩图片也是一个比较大的工作量，对于一些已经压缩过的图片，没必要进行二次压缩，此时可使用gulp-cache来缓存那些压缩过的图片，安装命令如下：npm install gulp-cache --save-dev
- 两个插件结合使用的代码如下：
        var gulp = require("gulp");
        var imagemin = require("gulp-imagemin");
        var cache = require("gulp-cache");
        gulp.task("images", (cb) => {
            gulp.src("./images/*.*")
            .pipe(cache(imagemin()))
            .pipe(gulp.dest("./dist/images/"));
            cb();
        });

## 9.检测代码修改，自动刷新浏览器
- 以上所有的任务，都是需要手动的在终端去执行，这样不方便。为了解决这个问题，可以使用gulp内置的watch方法解决
        //gulp4.0版本之前的写法
        gulp.task("watch", function() {
            gulp.watch("./css/*.css", ["css"]);
            //参数一指监听哪个文件，参数二指当监听的文件变化时执行哪个task
        });
        //gulp4.0版本之后有大的改动
        // 新版本中watch后面只能是gulp.parallel 或者gulp.series
        // 定义任务队列，使用gulp.parallel or gulp.series
        let taskSeries = gulp.series("css");

        // 定义一个监听的任务
        gulp.task("watch", function(done){
            // 监听所有的css文件，然后执行css这个任务
            gulp.watch("./css/*.css", taskSeries);
            done()
        });
        
        // 函数的不同表示
        gulp.task("watch", (cb) => {
        gulp.watch("./css/*.css", taskSeries);
        cb()
        });


- 以后只要在终端执行gulp watch命令即可自动监听所有的css文件，然后自动执行css的任务，完成相应的工作
### 补充gulp.series(),gulp.paralllel
- gulp.series 用于串行（顺序）执行
- gulp.parallel 用于并行执行
    - 两个函数接收两个参数：要执行的任务的名字，执行该任务之前需要执行的函数（即任务名）
- 这两个函数是用来解决老版本存在的一个问题，目前还没弄清楚该问题是啥

## 10，更改文件后，自动刷新浏览器
- 以上虽然自动执行处理了css文件，但是还是需要手动刷新浏览器，才能看到修改后的样式，为了让浏览器自动刷新，可以使用browser-sync。安装：npm install browser-sync --save-dev
        var bs = require("browser-sync").create();
        var gulp = require("gulp");
        var cssnano = require("gulp-cssnano");
        var rename = require("gulp-rename");
        
        gulp.task("css", (cb) => {
            gulp.src("./css/*.css")
            .pipe(cssnano())
            .pipe(rename({"suffix":".min"}))
            .pipe(gulp.dest("./dist/css/"))
            .pipe(bs.stream());  //要用bs刷新的对象需要加入bs.stream()
            cb();
        });
        
        let taskSeries = gulp.series("css");
        // 定义一个监听的任务
        gulp.task("watch", (cb) => {
            // 监听所有的css文件，然后执行css这个任务
            gulp.watch("./css/*.css", taskSeries);
            cb();
        });

        
        gulp.task("bs", function() {
            bs.init({
                server: {
                    baseDir: "./",
                },
                port: 8001,
                open: false,		
            })
        })
        let taskJihe = gulp.parallel(["bs", "watch"])
        gulp.task("server", taskJihe);
        在命令行中输入gulp server即可
        
        
        // 监听多个任务
        gulp.task("watch", (cb) => {
            gulp.watch(path.css+"*.css", gulp.series("css"));
            gulp.watch(path.html + "*.html", gulp.series("html"));
            gulp.watch(path.js + "*.js", gulp.series("js"));
            gulp.watch(path.images + "*.*", gulp.series("images"));
            cb();
        })
        
- 以上我们创建了一个bs的任务，这个任务会开启一个3000端口（默认，可以通过port设置来改变），之后再访问页面就通过http://127.0.0.1:3000 来访问。然后接下来还定义了一个server任务，这个任务会去执行bs和watch任务，只要修改了css文件，那么就会执行css的任务，然后就会自动刷新浏览器，更多browse-sync文档参见：http:browsesync.cn/docs/gulp

## gulp-util
- 该插件中有一个log方法用来打印出当js文件出错时的错误信息

## gulp-sourcemaps
- 由于引用的是min.css文件，所以打印出的错误信息会在一行，不知道源文件js是错的哪一行，用该插件就可以明确知道是哪一行出错
        // 定义一个处理js文件的任务
        gulp.task("js", (cb) => {
            gulp.src(path.js + "*.js")
                .pipe(sourcemaps.init())  //初始化js文件
                .pipe(uglify().on("error", ))
                .pipe(rename({"suffix": ".min"}))
                .pipe(sourcemaps.write())  //打印出错误信息的位置
                .pipe(gulp.dest(path.js_dist))
                .pipe(bs.stream());
            cb();
        });
