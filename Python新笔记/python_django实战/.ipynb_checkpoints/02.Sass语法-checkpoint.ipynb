# Sass语法
- 众所周知，css不是一门编程语言，它没法像js和Python你那样有逻辑处理的能力，甚至导入其他的css文件中的样式都不做不到，而Sass就是解决这些问题的。它允许使用变量，嵌套规则，mixins，导入等众多功能。并且完全兼容css语法。sass文件不能直接被网页所识别，写完sass后，还需要专门的工具转换为css才能使用

# Sass文件的后缀名
- Sass文件有两种后缀名，一个是scss，一个是sass，不同的后缀名，相应的语法也不一样。我们使用scss的后缀名，以下语法都是讲的该后缀名的语法

# 使用gulp将sass转换为css
- 将sass文件转换为css文件的工具有很多，这里使用gulp来实现。需要使用gulp-sass插件，安装：npm install gulp-sass --save-dev


# sass基本语法
### 注释
- 支持/* comment*/ 和//

### 嵌套
- sass语法允许嵌套，比如#main下有一个类为：.header，那么可以写成以下形式
        #main{
            background: orange;
            .header{
                width: 20px;
                height: 20px;
            }
        }
- 这样写起来更加直观，一看就知道.header是main下面的

### 引用父选择器(&)
- 有时候，在嵌套的子选择器中，需要使用父选择器，那么此时就可以通过&来表示
        a{
            font-weight: bold;
            text-decoration: none;
            &:hover{
                color: red;
            }
        }
        
### 定义变量
- 在sass中可以定义变量，对于一些常用的值，可以通过变量存储起来，定义变量使用&符号
        &mainWidth: 1000px;
        #main{
            width: &mainWidth;
        }
### 运算：
- 在sass中支持运算，比如现在有一个容器总宽度是900，要在里面平均放三个盒子，那么就可以通过变量来设置它们的宽度
        &mainWidth: 900px;
        .box{
            width:&mainWidth/3;
        }
        
### @import 语法
- 在css中@import只能导入css文件，而且对网站的性能有很大的影响，而sass中的@import则是完全实现了一套自己的机制。可以直接将指定文件的代码拷贝到导入的地方
    - @import "index.css"
    
### @extend语法：
- 有时候我们一个选择器中，可能会需要另外一个选择器的样式，那么就可以通过extend来直接将指定选择器的样式加入进来
        .error{
            background-color: red;
            border: 1px solid orange;
        }
        .serious-error{
            @extend .error;
            border-width: 3px;
        }
        
### mixin语法
- 有时候一段样式代码，可能要用很多地方，那么我们可以把它定义成mixin，需要用的时候直接引用就可以了
        @mixin large-text {
            font: {
                family: Arial;
                size:  20px;
                weight: bold;
            }
            color: red;
        }
        - 如果其他地方想要使用这个mixin的时候，可以直接通过@include来包含进来
        .page-title {
            @include large-text;
            padding: 4px;
            margin-top: 10px;
        }
        - @mixin也可以使用参数
        @mixin sexy-border($color, $width) {
            border:  {
                color:  $color;
                width:  $width;
                style:  dashed;
            }
        }
        - 那么以后在include的时候就需要传递参数了
        p {
            @include sexy-border(blue, 1px);
        }
        
### sass语法更多教程参看：http://sass.bootcss.com/docs/sass-reference/

# 如何使用阿里巴巴矢量图标库
- 进入网站并登陆，找到自己所需要的图标并加入购物车。
- 等到所有的图标加入完毕，点击购物车，点击添加至项目，新建一个文件夹存放这些图标
- 存放确定后进入到一个新的页面，点击FontClass，然后点击暂无代码，点击生成这句话，生成一个css引用链接
- 将其复制到HTML文件中的link
- 然后就是引用图标，首先在图标出点击复制代码，然后使用i标签，将其类名赋值为复制的代码，再加上iconfont即可
    - 比如：i class="icon-personalcenterclick iconfont" ></i
