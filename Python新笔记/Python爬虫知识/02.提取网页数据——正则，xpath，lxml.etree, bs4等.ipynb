# 正则表达式re

## re常用函数
- re.match(规则， 所要操作的字符串)
    - 匹配是正则表达式的基础，在这基础上可以进行检索，替换等操作。match()从输入字符串头部开始匹配，如果匹配不成功，返回None；如果匹配成功，返回re.Match对象。
    - mathch实例方法:
    - start():匹配开始位置的index,对于match来说，该值始终为0
    - end():匹配结束位置的index,注意末尾字符真正的索引应该为index-1
    - span():返回一个包含两个元素的元组，第一个元素表示匹配开始的位置，第二个元素表示匹配结束的位置，
    - groups():得到所有的子分组信息,分组方法就是在模式中将某一部分用()包裹起来，就可以得到一个组。
    - group():得到一个分组，在括号中加入数字表示第几个子分组，当数字为0时和不加数字的结果一样
        - group(1)表示第一个分组，group(0)表示整个分组
- re.search(规则，字符串)接口函数在整个字符串内查找能够匹配上的子字符串。如果找到匹配的字符串，返回一个re.Match对象；如果未找到，返回None。
- re.sub(匹配模式，替换字符串，输入字符串)
    - 替换想替换的内容，如果匹配不成功，则返回值为输入字符串。默认情况下，该接口对所有符合条件的子字符串进行替换操作。
    - re.sub(r"a", "A", "abc123abcABC")，参数count该参数表示最多替换的次数
    

- re.compile(表达式,flags):返回的是一个匹配对象，它单独使用就没有任何意义，需要和findall(), search(), match(）搭配使用。
    - 表达式前面一般要加个r.
    - flags表示编译标志位，用于修改表达式的匹配方式，如是否区分大小写，多行匹配等
        - re.S(DOTALL)使 . 匹配包括换行在内的所有字符
        - re.I(IGNORECASE)使匹配对大小不敏感
        - re.L(LOCALE)做本地化失败匹配，语法等
        - re.M(MULTILINE)多行匹配，影响^和"美元符号"
        - re.X(VERBOSE)，添加注释
        - re.U 根据Unicode字符集解析字符，这个影响\w,\W,\b,\B
- re.spilt(表达式，str):使用表达式分割字符，
- re.findall():查找所有返回一个列表
- re.search(str,pos, endpos):在字符串中查找匹配，pos和endpos表示起始位置
- re.finditer():查找返回一个iter结果
- 正则符号的含义
    -  .(点号)：可以表示除换行在外的任意字符
    -  \:转义字符
    - |：或
    -  [候选字符列表]:匹配列表中的任一字符
        - [^a-z]:表示不匹配a-z中的任一字符
        - ^[a-zA-Z]：表示一个以字母开头的字符串；
    - \d:表示单个数字字符，等同于[0-9]
    - \D单个非数字字符，等同于[^0-9]
    - \s:空白字符，空白字符包含空格、回车\r、换行\n、制表\t、垂直制表\v、换页\f。
    - \S:表示非空白字符，使用\S+表示不包含空格字符的字符串。
    - \w:单词字符,如大小写字母，数字，下划线，不包含空格、#，美元符号
    - \W:除开\w包含的
    - \b:匹配单词的边界
    - \A:只匹配字符串的开头，\Aabcd,则abcd
    - \Z:仅匹配字符串的末尾：abcd\Z,则abcd
    - ():对正则表达式进行分组，从第一个括号开始，编号逐渐增大
    - *:表示任意个(0个或者多个)前面的字符
    - +:表示1个或者多个前面的字符。
    - ？:表示0个或者1个前面的字符。
    - {m}:（出现指定次数）,表示m个前面的字符。m可以为0，表示不出现该字符。
    - {m,n}（指定出现次数的范围）
    - ^表示以xxx开头，若放在中括号中，表示取反操作
    - $:表示以xxx结尾，如字符abcd用abc$匹配的结果为abc，而不是abcd
    - /^[0-9]$/:表示输入的只能是数字
    - (?P<name>...):分组：除了原来的编号再制定一个别名(?P<id>12345)
    - (?P=name):引用分组    $

## 正则表达式中的贪婪模式与非贪婪模式
- 贪婪模式： 正则表达式一般趋向于最大长度匹配，总是尝试匹配尽可能多的字符 
     - 贪婪模式的量词，也叫做匹配优先量词，包括： 

            “{m,n}”、“{m,}”、“?”、“*”和“+”。 

- 非贪婪模式：就是匹配到结果就好，总是尝试匹配尽可能少的字符
    - 在匹配优先量词后加上“?”，即变成属于非贪婪模式的量词，也叫做忽略优先量词，包括： 

            “{m,n}?”、“{m,}?”、“??”、“*?”和“+?”。 
            
- 举例：
            源字符串：aa<div>test1</div>bb<div>test2</div>cc 

            正则表达式一：<div>.*</div> 

            匹配结果一：<div>test1</div>bb<div>test2</div> 

            正则表达式二：<div>.*?</div> 

            匹配结果二：<div>test1</div>（这里指的是一次匹配结果，所以没包括<div>test2</div>） 

```python
import re

s = r"c:\bdxabcdfhjs"
print(s)
result = re.match(r"c:\\bd",s)
print(type(result))
begin = result.start()
end = result.end()
print(s[begin:end])
```

# 选取节点
- 在Xpath中，有七种类型的节点：元素，属性，文本，命名空间，处理指令，注释及文档节点（或称为根节点）
- nodename：选取此节点的所有子节点
- /root:从根节点开始选取，若用在不是根节点的地方无返回结果
- //ndname:选取名为ndname的所有节点，不考虑位置。一般返回列表
- .:选取当前节点
- ..:选取当前节点的父亲节点
- @：选取属性
- Xpath查找时按照路径方法查找，路径可分为绝对路径（绝对路径起始于/,/step/step/...）和相对路径(没有起始于/,step/step/...)举例如下
        People/Teacher:返回Teacher节点
        //Teacher:选取所有名为Teacher的节点，不论在那个位置
        People//Age:选取People后代中所有名为Age的节点，包括孙子节点
        //@money:选取所有含有money的属性
        //Age[@Detail]:选取带有属性Detail的Age元素
        

# 谓语—predicates
- 用来查找某个特定的节点 ，就写在方括号中。进一步筛选
- /School/Student[1]:选取School下面的第一个Student节点
- /School/Student[last()]:选取School下面的最后一个Student节点
- /School/Student[last() - 1]：选取倒数第二个节点
- /School/Student[position() - 3]:选取School下面的前两个Student节点
- //Student[@score]:选取带有属性score的Student节点
- //Student[@score="99"]:选取带有属性score且值为99的Student节点
- //Student[@score]/Age:选取带有属性score的Student节点的子节点

# Xpath的运算符
- |:或者
        //Student | //Teacher: 把节点名为Student和Teacher的节点都选取
- 其余不常见的Xpath的运算符号包括：加减乘除(除用div表示)等

# Xpath通配符
- 通配符可用来选取未知的xml元素
    - *：匹配任何元素节点
    - @*：匹配任何属性节点
    - node():匹配任何类型的节点
    
# Xpath轴
- 轴可定义相对于当前节点的节点集
            ancestor	选取当前节点的所有先辈（父、祖父等）。
            ancestor-or-self	选取当前节点的所有先辈（父、祖父等）以及当前节点本身。
            attribute	选取当前节点的所有属性。
            child	选取当前节点的所有子元素。
            descendant	选取当前节点的所有后代元素（子、孙等）。
            descendant-or-self	选取当前节点的所有后代元素（子、孙等）以及当前节点本身。
            following	选取文档中当前节点的结束标签之后的所有节点。
            namespace	选取当前节点的所有命名空间节点。
            parent	选取当前节点的父节点。
            preceding	选取文档中当前节点的开始标签之前的所有节点。
            preceding-sibling	选取当前节点之前的所有同级节点。
            self	选取当前节点。
            
- 步：根据当前节点集之中的节点来进行计算
- 步包括：
    - 轴，节点测试，零个或者更多谓语
- 步的语法：
    - 轴名称::节点测试[谓语]，举例如下
            child::book	选取所有属于当前节点的子元素的 book 节点。
            attribute::lang	选取当前节点的 lang 属性。
            child::*	选取当前节点的所有子元素。
            attribute::*	选取当前节点的所有属性。
            child::text()	选取当前节点的所有文本子节点。
            child::node()	选取当前节点的所有子节点。
            descendant::book	选取当前节点的所有 book 后代。
            ancestor::book	选择当前节点的所有 book 先辈。
            ancestor-or-self::book	选取当前节点的所有 book 先辈以及当前节点（如果此节点是 book 节点）
            child::*/child::price	选取当前节点的所有 price 孙节点。

### beautifulsoup的四种对象

#### Beautifulsoup：表示的是一个文档的全部内容，可看作是特殊的tag对象

#### comment:特殊类型的NavigableString

## 搜索文档树
- fing_all方法：find_all( name , attrs , recursive , text ,limit, **kwargs )
    - 搜索当前tag的所有tag子节点,并判断是否符合过滤器的条件
    - name参数：
        - 传字符串：就是tag名字
        - 传正则表达式：Beautiful Soup会通过正则表达式的 match() 来匹配内容
        - 传列表：返回列表中包含的所有标签，没找到就算了
        - 传True：可以匹配任何值：但不会返回字符串节点
    - kwargs参数：
        - 如果一个指定名字的参数不是搜索内置的参数名,搜索时会把该参数当作指定名字tag的属性来搜索。注意若传入的字符是class，需要是这样class_，是为了和Python关键字区分
        - 上面这种方法和传入attrs参数效果一样，不过传入的attrs需要是字典类型
    - text参数：
        - 通过 text 参数可以搜索文档中的字符串内容.与 name 参数的可选值一样, text 参数接受 字符串 , 正则表达式 , 列表, True：
    - limit参数：
        -  limit 参数限制返回结果的数量.效果与SQL中的limit关键字类似,当搜索到的结果数量达到 limit 的限制时,就停止搜索返回结果.
    - recursive参数
        - 调用tag的 find_all() 方法时,Beautiful Soup会检索当前tag的所有子孙节点,如果只想搜索tag的直接子节点,可以使用参数 recursive=False：

- find方法：find( name , attrs , recursive , text , **kwargs )
    - 它与 find_all() 方法唯一的区别是 find_all() 方法的返回结果是值包含一个元素的列表,而 find() 方法直接返回第一个结果
    
- find_parents()  find_parent()：
    - find_all() 和 find() 只搜索当前节点的所有子节点,孙子节点等.. find_parents() 和 find_parent() 用来搜索当前节点的父辈节点,搜索方法与普通tag的搜索方法相同,

```python
from bs4 import BeautifulSoup

html = """
<html>
<head>
<meta charset="utf-8">
<title>Hello, 2020 happy new year</title>
<meta name="chenc" content="Runoob">
<meta name="description" content="时钟 时间 & 钟表 日历">
<meta name="viewport" content="width=device-width, initial-scale=1.0,minimum-scale=1.0,maximum-scale=1.0, user-scalable=no">
<link rel="stylesheet" type="text/css" href="css/style.css">
<link rel="stylesheet" type="text/css" href="css/font_1191451_h720mljzrsc.css">
<link rel="stylesheet" href="css/bootstrap.css">
<script src="js/jquery.min.js"></script>
<script src="js/bootstrap.min.js"></script>
<script src="js/jweixin-1.4.0.js"></script> 
<script src="js/app.js"></script>

</head>
<body>
    <div class="key">
        <div class="iocnBox"><i class="iconfont icon-delete"></i></div>
        <div class="empty">清空</div>
        <textarea placeholder="在此述说新年寄语..." rows="1" class="van-field__control"></textarea>
        <div class="buts">发送</div>
    </div>
    <div class="today">
        <div class="clock">
            <div class="pos SS"></div>
            <div class="pos MM"></div>
            <div class="pos HH"></div>
            <div class="spot"></div>
        </div>
        <!-- <p>北京时间</p> -->
        <!-- <div class="day"></div>         -->
        <div class="time"></div>
        <div class="sydate"></div>
    </div>
    <div class="Barrage"></div>
    <div class="message">发送弹幕</div>


    <div class="music">
        <audio class="play" id="indexVideo" controls autoplay>
            <!-- <source src="horse.ogg" type="audio/ogg"> -->
            <source src="mp3/akon.mp3" type="audio/mpeg">
            您的浏览器不支持 audio 元素。
        </audio>
        <marquee direction="left">akon.mp3</marquee>
    </div>


<div style="text-align:center;">
<p>更多源码：<a href="http://www.mycodes.net/" target="_blank">源码之家</a></p>
</div>
    


</body>
</html>
"""

soup = BeautifulSoup(html, "lxml")
metas = soup.find_all("meta")
for meta in metas:
    print(meta)
print("=" * 40)
#limit限制查找的个数
divs = soup.find_all("div", class_="clock", limit=2)
for div in divs:
    print(div)

print("=" * 40)
# 多个属性筛选条件,两种方法
links = soup.find_all("link", rel="stylesheet",type="text/css")
links = soup.find_all("link", attrs={"rel":"stylesheet", "type":"text/css"})
for link in links:
    print(link)

# 获取a中的链接的两种方法
alist = soup.find_all("a")
for a in alist:
    #第一种获取方法
    href = a['href']
    #第二种获取方法
    href = a.attrs['href']
    print(href)

# 获取元素的text内容
p_text = soup.find_all("div", class_="message")
for p in p_text:
    print(p)
    print(p.string)

div_text = soup.find_all("div")
for div in div_text:
    div_genrator = div.strings #strings是一个生成器
    for d in div_genrator:
        print(d)    #得到的字符中会包含空格，所以使用stripped_strings
    print("++" * 20)
    print(list(div.stripped_strings))
```

```python
# 查找所有类属性为“clock”的div标签
clock = soup.find_all("div", class_="clock")
print(list(clock))

print("=" * 40)
clock = soup.find_all("div", attrs={"class":"clock"})
print(list(clock))
```
