# Scrapy整体框架介绍



# scrapy安装
- 安装：pip install scrapy
- 注意：
    - 在ubuntu上安装scrapy之前，需要安装以下依赖：
            sudo apt-get install python-dev python-pip libxml2-dev libxsltl--dev zliblg-dev libffi-dev libssl-dev,然后再pip install scrapy

# 框架
- 包含各个部件
    - ScrapyEngine：神经中枢
    - Scheduler调度器：负责接受引擎发送过来的Request请求，并按照一定的方式进行整理排队，当引擎需要时，交还给引擎。
    - Downloader下载器：负责抓取网页，并把得到的response传给spider
    - Spider爬虫：解析response，得到items和urls
    - ItemPipeline管道：详细处理item，负责清洗，验证等
    - DownloaderMiddlerware下载中间体：处理传送到下载器的request和传送到引擎的response
    - SpiderMiddlerware爬虫中间体：处理抓取器的输入和输出
            
            
# scrapy使用步骤
- 1、新建scrapy项目，需要通过命令行实现。首先你先进入你要存放该项目的目录下，然后使用  scrapy startproject [项目名称] [项目地址] 创建。
    - 项目目录结构：
        - items.py:用来存放爬虫爬取下来的数据
        - middlewares.py:用来存放各种中间件的文件
        - pipelines.py:用来将items模型存储到本地磁盘中
        - settings.py:包含爬虫的一些配置信息（比如请求头，IP代理池）
        - scrapy.cfg:项目的配置文件
        - spiders包：以后所有的爬虫，都存放在这里面
- 2、创建爬虫：进入到项目所在的路径，执行命令：scrapy genspider [爬虫名字] [爬虫域名]
- 3、执行爬虫
    - scrapy crawl 爬虫名字
- 注：为了简便不使用命令行运行爬虫，可以在爬虫文件的根目录下新建一个start.py文件（就是与那个同名文件夹即项目地址在同一级目录），文件中写入;
        from scrapy import cmdline
        cmdline.execute("scrapy crawl 爬虫名".split())

### Spider的子类：CrawlSpider
- 继承自Spider类，功能更强大。可以定义爬取的url的规则，只要满足某个条件的url，就可以帮我们进行爬取，不用手动的request
- 创建crawlspider：scrapy genspider -t crawl [爬虫名字] [域名]
- 其他方法和spider一样
- 包含模块：    
    - 1、LinkExtractors：链接提取器，这个可以在所爬取的页面中找到满足规则的url，实现自动爬取
        - class scrapy.linkextractors.LinkExtractor(
            allow = (),
            deny = (),
            allow_domains = (),
            deny_domains = (),
            deny_extensions = None,
            restrict_xpaths = (),
            tags = ('a', 'area'),
            canonicalize = True,
            unique = True,
            process_value = None)
        - 主要参数讲解：
            - allow:允许的url，所有满足这个正则表达式的url都会被提取
            - deny：禁止的url，与上相反
            - allow_domains:允许的域名，只有在这个里面指定的域名的url才会被提取
            - restrict_xpaths:严格的xpath，和allow共同过滤链接
    - 2、Rule类模块
        - 定义爬虫的规则类

            link_extractor,

            follow = None,
            process_links = None,
            process_request = None)

            - link_extractor;一个linkExtractor对象

            - follow：指定根据该规则从response中提取的链接是否需要跟进
            - process_links：从link_extractors中获取到链接后会传递给这个函数，用来过滤不需要爬取的链接

#### scrapy.Request对象
- scrapy.Request(url, callback, method)：Reqeuest对象在需要重新定向到一个新的url时调用，这个类需要传递一些参数，其中比较常用的参数有：
    - url:Request对象发送请求的url
    - callback:在下载器下载完相应的数据后执行的回调函数,默认为pares函数
    - method：请求的方法，默认为GET请求，一般不需要指定。若需要POST请求，用FormRequest即可，在settings.py中指定
        - scrapy.FromRequest():FormRequest 类为Request的子类，用于POST请求
            - 这个类新增了一个参数 formdata，其他参数与Request一样
            - 示例：yield scrapy.FormRequest(url="http://www.example.com/post/action",
                    formdata={'name': 'Zarten', 'age': '27'},
                    callback=self.after_post)
    - headers：请求头，一般在settings中设置即可，也可在middlewares中设置
    - body:str类型，为请求体，一般不需要设置（get和post其实都可以通过body来传递参数，不过一般不用）
    - meta：比较常用，用于在不同的请求之间传递数据用的，用户可自定义从Request到Response传递参数，这个参数一般也可在middlewares中处理
    - cookies:dict或list类型，请求的cookie
        - 列表方式：cookies = [{'name': 'Zarten', 'value': 'my name is Zarten', 'domain': 'example.com', 'path': '/currency'}]
    - encoding
    - dont_filter：表示不由调度器过滤，默认为False，若设置为True，这次请求将不会过滤（不会加入到去重队列中），可以多次执行相同的请求
    - flags：list类型，一般不会用到，发送请求的标志，一般用于日志记录
    - errback：在发生错误的时候执行的函数
    
#### Response对象
- 一般是由scrapy主动构建的，不需要关心如何创建它，知道如何使用就行
- 方法：
    - meta：从其他请求传过来的meta属性，可以用来保持多个请求之间的连接
    - encoding：返回当前字符串编码和解码的格式
    - text：将返回的数据作为Unicode字符串返回
    - body:将返回来的数据作为bytes字符串返回
    - xpath:xpath选择器
- response是一个scrapy.http.response.html.HtmlResponse对象，可以用xpath和css语法来提取数据
    - response.xpath(), response.css()
    - 提取出来的数据是一个selector或者selectorlist，如果想要获取其中的字符串，执行getall或者get()方法
        - getall()方法：返回selector一个列表包含所有文本
        - get()方法：获取selector其中的第一个文本
    - 数据解析出来后，要传给pipline处理，可以使用yield来返回，或者是收集所有的item，最后统一用return返回
- item，在items.py中定义好专门存储数据的模型
- pipline:专门用来保存数据的，其中有三个方法常用：
    - open_spider(self, spider):当爬虫被打开的时候调用
    - process_item(self, spider):当爬虫有item传过来的时候调用
    - close_spider(self, spider):当爬虫关闭的时候被调用)
    要激活pipline，要在settings.py中设置ITEM_PIPLINES

### 下载文件和图片
- scrapy为下载item中包含的文件提供了一个可重用的item pipeliness.这些pipelines有些共同的方法和结构（称为media pipeline)，一般来说会使用Files Pipeline或者Images Pipeline
#### 下载文件的Files Pipeline：
- 当使用FilePipeline下载文件的时候，按照以下步骤：
    - 1.定义好一个Item，然后在这个item中定义两个属性，分别为file_urls以及files。file_urls是用来存储需要下载的url链接，是一个列表
    - 2.当文件下载完成后，会把文件下载的相关信息存储到item的files属性中，比如下载路径，下载的url和文件的校验码等
    - 3.在配置文件setting.py中配置FIFES_STORE,这个配置是用来设置文件下载的路径
    - 4.启动pipeline：在ITEM_PIPELINES中设置scrapy.pipeline.files.FilesPipeline:1
#### 下载图片的Images Pipeline
- 当使用Images Pipeline下载文件的时候，按照以下步骤来完成
- 1.定义好一个Item，然后在这个item中定义两个属性，分别为image_urls以及images。image_urls是用来存储下载的图片的url链接，是一个列表
- 2.当文件下载完成后，会把文件下载的相关信息存储到item的images属性中。
- 3.在settings.py中配置IMAGE_STORE,它是用来设置图片下载下来的路径
- 4.启动pipeline：在ITEM_PIPELINES中设置scrapy.pipelines.images.ImagesPipeline:1


#### 将存到item的数据输出成json格式
- JsonItemExporter和JsonLinesItemExporter
- 导入：from scrapy.exporters import JsonLinesItemExporter,JsonItemExporter
    - JsonItemExporter:这个是每次把数据都添加到内存中，最后统一写入到磁盘中，好处是，存储的数据是一个满足json规则的数据，缺点是如果数据量比较大，会比较耗内存
    - JsonLinesItemExporter：这个是每次调用export_item的时候就把这个item存储到硬盘中，缺点是每个字典是一行，整个文件不是满足一个json格式的文件，优点是处理的数据直接存储到了硬盘中，不会耗内存，数据也比较安全
- 示例：
        class QsbkPipeline(object):
            def __init__(self):
                self.fp = open("duanzi.json", "wb")
                self.exporter = JsonLinesItemExporter(self.fp, ensure_ascii=False, encoding="utf-8")
            def open_spider(self, spider):
                print("爬虫开始")
            def process_item(self, item, spider):
                self.exporter.export_item(item)
                return item
            def close_spider(self, spider):
                self.fp.close()
                print("爬虫结束")

## 3.scrapy框架之Middlewares
- 可以有很多个，按顺序加载执行
- 在middlewares文件中
- 需要在settings中设置以生效：DOWNLOADER_MIDDLEWARES，SPIDER_MIDDLEWARES
- 一个中间件完成一项功能
- 必须实现以下一个或多个方法
### (1)downloader middlewares:下载器中间件
- 它是引擎和下载器之间通信的中间件，在这个中间件中我们可以设置代理，更换请求头等来达到反反爬虫的目的。可以在下载器中间件中实现两个方法，一个是process_request(self,request,spider),这个方法是在请求头发送之前会执行的，还有一个是process_response(self, request, reposnese, spider)，这个方法是数据下载到引擎之前执行。具体使用如下：

#### process_request(self,request， spider）方法
- 该方法是向下载器发送请求之前执行的，一般可以在这个里面设置随机代理和IP，以及user_agent
- 1、参数：
    - request:发送请求的request对象
    - spider：发送请求的spider对象
- 2、返回值：
    - 返回None，如果返回None，scrapy将继续处理该request，执行其他中间件中的相应方法，直到合适的下载器处理函数被调用。
    - 返回response对象;scrapy将不会调用任何其他的peocess_reqeuest方法，将直接返回这个response对象，已经激活的中间件的process_responsse（）方法则会在每个response返回时被调用
    - 返回request对象，不再使用之前的request对象去下载数据，而是根据现在返回的request对象返回数据
    - 如果这个方法抛出了异常，将会调用process_exception方法
    
#### process_response(self,request， spider)方法
- 这个是下载器下载的数据到引擎中间会执行的方法 
- 1、参数
    - request：request对象
    - response：被处理的response对象
    - spider：spider对象
- 2、返回值：
    - 返回response对象，会将这个新的response对象传给引擎，最终传给爬虫
    - 返回request对象，下载器链接被切断，返回的request会重新被下载器调度下载
    - 如果抛出一个异常，那么调用request的errback（）方法，如果没有指定这个方法，那么会抛出一个异常
    
##### 随机请求头举例
- 为了防止被识别出来机器爬虫，可以在中间件中实现随机请求头中的随机用户
- 首先在settings中设置USER-AGENT=["",""]，然后在中间件中定义一个类
        from scrapy.downloadermiddlewares.useragent import UserAgentMiddleware
        import random
        
        class UserAgentMiddleware(UserAgentMiddleware):
          def __init__(self, user_agent):
              self.user_agent = user_agent

          @classmethod
          def from_crawler(cls, crawler):
              return cls(user_agent=crawler.settings.get('USER_AGENT'))

          def process_request(self, request, spider):
              agent = random.choice(self.user_agent)
              if agent:
                  print(agent)
              request.headers.setdefault('User-Agent', agent)
- 在settings.py中找到DOENLOADER_MIDDLEWARES,把刚才制定的类导入进去
    - DOWNLOADER_MIDDLEWARES = {'lyw.middlewares.UserAgentMiddleware': 543,}

##### IP代理池举例
- 代理购买：芝麻代理，太阳代理，快代理，蚂蚁代理，
- 使用代理
- 实例：需要在settings中注册和写入PROXIES(IP地址列表）
        class ProxyMiddleware(object):
            '''
            设置Proxy
            '''
            def __init__(self, ip):
                self.ip = ip
            @classmethod
            def from_crawler(cls, crawler):
                return cls(ip=crawler.settings.get('PROXIES'))
            def process_request(self, request, spider):
                ip = random.choice(self.ip)
                request.meta['proxy'] = ip
- 注册：DOWNLOADER_MIDDLEWARES = { 'myproject.middlewares.ProxyMiddleware': 544,}

##### cookies的设置举例
- 先获取headers头文件中的cookies，将网站返回的Headers保存到Redis中
- 可以从Redis中读取Cookies并给爬虫换上
- 如果有某网站的100个账号，那么单独写一个程序，持续不断地用Selenium和ChromeDriver或者Selenium 和PhantomJS登录，获取Cookies，并将Cookies存放到Redis中。爬虫每次访问都从Redis中读取一个新的Cookies来进行爬取，就大大降低了被网站发现或者封锁的可能性。
- 这种方式不仅适用于登录，也适用于验证码的处理。
        
        class LoginMiddleware(object):
            def __init__(self):
                self.client = redis.StrictRedis()

            def process_request(self, request, spider):
                if spider.name == 'loginSpider':
                    cookies = json.loads(self.client.lpop('cookies').decode())
                    request.cookies = cookies

### (2)spider middlewares:爬虫中间件
- 处理抓取器的输入和输出
- 方法
    - from_crawler(crawler)：
        - 用来创建你的spider
    - process_spider_input(response, spider)：
        - 在下载器中间件处理完成后，马上要进入某个回调函数parse_xxx()前调用
    - process_spider_output(response, result, spider)
        - 在爬虫运行yield item或者yield scrapy.Request()的时候调用。
        - 该方法处理完成以后，数据如果是item，就会被交给pipeline；如果request，就会被交给调度器，然后下载器中间件才会开始运行。
        - 该方法可以进一步对item或者request做一些修改。这个方法的参数result就爬到的item或者scrapy.Request()。由于yield得到的是一个生成器，所以result可迭代
    - process_spider_exception(response, exception, spider)
        - 当爬虫本身的代码出现了Exception的时候，爬虫中间件的process_spider_exception()方法被调用。
        - 可以返回None，也可以运行yield item语句或者像爬虫的代码一样，使用yield scrapy.Request()发起新的请求。如果运行了yield item或者yield scrapy.Request()，程序就会绕过爬虫里面原有的代码。
        - 可以将遇到的错误的详细信息获取保存到item
    - process_start_requests(start_requests, spider)
        - 当运行到start_requests()的时候
    - spider_opened(spider)

# scrapy整体流程梳理
        1.spider的yeild将request发送给engine
        2.engine对request不做任何处理发送给scheduler
        3.scheduler，生成request交给engine
        4.engine拿到request，通过middleware发送给downloader
        5.downloader在\获取到response之后，又经过middleware发送给engine
        6.engine获取到response之后，返回给spider，spider的parse()方法对获取到的response进行处理，解析出items或者requests
        7.将解析出来的items或者requests发送给engine
        8.engine获取到items或者requests，将items发送给ItemPipeline，将requests发送给scheduler（只有调度器中不存在request时，程序才停止，及时请求失败scrapy也会重新进行请求）

# Scrapy-Redis分布式爬虫组件
- scrapy本身不支持分布式爬虫。想要实现，就需要借助一个组件叫做scrapy-redis，这个组件正是利用了Redis可以分布式的功能，集成到scrapy框架中，使得爬虫可以进行分布式，可以充分的利用资源，更多宽带，同步爬取来提高爬虫的效率
- 优点：
    - 1.可以充分利用多态机器的带宽
    - 2.可以充分利用多态机器的IP地址
    - 2.多台机器效率更高
    
- 分布式爬虫必须要解决的问题：
    - 1.分布式爬虫只是几台机器同时运行，如何保证不同的机器爬取的页面不重复
    - 2.爬完后如何实现数据保存在同一地方
- 安装：pip install redis

# Redis教程
- 概述：
    - Redis是一种支持分布式的nosql数据库，他的数据库是保存在内存中，同时Redis定时把内存数据同步到磁盘，即可以将数据持久化，并且他比memcached支持更多的数据结构（string，list列表（队列和栈），set（集合）， sorted set（有序结合），hash（hash表）），相关参考文档：http://redisdoc.com/index.html

## Redis的使用场景
- 1.登录会话存储，存储在Redis中，与memcached相比，数据不会丢失
- 2.排行版/计数器：比如一些秀场类的节目，经常会有一些前多少名的主播排名，还有一些文章阅读量的技术，或者新浪微博的点赞数等
- 3.作为消息队列，比如celery就是使用Redis作为中间人
- 4.当前在线人数
- 5.一些常用的数据缓存，比如我们的bbs论坛，板块不会经常变化的，但是每次访问首页都要从MySQL中获取，可以在Redis中缓存起来，不用每次请求数据库
- 6.把前200篇文章缓存或者评论缓存
- 7.好友关系，微博的好友关系使用Redis实现
- 8.发布和订阅功能，可以用来做聊天软件

## Redis和memcached的比较
                memcached                       redis
    类型         纯内存数据库                   内存磁盘同步数据库
    数据类型      在定义value时就要固定数据类型           不需要
    虚拟内存      不支持                          支持
    过期策略        支持                         支持
    存储数据安全        不支持                   可以将数据同步到dump.db中
    灾难恢复           不支持                  可以将磁盘中的数据恢复到内存中
    分布式            支持                    主从同步
    订阅与发布         不支持                      支持
    
    

## Redis在Windows系统中的安装与启动
- 1.下载：Redis是不支持Windows操作系统的，所以在:https://github.com/MicrosoftArchive/redis/releases
- 2.启动：命令提示符进入到Redis安装所在的路径然后执行redis-server.exe redis.windows.conf就可以运行
- 3.连接：redis和MySQL以及mongo是一样的，都提供了一个客户端进行连接，输入命名redis-cli就可以连接到redis服务器了（redis的安装路径需要加入到环境变量中）

## 其他机器访问本地redis服务器
- 想要让其他的机器访问本机的redis服务器，那么就要修改redis.conf的配置文件，将bind改成bind [自己的IP地址或者0.0.0.0]，其他机器才能访问
- 注意：bind绑定的是本机网卡的IP地址，而不是想让其他机器连接的IP地址，如果有多块网卡，那么可以绑定多个网卡的IP地址，如果绑定到的是0.0.0.0，意味着其他机器可以通过本机所有的IP地址进行访问

## 对redis的操作
- 对redis的操作有两种方式，第一种是采用redis-cli，第二种是采用编程语言，Python，PHP，Java等
- 1.使用redis-cli对redis进行字符串操作
        启动redis
        service redis-server start
        连接上redis-server
        redis-cli -h [ip] -p [端口]
        添加
        set  key value 比如：set username taotao
        将字符串关联到key，如果key已经持有其他值，set命令就要覆盖旧值，无视其类型，并且默认的过期时间是永久，即永远不会过期
        删除
        del key  如：del username
        设置过期时间
        expire key timeout(单位为秒)
        也可以设置过期时候，一同指定过期时间
        set key value EX timeout
        setex key timeout value
        查看过期时间
        ttl key
        查看当前redis中的所有key
- 2.列表操作
        在列表左边添加元素
        lpush key value
        将值value插入到列表key的表头，如果key不存在，一个空列表就会被创建并执行lpush操作，当key存在但不是列表类型时，将返回一个错误
        在列表右边添加元素
        rpush key value
        将值value插入到列表key的末尾，如果key不存在，同上
        查看列表中的元素
        lrange key start stop
        返回列表key中指定区间内的元素，区间以偏移量start和stop指定如果要左边的第一个到最后一个lrange key 0 -1
        移除列表中的元素
                移除并返回列表key的第一个元素
                lpop key
                移除并返回列表的最后一个元素
                rpop key
                移除并返回列表key的中间元素
                lrem key count value
                将删除这个列表中count个值为value的元素
        指定返回第几个元素
        lindex key index, 将返回key这个列表中，索引为index的这个元素
        获取列表中的元素个数
        llen key
        删除指定的元素
        lren key count value
        根据参数count的值，移除列表中与参数value相等的元素，count的值可以是以下几种：
            - count>0:从表头开始向表尾搜索，移除与value相等的元素，数量为count
            - count<0:从表尾开始向表头搜索，移除与value相等的元素，数量为count的绝对值
            - count=0:移除表中所有与value相等的值

- set集合的操作
        添加元素：
            add set value1 value2... 
        查看元素
        smembers set
        移除元素
        srem set member...
        查看集合中的元素个数
        scard set
        获取多个集合的交集
        sinter set1 set2
        获取多个元素的并集
        sunion set1 set2
        获取多个集合的差集
        sdiff set1 set2
- hash哈希操作
        添加一个新值
        hset key field value：如：hset website baidu baidu.com
        将哈希表key中的域field的值设置为value，如果key不存在，一个新的哈希表被创建并进行hset操作，如果field已经存在于哈希表中，旧值将被覆盖
        获取哈希中的field对应的值
        hget key field
        删除field中的某个field
        hdel key field
        获取某个哈希中所有的field和value
        hgetall key
        获取某个哈希中所有的field
        hkeys key
        获取某个哈希中所有的值
        hvals key
        判断哈希中是否存在某个field
        hexists key field
        获取哈希中总共的键值对
        hlen field
- 事务操作：redis事务可以一次执行多个命令，事务具有以下特征
    - 隔离操作：事务中的所有命令都会序列化，按顺序的执行，不会被其他命令打扰
    - 原子操作：事务中的命令要么全部被执行，要么全部都不执行
        - 开启一个事务：
            multi
            以后执行的所有命令，都在这个事务中执行
        - 执行事务
            exec，会将在multi和exec中的操作一并提交
        - 取消一个事务
            - discard:会将multi后的所有命令取消
    - 监视一个或者多个key：watch key
        - 监视一个或多个key，如果在事务执行之前这个或这些key被其他命令所改动，那么事务将被打断
    - 取消所有key的监视：unwatch
    - 发布或订阅操作
        - 给某个频道发布消息
            - publish channel message
        - 订阅某个频道的消息
            - subscribe channel
