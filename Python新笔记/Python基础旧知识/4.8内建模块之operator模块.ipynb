# operator模块
- operator模块输出一系列对应Python内部操作符的函数
- perator模块也定义了一些广义属性和项目查找的工具。它们常用于为诸如map()、sorted()、itertools. groupby()或其他需要函数作为参数的函数提供参数，该参数为一个高速的字段提取器。
- 函数的映射操作


# 通过公共键值对序列进行排序operator.itemgetter()


```python
from operator import itemgetter
a = [{'name':'liutao', 'age': 23, 'address':"jiangsu"},
    {'name':'Maxiaoxiao', 'age':22, 'address':'Shanghai'},
    {'name':'Jiaojiao', 'age':21, 'address':'Chengdu'}]
sort_by_name = sorted(a, key=itemgetter("name"))
print(sort_by_name)
# itemgetter可以传入多个参数
sort_by_age_name = sorted(a, key=itemgetter('age', 'name'), reverse=True)
print(sort_by_age_name)
# 可以用lambda表达式代替itemgetter，但是itemgetter更快
# sort_by_age_name = sorted(a, key=lambda d: d['age'], reverse=True)
# itemgetter同样适用于max,min函数
print(max(a, key=itemgetter("age")))
```

# 对不原生支持比较操作的对象进行排序operator.attrgetter()

```python
from operator import attrgetter

class User():
    def __init__(self, user_id):
        self.user_id = user_id
        
    def __repr__(self):
        return "User({0})".format(self.user_id)
    
users = [User(3), User(40), User(9)]
# 仍可用lambda对其进行排序
user_sort1 = sorted(users, key=lambda u: u.user_id)
print(user_sort1)
user_sort2 = sorted(users, key=attrgetter("user_id"))
print(user_sort2)
# 同样，attrgetter也可以传入多个参数
```
