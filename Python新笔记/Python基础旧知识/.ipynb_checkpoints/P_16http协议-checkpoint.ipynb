# 基础知识
- 超文本：
    - 包含有超链接和各种多媒体元素标记的文本。这些超文本彼此链接，形成网状，因此又被称为网页，这些链接用url表示，最常见的超文本格式就是超文本标记语言HTML
    
- 超文本传输协议HTTP
    - 是一种按照url提示，将超文本文档从一台主机（web服务器）传输到另一台主机（浏览器）的应用层协议，以实现超链接的功能

- 传输HTTP的具体情况

- 非持久连接，持久性连接
    - 非持久性连接：即浏览器每完成请求一个web文档，就创建一个新的连接，当文档传输完毕后，连接就立刻被释放。HTTP1.0和0.9采用此方式
    - 持久性连接：在一个连接中，可以多次进行请求和相应，服务器在完成文件传输后，并不立刻释放连接，可以持续进行请求，连接保持的时间可以由双方进行协商

- HTTP协议：无状态性：
    - 指同一个客户端（浏览器）第二次访问同一个web服务器上的页面时，服务器无法知道这个客户端曾经访问过，HTTP的无状态性简化了服务器的设计，使其更容易支持大量并发的HTTP请求
    
### HTTP报文结构
- 请求报文
- 返回报文
- 请求报文中的方法
- 响应报文中的状态码
- 首部字段或消息头
- 报文结构实例

#### 浏览器向服务器提供的报文格式
- 报文主要有三部分组成：
    - 请求行
    - 首部行：用来说明浏览器，服务器或报文主体的一些信息
    - 报文主体


#### 服务器向浏览器提供的报文格式


# 响应报文中的状态码
- 1xx：通知信息： 比如100=服务器正在处理客户请求
- 2xx：成功     200 = 请求成功
- 3xx：重定向    301 = 页面改变了位置
- 4xx：客户错误， 403 = 禁止的页面，404=页面未找到
- 5xx：服务器错误 500 = 服务器内部错误，503=以后再试

# 首部行（首部字段或消息头）
- User-Agent：    请求    关于浏览器和它的平台消息
- Accept：       请求    客户能处理的页面的类型
- Accept-Charset   请求    客户可以接受的字符集
- Accep-Encoding    请求    客户能处理的页面编码方法
- Accept-Language   请求   客户能处理的自然语言
- Host：         请求      服务器的DNS名称，从url中提取出来，必须有
- Authoriaztion：   请求    客户的信息凭据列表
- Cookie:         请求    将以前设置的Cookie送回服务器，可用来作为会话信息
- Date：          双向   消息被发送时的日期和时间
- Server：         响应   关于服务器的信息
- Content-Encoding：  响应    内容是如何被编码的
- Content-Length：    响应    以字节计算的页面长度
- Content-Language：   响应    页面所使用的自然语言
- Content-Type：     响应    页面的MIME类型
- Last-Modified：     响应    页面最后被修改的日期和时间，在页面缓存机制中意义重大  
- Location：       响应    指示客户端请求发送给别处，即重定向到另一个url
- Set-Cookie:      响应    服务器希望客户保存一个Cookie

# http代理
- 又称Web缓存或代理服务器，是一种网络实体，能代表浏览器发送HTTP请求，并将最近的一些请求和响应缓存在本地磁盘中，当请求的web页面先前暂存过，则直接将缓存的页面发给客户端（浏览器），无须再次访问Internet
