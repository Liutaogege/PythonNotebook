# Tkinter常用组件
- 按钮组件
        Button      按钮控件；在程序中显示按钮。
        Radiobutton	单选按钮控件；显示一个单选的按钮状态
        Checkbutton	多选框控件；用于在程序中提供多项选择框
        Listbox	    列表框控件；在Listbox窗口小部件是用来显示一个字符串列表给用户
- 文本输入组件
        Entry	    输入控件；用于显示简单的文本内容
        Text	    文本控件；用于显示多行文本
- 标签组件
        Label      标签组件 ，可以显示文本和图片（位图）
        Message    标签组件；用来显示多行文本，与label比较类似
- 菜单组件        
        Menu	    菜单控件；显示菜单栏,下拉菜单和弹出菜单
        Menubutton	 菜单按钮控件，用于显示菜单项。
- 滚动条
        Scale	    滑动组件
        Scrollbar	滚动条控件，当内容超过可视化区域时使用，如列表框。
- 其他组件
        Canvas	    画布控件；显示图形元素如线条或文本
        Frame	    框架控件；在屏幕上显示一个矩形区域，多用来作为容器，将多个组件编组
        Toplevel	容器控件；创建子窗口容器组件，提供一个单独的对话框，和Frame比较类似
        Spinbox	    输入控件；与Entry类似，但是可以指定输入范围值
        PanedWindow	是一个窗口布局管理的插件，可以包含一个或者多个子控件。
        LabelFrame	是一个简单的容器控件。常用与复杂的窗口布局。
        tkMessageBox	用于显示你应用程序的消息框。
- 组件的共同属性   
    Dimension	控件大小；
    Color	控件颜色；
    Font	控件字体；
    Anchor	锚点；
    Relief	控件样式；
    Bitmap	位图；
    Cursor	光标；
### 组件的大致使用步骤
- 1.创建总面板
- 2.创建面板上的各种组件
    - a.指定组建的父组件，即依附关系
    - b.利用相应的属性对组件进行设置
    - c.给组件安排布局
- 3.同步骤2相似，创建多个组件
- 4.最后，启动总面板的消息循环

```python
import tkinter as tk
base = tk.Tk()

# 创建一个标题
base.wm_title("Python")
b = tk.Button(base, text="pickme",background='red') #指定其父组件为base
lb = tk.Label(base, text="button")



# 给相应组件指定布局
b.pack()
lb.pack()

base.mainloop()
```

# Button属性详解
- anchor:设置按钮中文字的对其方式，相对于按钮的衷心位置
- background：设置按钮的背景颜色
- foreground：设置按钮的前景颜色，即文字的颜色
- borderwidth：设置按钮边框宽度
- cursor：设置鼠标点击时触发的样式
- command：设置按钮点击时触发的函数
- bitmap：设置按钮上显示的位图
- font：设置按钮上文本的字体
- width：设置按钮的高度（可控制字符个数）
- height：设置按钮的高度（可控制字符个数）
- state：设置按钮的状态
- image：设置按钮上的图片

# 组件布局
- 控制组件的摆放方式
- 三种布局：
    - pack：按照方位布局
    - place：按照坐标布局
    - grid:网格布局
- pack布局：
    - 最简单，代码量少，依次摆放，默认从上到下
    - 通用使用方式：组件对象.pack（设置，，，，，）
    - 设置内容有
        - side:停靠方式：可选值为LEFT,TOP,RIGHT,TOTTON
        - fill:填充方式，X,Y,BOTH,NONE
        - expand:是否沿着fill的方向扩展其高度或者宽度，YES/NO
        - anchor:设置对齐方式：N,E,S,W,CENTER
        - ipadx:x方向的内边距
        - ipady:y
        - padx：x方向外边界
        - pady：y方向外边界
- grid布局
    - 通用使用方式：组件对象.grid(设置，，，，)
    - 利用row，column编号，都是从0开始
    - sticky：N，E，S，W表示上下左右，用来决定组件从哪开始排列
    - 支持ipadx，padx等参数，跟pack函数一样
    - 支持rowspan，columnspan，表示跨行，跨列数量
    
- place布局：不建议使用
    - 明确方位的摆放
    - 相对位置布局，分为绝对布局和相对布局，绝对布局使用x，y参数
    - 相对布局使用relx，rely，reheight，rewidth

```python
# pack布局案例
import tkinter as tk

base = tk.Tk()

btn1 = tk.Button(base, text='选择', background="red")
btn1.pack(side=tk.LEFT, expand=tk.YES, fill=tk.Y)

btn2 = tk.Button(base, text='删除', background="yellow")
btn2.pack(side=tk.TOP, expand=tk.YES, fill=tk.BOTH) 

btn3 = tk.Button(base, text='撤销', background="blue")
btn3.pack(side=tk.RIGHT, expand=tk.YES, fill=tk.NONE)

btn4 = tk.Button(base, text='复制', background='orange')
btn4.pack(side=tk.LEFT, expand=tk.NO, fill=tk.X)

btn5 = tk.Button(base, text='粘贴', background='gray')
btn5.pack(side=tk.TOP, expand=tk.YES, fill=tk.Y)  # 如果不扩展，fill就起不了作用了

btn6 = tk.Button(base, text='剪切', background='purple')
btn6.pack(anchor=tk.CENTER)

base.mainloop()
```

# 消息机制
- 消息的传递机制
    - 自动发出事件/消息
    - 消息由系统发送到队列
    - 由相关组件进行绑定/设置
    - 后端自动选择感兴趣的事件并作出相应反应
- 消息格式：
    - <[modifier-]---type-[-detail]>
    - <Button-1>:button表示一个按钮事件，1代表的是鼠标左键，2代表中建
    - <KeyPress-A>:表示按键盘A
    - <Control-Shift-KeyPress-A>:表示同时按下control，shift，A三个键位
    - <F1>:F键盘
    - [键位对应名称]
        - 网址：https://infohost.nmt.edu/tcc/help/pubs/tkinter/web/key-names.html

```python
# 模拟消息机制传递案例
import tkinter as tk

def baseLabel(event):
    global base
    print("hehahaha")
    lb = tk.Label(base, text='谢谢点击')
    lb.pack()
    
base = tk.Tk()
lb = tk.Label(base, text="按1键启动我")

# 绑定相应的消息和处理函数
# 自动获取鼠标左键点击，并启动相应的处理函数baselabel
lb.bind("<Button-1>", baseLabel)
lb.pack()

base.mainloop()
```

# tkinter绑定
- bind_all:全局范围的绑定，默认的是全局快捷键，比如F1是帮助文档
- bind_class:接受三个参数，类名，事件，操作
    - w.bind.class("Entry", "<Control-V>", my_paste）
- bind:单独对某一个实例绑定
- unbind：解绑，就只需一个需要解绑的事件参数

### Entry
- 输入框：功能单一
- entry["show"] = "*"，设置遮挡字符

```python
# 示例
import tkinter as tk
import time

def reg():
    name = e1.get()
    pwd = e2.get()
    
    t1 = len(name)
    t2 = len(pwd)
    
    if name == "111" and pwd == "222":
        lb3['text'] = "登录成功"
    else:
        lb3['text'] = '用户名或密码错误'
        
        # 删除输入错误的用户名和密码
        e1.delete(0, t1)
        e2.delete(0, t2)

        
base = tk.Tk()

lb1 = tk.Label(base, text="用户")
lb1.grid(row=1, column=1, stick=tk.W)

e1 = tk.Entry(base)
e1.grid(row=1, column=2, stick=tk.W)

lb2 = tk.Label(base, text="密码")
lb2.grid(row=2, column=1, stick=tk.W)

e2 = tk.Entry(base)
e2.grid(row=2, column=2)
e2['show'] = '*'

btn = tk.Button(base, text="登录", command=reg)
btn.grid(row=3, column=2, stick=tk.E)

lb3 = tk.Label(base, text="")
lb3.grid(row=4, column=10)

base.mainloop()
```

## 菜单
### 1.普通菜单
- 第一个Menu类定义的是parent
- add_command 添加菜单项，如果菜单是顶层菜单，则从左向右添加，否则就是下拉菜单
    - label：指定菜单项名称
    - command:点击后相应的调用函数
    - acceleteor：快捷键
    - underline：指定菜单信息下是否有横线
    - menu：属性制定使用哪一个作为顶级菜单
    

```python
# 普通菜单的代码

import tkinter as tk

baseFrame = tk.Tk()
baseFrame.wm_title("Testing Windows")
menubar = tk.Menu(baseFrame) # 创建一个菜单，但没有名字

for item in ["File", "Edit", "View", "Help"]:
    menubar.add_command(label=item)
    
baseFrame['menu'] = menubar
baseFrame.mainloop()
```

# 级联菜单
- add_cascade：级联菜单，作用是引出后面的菜单
- add_cascade的menu属性：指明把菜单级联到哪个菜单上
- label：名称
- 过程：
    - 1.建立menu实例
    - 2.add_command
    - 3.add_cascade

```python
import tkinter as tk

baseFrame = tk.Tk()

menubar = tk.Menu(baseFrame)
# 子菜单继承主菜单
setting_son_menu = tk.Menu(menubar) 
option_son_menu = tk.Menu(menubar)
view_son_menu = tk.Menu(menubar)

# 构建子菜单内容
for item in ["Copy", "Past", "Cut"]:
    setting_son_menu.add_command(label=item)

for item in ['Edit', 'Quit', 'Creat', 'Delete']:
    option_son_menu.add_command(label=item)

for item in ['Size', 'Zoom', 'Exchange']:
    view_son_menu.add_command(label=item)
    
# 将子菜单级联到主菜单
menubar.add_cascade(label="Options", menu=option_son_menu)
menubar.add_cascade(label="Settings", menu=setting_son_menu)
menubar.add_cascade(label="Views", menu=view_son_menu)
menubar.add_cascade(label="Help")
menubar.add_cascade(label="Escape")

baseFrame['menu'] = menubar
    
baseFrame.mainloop()
```

# 弹出式菜单
- 弹出式菜单也叫上下文菜单
- 实现的大致思路
    - 1.简单菜单并向菜单添加各种功能
    - 2.监听鼠标右键
    - 3.如果右键点击，则根据位置判断弹出
    - 4.调用Menu的pop方法
    
- add_separator:添加分隔符

```python
import tkinter as tk

def makeLabel():
    global baseFrame
    tk.Label(baseFrame, text="生命太短，就用Python，Python使我快乐").pack()
    
baseFrame = tk.Tk()
menubar = tk.Menu(baseFrame)
for x in ["CreatFile", "CopyFile", "Paste"]:
    menubar.add_separator()
    menubar.add_command(label=x)
    
menubar.add_command(label="Editing", command=makeLabel)

# 定义一个事件处理函数，该函数至少要有一个参数，且第一个参数表示的是系统事件
def pop(event):
    # 注意使用event.x和event.x_root的区别
    #menubar.post(event.x, event.y)
    menubar.post(event.x_root, event.y_root)
    
baseFrame.bind("<Button-3>", pop)

baseFrame.mainloop()
```

# canvas画布
- 画布：可以自由在其上面绘制图形
- 绘制对象时，使用create_xxxx, xxxx=对象类型，例如line, rectangle
- 画布的作用是把一定的组件画到画布上显示出来
- 画布所支持的组件：
    - arc
    - bitmap
    - image
    - line
    - oval
    - polygon
    - rectangle
    - text
    - window(组件)
- 每次调用create_xxxx都会返回一个创建的组件的ID，同时也可以用tag属性指定其标签
- 通过调用canvas.move实现一个一次性动作

```python
import tkinter as tk

baseFrame = tk.Tk()
cvs = tk.Canvas(baseFrame, width=300, height=300)
cvs.pack()

cvs.create_line(23,35, 92,90)
cvs.create_text(56,89, text="I love Python")


baseFrame.mainloop()
```

```python
import tkinter as tk
import math

baseFrame = tk.Tk()
w = tk.Canvas(baseFrame, width=400, height=400, background="black")
w.pack()

center_x = 200
center_y = 200
r = 200

points = [
    center_x - int(r * abs(math.sin(2 * math.pi / 5))),
    center_y + int(r * abs(math.cos(2 * math.pi / 5))),
    
    center_x + int(r * abs(math.sin(2 * math.pi / 5))),
    center_y + int(r * abs(math.cos(2 * math.pi / 5))),
    
    center_x - int(r * abs(math.sin(math.pi / 5))),
    center_y - int(r * abs(math.cos(math.pi / 5))),
    
    center_x,
    r + center_y,
    
    center_x + int(r * abs(math.sin(math.pi / 5))),
    center_y - int(r * abs(math.cos(math.pi / 5))),
    
    center_x - int(r * abs(math.sin(2 * math.pi / 5))),
    center_y + int(r * abs(math.cos(2 * math.pi / 5))),
]

w.create_polygon(points, outline="blue", fill="yellow")
w.create_text(200,200, text='五角星')
baseFrame.mainloop()
```

```python
#canvas动画实现
import tkinter as tk

baseFrame = tk.Tk()
cvs = tk.Canvas(baseFrame, width=400, height=400)
cvs.pack()

def btnClick(event):
    global cvs
    cvs.move(id_ball, 12,5)
    cvs.move("fall", 0,5)
    
cvs.bind("<Button-1>", btnClick)

# 创建组件后返回ID
id_ball = cvs.create_oval(20,20, 50,50, fill="yellow")
# 创建组件使用tag属性
cvs.create_text(123,47, fill='red', text="IlovePython", tag="fall")
#创建的时候如果没有指定tag可以利用addtag_withtag添加
# 同类的函数还有addtag_all, addtag_above, addtag_xxx等
id_rectangle = cvs.create_rectangle(56, 78, 173,110, fill='blue')
cvs.addtag_withtag("fall", id_rectangle)

baseFrame.mainloop()
```
