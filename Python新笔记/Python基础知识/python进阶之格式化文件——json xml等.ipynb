# xml（EXtensible Mark Language）：可扩展标记语言
- 标签没有被预定义，具有自我描述性
- 标记语言：语言中使用尖括号括起来的文本字符串标记
- 可扩展：用户可根据自己的需要定义标记
- 例如：
            <from> 
                    自定义标记from
                    在两个标记之间的任何内容都应该与from相关
            </from>
- 符合W3C制定标准

- xml文档的构成
    - 处理指令（可以认为一个文件内只有一个处理指令）
        - 最多只有一行且必须在第一行
        - xml文件声明及相关指令
        - 以xml关键字开头
        - 一般用于声明xml的版本号和采用的编码方式
            - version必须有
            - enconding
    - 根元素：（一个文件内有且只有一个根元素），该元素是其他元素的父亲
    - 子元素：相同层级的子元素成为同胞（兄弟姐妹），所有元素都可以有属性和文本内容
    - 属性：属性值必须加引号，属性值是一个键值对的形式
    - 内容
    - 注释
        - 起说明作用
        - 注释不能嵌套在标签里
        - 只有在注释的开始和结尾使用双短横线
        - 三短横线只能出现在注释的开头而不能用在结尾
             <!--这是注释-->
             <!--这-是-注-释-->
             <!---这-是-注-释-->
- 保留字符的处理：
    - xml中使用的符号可能跟实际符号相冲突，典型的就是左右尖括号
    - 使用实体引用<EntityReference>来表示保留字符
        <score> score&gt;80 </score> #实体引用，表示分数大于80
    - 把含有保留字符的部分放在CDATA块内部，CDATA块把内部信息视为不需要转义
                <![CDATA[
                  select name, age
                   from Stuent
                   where score > 80
                  ]]>
- 常用的需要转义的保留字符和对应实体引用
    - &: &amp;
    - <: &glt;
    - >:&gt;
    - ':&apos;
    - ":&quot;
   - 一共五个，每个实体引用以&开头并且以分号结尾
- xml标签的命名规则
   - Pascal命名法
        - 用单词表示，第一个字母大写
        - 大小写严格区分
        - 配对标签必须一致
    
# 命名空间
- 当两个不同的文档使用相同的元素名时，就会发生命名冲突。
- 为了防止命名冲突
- 解决办法：
    - 使用前缀来避免命名冲突
        <h:table>
        <h:tr>
        <h:td>Apples</h:td>
        <h:td>Bananas</h:td>
        </h:tr>
        在table之前加上各自的前缀
        <f:table>
        <f:name>African Coffee Table</f:name>
        <f:width>80</f:width>
        <f:length>120</f:length>
        </f:table>
        </h:table>
    - xlmns:xml name space的缩写
        <h:table xmlns:h="http://www.w3.org/TR/html4/">
        <h:tr>
        <h:td>Apples</h:td>
        <h:td>Bananas</h:td>
        </h:tr>
        </h:table>
        - 与仅仅使用前缀不同，我们为 <table> 标签添加了一个 xmlns 属性，这样就为前缀赋予了一个与某个命名空间相关联的限定名称。
    
# Python XML解析

## Python对xml文件的解析
- 常见的 XML 编程接口有 DOM 和 SAX，这两种接口处理 XML 文件的方式不同，当然使用场合也不同。
- Python有三种方法解析xml：SAX,DOM,ElementTree

### SAX:(Simple API for xml)
- 基于事件驱动
- 通过在解析XML的过程中触发一个个的事件并调用用户定义的回调函数来处理XML文件。
            利用SAX解析涉及到解析器和事件处理器。解析器负责读取文档，并向事件出处理器发送事件，如元素开始和元素结束事件。而事件处理器负责对事件作出相应，对传递的xml数据进行处理
- 在python中使用sax方式处理xml要先引入xml.sax中的parse函数，还有xml.sax.handler中的ContentHandler。
- 特点：块，流式读取，

#### ContentHandler类方法介绍
- characters(content)方法
    - 调用时机：
        从行开始，遇到标签之前，存在字符，content的值为这些字符串
        从一个标签，遇到下一个标签之前， 存在字符，content 的值为这些字符串。
        从一个标签，遇到行结束符之前，存在字符，content 的值为这些字符串。
        标签可以是开始标签，也可以是结束标签。
- startDocument()方法
    - 文档启动的时候调用
- endDocument（）方法
    - 解析器到达文档结尾的时候调用
- startElement(name, attrs)方法
    - 遇到xml开始标签时调用，name时标签的名字，attrs时标签的属性值字典
- endElement(name)方法
    - 遇到xml结束标签时使用
    
### xml.sax类方法介绍
- make_parser方法：以下方法创建一个新的解析器对象并返回
    - xml.sax.make_parser([parser_list]):parser_list,可选参数，解析器列表
- parser方法：以下方法创建一个SAX解析器并解析xml文档
    - xml.sax.parse(xmlfile, contenthandler[, errorhandler])
    - xmlfile:xml文件名
    - contenthandler：必须是一个ContentHandler的对象
    - errorhandler：如果指定该参数，errorhandler必须是一个SAX Errorhandler对象
- parseString方法：创建一个xml解析器并解析xml字符串
    - xml.sax.parseString(xmlstring, contenthandler[,errorhandler]
    - xmlstring:xml字符串
    
    
### DOM:(Document Object Model)
- 将 XML 数据先缓存在内存中并将其解析成一个树，通过对树的操作来操作读取XML。
- 作用：
    - 可定位浏览xml任何一个节点信息
    - 添加删除相应 内容
- 实现DOM的第三方工具
    - minidom包
        - minidom.parse(filename)：加载读取XML文件，filename也可以是xml代码
        - doc.documentElement：获取XML文档对象，一个xml文档只有一个文档对象
        - node.getAttribute(AttributeName)：获取XML节点属性值
        - node.getElementsByTagName(TagName)：获取XML节点对象集合
        - node.childNodes ：返回所有孩子节点列表。
        - node.childNodes[index].nodeValue：获取XML单个节点值
        - node.firstChild：访问第一个节点，等价于node.childNodes[0]
        - 返回Node节点的xml表示的文本：
            - doc = minidom.parse(filename)
            - doc.toxml(‘UTF-8’)
            - 访问元素属性：
            - node.attributes[tag_name]:a.name就是tag_name，a.value就是属性的值
            - root.nodeName/root.tagName：节点的名称
            - root.nodeValue：节点的值，文本节点才有值，其它节点返回的是None
            - root.nodeType：节点的类型
     - etree包
         - 以树形结构来表示xml
         - 解析器.getroot():得到根节点
         - 解析器.fromstring(字符串)：从字符串获取根节点
         - root.getiterator:得到相应的可迭代的node集合
         - root.iter:递归查询指定的或所有子元素　
         - find(node_name):查找指定node_name的节点，返回一个node
         - root.findall(nodee_name):返回多个node_name的节点
         - root.iterfind()   根据tag查找直接子元素，返回查到的所有元素的生成器
         - root.getchildren()：获取儿子元素
         - node.get()：获得属性值
         - node.tag: node对应的标签名字
         - node.text:node的文本值
         - node.attrib:是node的属性的字典类型的内容
         - 更多方法详见：https://www.cnblogs.com/awakenedy/articles/9305336.html

```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-
 
# 利用sax构建解析器解析movie.xml
import xml.sax
import os 

class MovieHandler( xml.sax.ContentHandler ):
    def __init__(self):
        self.CurrentData = ""
        self.type = ""
        self.format = ""
        self.year = ""
        self.rating = ""
        self.stars = ""
        self.description = ""
  
   # 元素开始事件处理
    def startElement(self, tag, attributes):
        self.CurrentData = tag
        if tag == "movie":
            print( "*****Movie*****")
            title = attributes["title"]
            print("Title:", title)
 
   # 元素结束事件处理
    def endElement(self, tag):
        if self.CurrentData == "type":
            print( "Type:", self.type)
        elif self.CurrentData == "format":
            print( "Format:", self.format)
        elif self.CurrentData == "year":
            print( "Year:", self.year)
        elif self.CurrentData == "rating":
            print( "Rating:", self.rating)
        elif self.CurrentData == "stars":
            print( "Stars:", self.stars)
        elif self.CurrentData == "description":
            print( "Description:", self.description)
        self.CurrentData = ""
 
   # 内容事件处理
    def characters(self, content):
        if self.CurrentData == "type":
            self.type = content
        elif self.CurrentData == "format":
            self.format = content
        elif self.CurrentData == "year":
            self.year = content
        elif self.CurrentData == "rating":
            self.rating = content
        elif self.CurrentData == "stars":
            self.stars = content
        elif self.CurrentData == "description":
            self.description = content
  


if __name__ == "__main__":
   
   # 创建一个 XMLReader
    parser = xml.sax.make_parser()
   # turn off namepsaces
    parser.setFeature(xml.sax.handler.feature_namespaces, 0)
 
   # 重写 ContextHandler
    Handler = MovieHandler()
    parser.setContentHandler( Handler )

    #parser.parse("movie.xml") #需要使用pycharm
    # 更多案例代码看pycharm
```

# JSON
- JavaScript Object Notation(JavaScript对象表示法)
- JSON是存储和交换文本信息的语法，类似XML
- JSON比xml更小，更快，更易解析
- 使用JavaScript语法来描述数据对象
- JSON语法是JavaScript语法的子集。由于这种相似性，无需解析器，JavaScript 程序能够使用内建的 eval() 函数，用 JSON 数据来生成原生的 JavaScript 对象。
    - JSON.parse（）：将一个JSON字符串转化为JavaScript对象
    - JSON.stringify():将JavaScript值转化为JSON字符串
- JSON 数据的书写格式是：
    - 名称/值对，用大括号包裹。如{"name" : "菜鸟教程"}，
    - JSON键必须是字符串
    - JSON值可以是数字，字符串，逻辑值，数组，对象，null
    - 多个键值对用逗号隔开
    - JSON对象可以嵌套
- json和Python格式的对应
    - 字符串：字符串
    - 数字：数字
    - 数组：list，tuple
    - 对象：dict
    - 布尔值(小写)：布尔值
    - null：None
- python中的json包
    - json和Python的相互转换
        - json.dumps()：对数据编码，把Python格式表示成json格式，不需要写入文件
        - json.loads():对数据解码，将json格式转换为Python，不需要写入文件
        - json.dump():把Python格式转化为json格式，同时把内容写入文件
        - json.load()：把json文件内容读入Python
- 参数
    - Skipkeys：默认值是False，如果dict的keys内的数据不是python的基本类型(str,unicode,int,long,float,bool,None)，设置为False时，就会报TypeError的错误。此时设置成True，则会跳过这类key

    - ensure_ascii：默认值True，如果dict内含有non-ASCII的字符，则会类似\uXXXX的显示数据，设置成False后，就能正常显示

    - indent：应该是一个非负的整型，如果是0，或者为空，则一行显示数据，否则会换行且按照indent的数量显示前面的空白，这样打印出来的json数据也叫pretty-printed json

    - separators：分隔符，实际上是(item_separator, dict_separator)的一个元组，默认的就是(',',':')；这表示dictionary内keys之间用“,”隔开，而KEY和value之间用“：”隔开。

    - encoding：默认是UTF-8，设置json数据的编码方式。

    - sort_keys：将数据根据keys的值进行排序。

    - Decode过程，是把json对象转换成python对象的一个过程，常用的两个函数是loads和load函数。区别跟dump和dumps是一样的。

```python
# json举例
import json
student = {
    'name':'liutao',
    'age':23,
    'address':'chengdu'
}
s_json = json.dumps(student)
print(s_json)
print(type(s_json))
print(type(student))
new_s = json.loads(s_json)
print(new_s)
print(type(new_s))
```

# 将字符串字典类型转换为字典类型的三种方法
- 第一种：利用json模块中的json.loads
    - 用法：
         s = "{"name":"liutao"}"
         s_dict = json.loads(s)
    - 存在问题：字符串字典类型如果像这样s = "{'name': 'liu'}"就不行，这和json模块要求的数据样式有关
- 第二种：使用eval
    - 用法：
        s = "{"name":"liutao"}"
        s_dict = eval(s)
    - 存在问题：不安全问题
        - 如果s是有input输入得到的，那我如果输入import(‘os’).system(‘dir’)，user_dict，eval就将会把他当成命令处理而执行
- 第三种：利用 ast模块中的ast.literal_eval
    - 用法：
        s = "{"name":"liutao"}"
        s_dict = ast.literal_eval(s)
    - 该种用法没有以上方法的缺点，建议使用
