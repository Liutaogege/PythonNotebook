```python
# 计算体脂率
class CalTizhi():
    def __init__(self):
        while True:
            try:
                self.sex = int(input("请输入你的性别（女生0，男生1）:"))
                if self.sex != 0 and self.sex != 1:
                    raise ValueError
                self.waist = float(input("请输入你的腰围(cm):"))
                self.weight = float(input("请输入你的体重（Kg)"))
                if self.weight:
                    break
            except:
                print("请输入正确的数字")

    def cal_tizhi(self):
        sex = int(self.sex)
        waist = float(self.waist)
        weight = float(self.weight)
        if int(sex) == 0:
            weight_value = weight * 0.082 + 34.89
        else:
            weight_value = weight * 0.082 + 44.74
        waist_value = waist * 0.74
        tizhi_value = waist_value - weight_value
        tizhi = format(abs((tizhi_value / weight) * 100), ".2f")
        print("你的体脂率是:", tizhi + "%")

    def run(self):
        self.cal_tizhi()

if __name__ == "__main__":
    cal_tizhi = CalTizhi()
    cal_tizhi.run()

```

```python
# 计算水仙花数
def sxh():
    for num in range(100, 1000):
        num_length = len(str(num))
        hundreds_digit = int(num / 100)
        tens_digit = int((num / 10) % 10)
        unions_digit = num % 10
        if hundreds_digit ** num_length + tens_digit ** num_length + unions_digit ** num_length == num:
            yield num
    return "Done"

def resulting():
    results = sxh()
    for res in results:
        print(res)

resulting()
```

```python
# 宠物管理实现
def add_pet():
    ID = input("请输入宠物的id：")
    name = input("请输入宠物的名字:")
    kind = input("请输入宠物的种类:")
    pet = {"ID": ID, "name": name, "kind": kind}
    LIST.append(pet)
    print("添加宠物成功")


def find_pet():
    ID = input("请输入你要查找的宠物的id：")
    for pet in LIST:
        if ID == pet['ID']:
            text = "你查找的宠物信息如下：id:{ID}/name:{name}/kind:{kind}".format(ID=pet['ID'], name=pet['name'], kind=pet['kind'])
            print(text)


def del_pet():
    ID = input("请输入你要删除的宠物的id：")
    is_right = input("你确定要删除该宠物信息吗？(y/n)")
    if is_right == "y" or is_right == "Y":
        for pet in LIST:
            if ID == pet['ID']:
                LIST.remove(pet)
                print("你已成功删除该宠物信息")
                break
    else:
        print("你已取消操作")


def list_pet():
    print("所有的宠物信息如下：")
    for pet in LIST:
        text = "你查找的宠物信息如下：{ID}/{name}/{kind}".format(ID=pet['ID'], name=pet['name'], kind=pet['kind'])
        print(text)


def exit_program():
    with open(FILENAME, "w", encoding="utf-8") as fp:
        lines = []
        for pet in LIST:
            text = "{}/{}/{}".format(pet['ID'], pet['name'], pet['kind'])
            lines.append(text)
        fp.writelines(lines)

# def saving():
#     with open(FILENAME, "w", encoding="utf-8") as fp:
#         for pet in LIST:
#             fp.write(pet['ID'] + '/' + pet['name'] + '/' + pet['kind'] + '\n')

# 宠物管理系统
def main():
    while True:
        print("=" * 30)
        print("1.添加宠物")
        print("2.查找宠物")
        print("3.删除宠物")
        print("4.列出宠物")
        print("5.退出程序")
        print("=" * 30)
        option = input("请输入你的操作:")
        if option == "1":
            add_pet()
        elif option == "2":
            find_pet()
        elif option == "3":
            del_pet()
        elif option == "4":
            list_pet()
        elif option == "5":
            exit_program()
            break
        else:
            print("请输入正确的数字")


if __name__ == "__main__":
    LIST = []
    FILENAME = "pets.txt"
    try:
        with open(FILENAME, "r", encoding='utf-8') as fp:
            for line in fp:
                pet_info = line.split("/")
                pet = {"ID": pet_info[0], "name": pet_info[1], "kind": pet_info[2]}
                LIST.append(pet)
    except:
        fp = open(FILENAME, 'w', encoding='utf-8')
        fp.close()
    main()

```

```python
# 宠物管理用类进行封装
class Pet():
    """
    宠物类
    """
    def __init__(self, pet_id, pet_name, pet_kind, pet_price):
        self.pet_id = pet_id
        self.pet_name = pet_name
        self.pet_kind = pet_kind
        self.pet_price = pet_price

    def pet_foramt(self):
        # 格式化pet的信息
        return "{id}|{name}|{kind}|{price}\n".format(id=self.pet_id, name=self.pet_name,
                                                   kind=self.pet_kind, price=self.pet_price)

    @classmethod
    def pet_beautify(cls, result):
        new_result = result.replace("\n", "")
        return "|".join(new_result.split("|"))


class PetManager():
    """
    宠物管理
    """
    __filename = "petsinfo.txt"
    # 单例设定
    def __new__(cls, *args, **kwargs):
        if not hasattr(cls, "_instance"):
            cls._instance = super(PetManager, cls).__new__(cls, *args, **kwargs)
        return cls._instance

    def add_pet(self):
        while True:
            try:
                pet_id = int(input("请输入宠物的id（数字）:"))
                pet_name = input("请输入宠物的名字:")
                pet_kind = input("请输入宠物的种类:")
                pet_price = float(input("请输入宠物的价格:"))
                if pet_price:
                    break
            except:
                print("请输入正确的格式")
        pet = Pet(pet_id, pet_name, pet_kind, pet_price)
        with open(PetManager.__filename, "a", encoding="utf-8") as fp:
            fp.write(pet.pet_foramt())
        print("恭喜，添加宠物成功")

    def search_pet(self):
        pet_id = input("请输入你要查询的宠物的id：")
        with open(PetManager.__filename, "r", encoding="utf-8") as fp:
            has_not = False
            for line in fp:
                if line.startswith(pet_id+"|"):
                    has_not = True
                    print("宠物信息如下：")
                    print("pet_id | pet_name | pet_kind | pet_price")
                    result = Pet.pet_beautify(line)
                    print(result)
                    break
            if not has_not:
                print("抱歉没有该宠物的信息")

    def list_all_pets(self):
        print("所有的宠物信息如下：")
        with open(PetManager.__filename, "r", encoding="utf-8") as fp:
            print("宠物信息如下：")
            print("pet_id | pet_name | pet_kind | pet_price")
            for line in fp:
                result = Pet.pet_beautify(line)
                print(result)

    def del_pet(self):
        pass


class ManagerApplication():
    """
    应用管理
    """
    def __init__(self, manager):
        self.manager = manager

    def __init_app_info(self):
        while True:
            print("="*40)
            print("0:退出程序")
            print("1:添加宠物")
            print("2:查找宠物")
            print("3:查看所有宠物")
            print("4:删除宠物")
            option = input("请输入你的操作：")
            print("="*40)
            if option == "0":
                break
            elif option == "1":
                self.manager.add_pet()
            elif option == "2":
                self.manager.search_pet()
            elif option == "3":
                self.manager.list_all_pets()
            elif option == "4":
                self.manager.del_pet()
            else:
                print("请输入正确的数字")

    def run(self):
        self.__init_app_info()


def main():
    manager = PetManager()
    app = ManagerApplication(manager)
    app.run()


if __name__ == "__main__":
    main()
```
