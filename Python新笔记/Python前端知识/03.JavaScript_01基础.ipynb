# JavaScript
- js的全称就是JavaScript，它和Java没有半毛钱关系
## 如何使用js代码
- script属性
    - src属性： 用于引入js文件的地址
    - language属性：设置脚本的语言类型（已经废弃）
    - type属性：用于引入文件的MIME类型
    - charset属性：用于设置引入js文件的字符集类型
    - defer属性：延时执行属性，可以使得js在HTML完全加载完毕之后才执行（对外部js文件有效）
    - async属性：异步加载属性，js下载中的VIP，注意，仅改变下载顺序不改变执行顺序
    
## 三种js的使用方式
- 外链式
        script src="文件地址" /script
- 嵌入式
        script
            书写js代码
        /script
- 内联式
        button 事件属性="js代码"> 按钮名称 /button
        
- 推荐格式：
    - 1.js的框架文件，一般使用外链式js
    - 2.局部性的js代码可以使用嵌入式js
    - 3.内联式尽量少用
    

# js的基本语法
- 变量
- 声明规范：
    - var 变量名 = 值;  声明并且赋值
    - var 变量名;  声明但不赋值

# js中的数据类型分类
- 一种类型：对象
    - js中的所有数据都是对象创建的，因为js是一门基于对象的语言
- 两种类型：（简单数据类型和复杂数据类型/ 基本数据类型和引用数据类型）
    - 简单数据类型（基本数据类型）：
        - 常见的数值，整数，小数，字符都是简单数据类型
    - 复杂数据类型（引用数据类型）
        - 对象
- 六种类型：使用数据监测法typeof监测的结果
- undefined  类型
- string  字符串类型
- number 数值类型
- boolean  布尔类型
- function  函数类型
- object  对象类型
- symbol(新增的数据类型)：先不管

### 1.undefined类型
- 只有一个值，就是undefined本身
- 获得undefined的方式有两种：
    - 声明变量但是不赋值，默认值为undefined
    - 为变量直接赋值undefined
    
### 2.string字符串类型
- 声明用单引号或双引号
- 转义字符：
    - \r:表示回到当前行的开头

### 3.number类型
- 整型，浮点型，NaN类型，Infinity
-  NaN类型
    - 获得方式：var value = 7 - "woaini";或者var value = NaN;
    - 尽量避免获得NaN值，它是一个错误
    - NaN的特点
        - 1.不等于任何数值，包括NaN自己，NaN == NaN会得到false
        - 2.NaN具有传染性，和他有联系的运算都是NaN类型
        - 唯一检测NaN的办法是使用isNaN（）功能

### 4.Boolean类型
- 真假，用小写
- undefined, null, 0,-0, NaN, ""都为假

### 5.function类型
- 分为两类：声明函数和表达式函数
        方法一：function calculate(){
            var a = 5
            var b = 6
            console.log(a+b)
            }
        方法二：var calculate = function (){
            var a = 5
            var b = 6
            console.log(a+b)
            }
            
### 6.object类型
- 对象是数据和功能的集合，null也是object类型。相当于python中的字典
    - 注：在js中对象就是类实例，对象原型是父类，没有类这种说法
- 创建一个简单的对象
    - 第一种方法：var Stu = new Object():这是空对象
    - 第二种方式：var Stu = {}; 可以在括号里面加属性和方法
            var Student = {
                "name": "taotao",
                "age": 23,
                "address": "chengdu",
                testing:function(){
                    console.log("学生会考试");
                    return "haha"
                }
            };
            s = Student;
            console.log(s.name);
            console.log(s.testing());
- 对象中的特殊对象null
    - null是对象类型的空指针
    
- 注：undefined和null的区别
    - null表示一个空的值，它和0以及""不同，""表示长度为0的字符串，而null为空
    - 1.undefined是undefined类型，null是对象类型，null是js中的一个关键字， undefined不是js中的关键字，其是一个全局变量，是Global的一个属性,是一种数据类型，undefined类型
    - 2.undefined和null的值是相同的
    - undefined表示不知道自己未来是什么数据类型，null表示自己知道将来是对象，但现在还是空对象，undefined仅仅在判断函数参数是否传递的情况下有用
    - Number(null)得到0，Number(undefined)得到NaN
    - 建议大多数情况下使用null

# 数据类型的转换
- 分类：
    - 显式数据类型转换（强制数据类型转换）
    - 隐式数据类型转换（自动）
    
## 1.强制数据类型转换
### a.Number数据类型转换
#### parseInt：强制将一个字符串类型转换为整型（仅针对字符串类型有效）
- 转换格式：新变量 = parseInt（原变量）
- 转换规则：
    - 1.如果是纯数字字符串，直接转换（取消小数点后面的数值）
    - 2.如果是以数值开头的字符串，将第一个非数值之前的数值部分保留，，删除所有第一个非数值字符串之后的内容，然后将保留的数值部分转换为整型（取消小数点后面的数值）
    - 3.如果不是数值开头的字符串，转换结果为NaN，undefined也转换为NaN
    - 4.boolean类型转换为NaN类型
- 注意：凡是有在字符串数字有加减乘除的计算，就可以强制转换为number类型


#### parseFloat：
- 将一个字符串类型转换为浮点型

#### Number：将任意类型数据转换为number类型
- 转换规则：
    - 新变量= Number（变量）
    - undefined变成NaN
    - string，纯数字就转换成number，非纯数字就NaN
    - Boolean：true1 false0
    - function变成NaN
    - Object:null变成0，其余为NaN
#### 其他方法
- toExponential()	把对象的值转换为指数计数法。
- toFixed(几位小数)	把数字转换为字符串，结果的小数点后有指定位数的数字。
- toPrecision()	把数字格式化为指定的长度。

### b.Boolean类型转换
- 格式：新变量= Boolean（变量）
- 转化规则：
    - 1.Number -> 0/0.0/NaN 都是false
    - 2.String -> 空字符串 ""为false
    - 3.Undefined -> 永远是false
    - 5.Function -> 永远为真
    - 6.Object -> null为false，其余为真
    
### c.String类型转换
- 转换规则：
    - 1.undefined，number，string， Boolean，function，object，类型转变，但内容不变

## 3.数据类型的判断

- 判断Array要使用Array.isArray(arr)；

# 运算
### 运算符
- &&：与运算符
- ||:或运算符
- !:非运算符
- ==：表示值的比较，自动转换数据类型再比较，很多时候，会得到非常奇怪的结果
- ===：数据类型和值的判断，不会自动转换
    - 注：建议坚持使用===比较
    - 且NaN === NaN得到是false，唯一能判断某个值是NaN的方法是isNaN(value)
- ++：自增运算
- --：自减运算
- new:实例化对象运算符
- instanceof：对象类型检测运算符
- delete：删除对象的成员或者删除数组的值

### with语句
- with:用于 绑定作用范围在某个对象当中
- 格斯：
        with (对象){
            该对象的所有属性都可以直接用名字直接访问，不用加对象名了
        }
        with (Student){
            console.log(name)
            console.log(age) 而不用像这样console.log(Student.age)
            }





# 操作字符串的方法
- toUpperCase():全部变成大写
- toLowerCase():全部变成小写
- indexOf():搜索指定字符串出现的位置,没找到就返回-1。注意：若查找的是一个单词，返回的是该单词首字母的位置
        var s = "我爱你"
        console.log(s.indexOf("爱"))
- substring():返回指定索引区的子串
    - s.substring(3, 5),若只有一个数字，类似列表，意思是指取该数直至结束
