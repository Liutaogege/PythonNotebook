## 块标签和行内标签的特征
- 块标签，又称块元素，块级标签，块级元素：
- 行内标签，又称行标签，行级元素，行内元素，内联元素
- 两者区别：
    - 块元素可以设置宽度和高度，行内元素不可以
    - 块状元素不可以和其他元素一行共存，内联元素可以和内联元素仪器共存
    - 块状元素可以包含内联元素，内联元素不可以包含块状元素
    - div标签时块元素的代表，span是行内元素的代表

## 超链接标签
- a标签：属性
    - href属性：用于设置超链接跳转或者打开的文件地址
        - #：当href="#"时或者什么都不写的时候表示回到网页顶部，HTML中#代表本网页
        - ####：表示什么都不做，不要跳转
        - javascript:void(0); :这种写法和####一样，是企业级标准写法。在body中的顶部一般会写，为了让css修饰网页成功
    - target属性：设置超链接的打开方式
        - _self:当前自己页面打开（默认值）
        - _blank:在新的页面打开（在框架集中使用）
        - _top:在页面顶部打开（在框架集中使用）
        - _parent:在父级页面打开（在框架集中使用）
        
- 超链接的三种使用方式：
    - 1.普通超链接
        - 即网址
    - 2.锚点连接
        - 第一步：在需要跳转到的目的地，使用a标签和name属性（或者id属性）定义锚点位置和名称
        - 第二步：在需要点击跳转的位置，使用正常的超链接设置跳转位置，必须以#开头
        - 注意：锚点的设置可以使用包含id名称的其他标签来代替a标签和name属性设定
                <a href="javascript: void(0);" name="top_a">这是顶部标签</a>
                <a href="#top_a">这是链接跳转到另一个链接</a> #相当于实现页面滚动的效果
                下面这种不会跳转
                <h2 name="h2_tag">这是测试啊</h2>
                <a href="#h2_tag">测试能不能跳转到h2标签</a>
        - 若想实现跳转，需要像下面这样。同样id也适用于a标签
                <h2 id="h2_tag">这是测试啊</h2>
                <a href="#h2_tag">测试能不能跳转到h2标签</a>
                
    - 3.邮箱超链接:使用mailto
        - href="mailto:liutao@haha.com"

## 表单标签form
- 用于连接用户和后台服务器的窗口，负责接收用户的数据和操作，将数据提交到后台服务器页面的标签
- form标签： 设定信息即表单范围标签
- 属性如下：
    - action属性：设置当前表单数据提交的地址
    - method属性：设置当前表单提交数据的方式
        - get方式：使用明文传输，提交的数据长度有限
        - post方式：不会明文显示在地址栏中，在head头信息中发送
            - enctype属性：设置当前表单提交数据的编码方式
                - enctype为：application/x-www-from-urlencode时表示他就是url编码（默认值）
                - 为multipart/form-data:表示上传文件必须有值
                - 为text/plain:基本不编码
    - target属性：值和a标签的href属性一样


### form下的内容
- input标签：输入表单标签
    - name:设置当前表单提交数据的名称，必须有
    - value:设置当前表单的默认值或提示信息（推荐作为默认值使用）
    - disable:禁用属性，该数据不会发送给浏览器
    - readonly：只读
    - checked:选择默认选项
    - placeholder：做提示
    - maxlength:输入数据的最大长度
    - type：通过type属性定义不同类型的表单控件，此标签必须要加入name属性，否则提交会出错。type的值为以下：
         - text:普通文本输入框，value属性表示默认值
         - password:密码输入框，value属性表示默认值
         - radio：单选按钮,必须要有name属性和value属性，且一组单选框name值必须相同。因为单选框和多选框无法输入内容，所以value是该复选框提交给后台程序传过去的值，所以必须要有value
         - checkbox：多选按钮,必须指定name属性和value属性，并且一组多选框name属性的值要相同
             - checked：默认选中
         
         - file：文件上传选框，一般不写value
         - textarea:文本域，不需要input引入
         - submit:提交按钮，name不写value设置当前按钮提交的名称，和buttton提交效果一样。要放在表单域中才能正确提交
         - reset：重置按钮，name可以不写，value为按钮名称
         - hidden：隐藏域，要和表单一起提交的信息，不需要用户看见，name和value必须有
- select下拉列表框，必须要配合option标签来使用
    - select属性
        - multiple:设定可以多选择几个值，按crtl键进行选择
        - size:表示下拉每次呈现几个选择
        - disabled：只要有disabled属性，浏览器就不会帮你发 送这个数据
    - option选择项标签，属性：
        - selected:默认选择的值
        - value：提交的值
            - 注意：如果option标签中不使用value属性，那么提交的时候会将option标签的text提交上去，不推荐这样
- textarea 多行文本
    -textarea name="" id="" cols="30" rows="10"
    - 没有value属性，可在标签中间写，起提示作用
- button 标签 多功能按钮标签
    - type属性：
        - 值可以为submit,reset,button(普通按钮)
    - 小技巧：将文字和表格绑定起来
        - 使用label标签
        - <label for="man">男</label>：man是某个input标签的id

# 框架和框架集
- frameset是给后台布局使用的，前台不用
- 框架集：当前页面完全由其他页面组成，当前页面没有自己的内容。该框架集不能与html，head等使用，是相冲突的
- 1、frameset标签：框架集合标签
    - 每一个页面中包含多个子页面，从而组成一个框架集合页面
    - 属性：
        - cols：纵向切割页面属性，值为多个数值组成，使用逗号分隔
                cols="40%,60%"纵向切割，左侧40%，右侧60%
        - rows：横向切割页面属性，值为多个数值组成，使用逗号分隔
                rows="20%,80%"横向分割，顶部20%，底部80%
        - frameborder：设置框架集合是否使用边框
            - 1使用边框（默认使用），0不适用
        - border：设置框架集合边框的粗细
            - 值为像素值，不需要单位。只有frameborder值为1时才设置border属性
- frameset下的frame标签：引入框架页面标签
    - 在框架集合页面用于引入其他子页面使用。每个子页面是独立的，可以独立浏览
    - 属性
        - src属性：设置引入页面的地址
        - noresize属性：禁止拖动页面属性
        - scrolling属性：设置是否允许出现滚动条
            - auto值：自动检测，如果内容超出高度就出现滚动条，否则就不出现
            - no值：一直都不出
            - yes值：一直都有滚动条，但不一定有滑块
        - 注意：在框架集合中，不允许出现body标签或者body标签中的内容，否则就会出错。说白了就是不能出现text内容
        
- 2、iframe标签，可以和html，head一起使用，为自己的HTML页面引入其他页面
    - 属性
        - src属性：设置引入页面的地址
        - width:设置引入页面的宽度
        - height：引入页面的高度
        - noresize属性：禁止拖动页面属性
        - scrolling属性：设置是否允许出现滚动条
            - auto值：自动检测，如果内容超出高度就出现滚动条，否则就不出现
            - no值：一直都不出
            - yes值：一直都有滚动条，但不一定有滑块

# head头中的主要标签
- title标签
    - 用于定义当前页面的标题，会在浏览器的标签栏中显示
    - title对网站的SEO优化有一定的影响
    - 首页：一般设置为网站名称即可
    - 内容页：当前页面的主要标题，频道等名称
    - 频道列表页：当前频道名称
    - 注意：title中可以出现当前网站的关键字，但是不要太多，不要重复，要精简
- meta标签：对于浏览器进行设置的标签
    - 1.设置浏览器的声明字符集
        - meta http-equiv="content_type" content="text/html";charset="utf-8":这是html4的书写方式，HTML5直接就写一个charset就行
    - 2.设置浏览器的刷新和跳转方式
        - meta http-equiv="refresh" content="时间;url=地址" />
    - 3.为搜索引擎进行的信息设置
        - a.设置关键字
            - meta name="keywords" content="关键字,关键字;.."
            - 注意：关键字之间必须使用英文逗号分隔，字数尽量保持在60-80之间
        - b.设置网址描述信息：即搜索引擎的页面信息介绍
            - meta name="description" content="当前网站的描述，要使用通顺的语言"，140词左右
        - c.设定作者
            - meta name="author" content="作者名字"
        - d.设置开发工具
            - meta name="generator" content="开发工具信息"
        - e.设置网站修订时间及作者
            - meta name="revised" content="作者,时间"
- base标签
    - 基准连接标签
    - 属性
        - href属性：设置基准连接属性
        - 注意：为页面中所有的相对链接添加href属性的值，组成一个新的绝对路径，对绝对路径不起作用
        - target属性：设置连接的打开方式
            - 注意：为页面中所有没有设置打开方式的链接，添加打开方式，主要设置了打开方式，就不会受到了该属性的影响
            
- link标签：
    - 用于在页面中引入css文件标签
    - 属性
        - href：设置引入css文件地址
        - type：设置引入time的MIME类型，MIME是浏览器区分文件的标识，类似于文件后缀
        - rel：设置引入文本和当前页面之间的关系，stylesheet样式表单
        - 注意：rel属性和src属性天生冲突不能共存，所以link采用href来引入文件地址。引入css文件时，必须要把rel设置为stylesheet，type="text/css"
        - link的其他用法：
            - 设置标签栏的图标,图片要为ico格式
                - link rel="icon"/logo.ico type="image/x-iicon"

## XHTML规范
- 1.所有标签名必须使用小写字母
- 2.所有标签必须闭合，单标签自封闭
- 3.所有属性必须有值，单属性值为属性名disabled
- 4.所有属性值必须使用双引号
- 5.所有标签必须合理嵌套
- 6.所有特殊字符必须使用实体字符表示
- 7.所有img标签必须加alt属性
- 8.所有注释内容中不允许出现-等信息，如果需要使用-，可以使用=代替
