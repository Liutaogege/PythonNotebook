# BOM：浏览器对象模型
- BOM:browser object model浏览器对象模型
- 作用：
    - 1.用于连接js语言和浏览器行为的一个中间结构
    - 2.js可以通过操作BOM实现浏览器对应的行为变化
    - 注：如果浏览器是空调的话，BOM就是遥控器
- 本质：
    - BOM是js可以操作的内容，所以他的本质上是对象,BOM在js中是一个对象。window对象就是BOM本身
    
# window对象分析
- window对象代表浏览器本身（BOM浏览器对象模型）：
    - document：表示浏览器的文档部分（文档对象）DOM
    - location  表示浏览器位置相关的对象
    - frames   表示浏览器框架集相关信息的对象
    - history  表示浏览器历史的信息对象
    - screen   表示浏览器屏幕信息对象
    - navigator  表示浏览器信息对象
    - 注意：以上这些window子对象内容已经是存在的变量，不可当成变量再使用

### window对象的自身功能
- 不属于以上六大对象的属性和方法，在六大子对象之外，还有一些其他的对象方法
- window.innerWidth:表示浏览器视窗的宽度
- window.innerHeight：表示浏览器视窗的高度

### 浏览器位置信息属性
- 新Firefox浏览器及其他浏览器识别属性
    - screenLeft：浏览器距离屏幕左边的距离
    - screenTop：
- 老版firefox浏览器
    - screenX:浏览器距离屏幕左边的距离
    - screenY:浏览器距离屏幕顶端的距离

### 浏览器窗口大小调整方法
- resizeTo(width, height)将浏览器设置为指定的宽度和高度
- reesizeBy(水平长度，垂直高度):将浏览器在当前的宽高的基础上增加或者减少指定长度的宽高
    - 如果长度为正数，则增加，否则减少
- 注意：以上方法必须在单标签浏览器中使用

### 浏览器位置大小调整方法
- moveTo(水平坐标， 垂直坐标):将浏览器移动到指定的坐标
- moveBy(水平长度，垂直长度)将浏览器在当前位置基础上移动指定的长度
### 定时执行
- setTimeout()：设置定时执行操作,注意时间是以ms为单位的，有一个返回值是数字
    - 格式：定时变量 = setTimeout(回调函数，时间)
    - 方法1：setTimeout（函数名，时间）
    - 方法2：setTimeout（匿名函数，时间）
    - 方法3：setTimeout（调用函数的字符串，时间）
    - 实例：
        var flag = window.setTimeout(function(){
                    console.log("哈哈我终于出来了");
                }, 3000
                )
        flag是他的一个标识，是数字
    
### 循环执行
- setInterval()循环执行方法，有一个返回值，是数字
- 格式：循环变量= setInterval（回调函数，时间）
- 方法1：setInterval（函数名，时间）
- 方法2：setInterval（匿名函数，时间）
- 方法3：setInterval（“调用函数的字符串”,时间）
### 清除定时执行
- clearTimeout（flag）取消定时执行的操作
    - flag就是setTimeout返回的那个值，只要把这个值传入，setTimeout就会停止执行
- 格式；clearTimeout(定时变量)。
- 不能直接向下面一样在里面把clearTimeout加进去，会立即停止执行的。在外部添加个按钮来控制

### 清除循环执行
- clearInterval(flag):取消循环执行的操作
- 实例：
        var flag = window.setInterval(
                function(){
                    for (var i=0;i<10;i++){
                        console.log("哈哈我终于出来了", i);
                    }
                    if (i===10){
                        console.log("不好意思，停止循环了");
                        window.clearInterval(flag);
                    }
            }, 1000
            )
            
### 浏览器的三种弹窗

# DOM文档对象模型
- 是js操作文档的一个中间接口
- DOM操作的范围就是视窗区域内，也就是浏览器的白色区域，专门用于显示HTML body相关内容的位置
- DOM的本质就是window对象中的一个子对象：document对象

#### 概念
- 节点：页面中的所有内容都是节点，无论是标签，文字，声明，注释，js都是节点
- 节点层次：节点之间的关系就是节点层次

# Node对象——节点对象
- 在js中叫做Node对象，在DOM中叫做Node节点
- Node是一个构造方法，和Number，Boolean是一样的
- 在js中所有内容都是对象创建的
- 在DOM中所有的内容都是节点创建的，节点是对象创建的

## 1.Node对象的基本属性

### 描述属性
#### nodeType节点类型（必须要记住它们的数字）
- 节点类型是用来表示内容创建使用的节点，重点记忆以下几个
    - ELEMENT_NODE  元素节点  1  相当于HTML中的标签
    - ATTRIBUTE_NODE  属性节点 2  相当于标签的属性
    - TEXT_NODE  文本节点  3  相当于标签下的文本内容
    - DOCUMENT_NODE  文档节点 9  文档节点  document对象
#### 节点常用属性
- 节点名称：节点对象.nodeName
- 节点类型：节点对象.nodeType，是数字表示
- 节点值： 节点对象.nodeValue(注；元素节点的该属性是null)
- childNodes:获取该节点的所有子节点，返回一个列表：
    - 格式：节点对象.childNodes
    - 对于这个属性来说，文本也算是一个子节点，通过节点对象.childNodes[index]就可以访问任意一个子节点，若子节点是文本，则直接返回文本内容，若仍是节点对象，返回该节点对象，可以对节点进行操作
    - 不可见对象仍被处理为一个节点对象，比如换行符
    - childNodes.length：返回子节点对象的长度
    
#### 关系属性
- firstChild:获取元素/文档节点的第一个子节点
- lastChild
- childNodes获取指定元素的节点集合列表
- parentNode获取当前节点的父节点
- previousSibling获取指定节点的上一个兄弟节点
- nextSibling获取指定节点的下一个兄弟节点

## 2.Node对象的基本方法
- 注意：这些方法是给父节点使用的，对节点的操作先要获取父节点，通过父节点来进行增删
- appendChild:在父节点的子节点列表的最后插入定制的节点
    - 格式：父节点.appendChild(要插入的节点)
- insertBefore在父节点的子节点列表的指定子节点之前插入新的节点
    - 格式：父节点.insertBefore(要插入的节点， 定位的节点)
- replaceChild使用新节点替换指定的节点
    - 格式：父节点.replaceChild(新节点，被替换的节点)
- removeChild:移除指定的节点
    - 格式；父节点.removeChild(被移除的子节点)
    - 删除后的节点虽然不在文档树中，但其实还在内存中，可以随时再次被添加到别的位置
    - 节点删除后，父节点.children的长度改变，index也会改变
- cloneNode：克隆子节点
    - 格式1:节点.cloneNode():仅仅复制当前元素的节点本身
    - 格式2：节点.cloneNode(true):克隆当前节电及当前节点的内部所有节点
- 注：js对id有唯一性要求，js中的任一一个id名称都会在浏览器解析页面的时候变成一个变量，若某个标签的id为IDD，可通过console.log(IDD)访问

## 3.元素节点
- 元素节点：是创建一切元素的基本节点

### 元素节点属性
- attributes:获取当前元素的所有属性节点集合
- tagName：获取当前元素节点的标签名称（值为大写字母）

### 元素节点方法
- 以下三个方法都是操作元素节点属性的方法
    - getAttribute():
        - 作用：获取元素节点的属性值
        - 格式：元素节点.getAttribute(“属性名”)
    - setAttribute():
        - 作用：获取元素节点的属性值
        - 格式：元素节点.setAttribute("属性名"， vaule)
    - removeAttribute():
        - 作用：移除指定的属性
        - 格式：元素节点.removeAttribute("属性名")

        
## 4.HTML元素节点
- 是在元素节点的基础上针对HTML文档专门设计的一套操作HTML元素节点的方法和属性，更加快捷和方便
- HTMLDOM的相关转换规则
    - 1.HTML标签所具备的属性会变成js对象的成员属性直接使用
    - 2.class属性在转换之后会变成对象的成员属性className（因为class是保留字之一）
    - 3.style属性转换之后会得到一个所有css样式和值组成的集合对象（浏览器为用户操作提前解析了css）
    - 4.在style属性对象访问指定的css样式时，如果样式名称带有-，需要变更为：去掉-，将-后面的第一个字母大写
    

## 5.属性节点
- 页面中任何元素的属性都是使用属性节点创建
- 属性节点一般都是在元素内部出现，单独脱离了元素无法生效
- 成员方法：
    - 以下三个方法都是操作元素节点的属性节点
        - getNamedItem()获取指定名称的属性节点
            - 格式：元素节点.attributes.getNamedItem(属性名)
        - setNamedItem()设置或者添加指定的属性节点
            - 格式：元素节点.attributes.setNamedItem(属性节点)
        - removeNamedItem()移除指定名称的属性节点
            - 格式：元素节点.attributes.removeNamedItem(属性名)

## 6.文本节点textNode
- 在HTML中一个页面中的文字都是由浏览器中的文本节点创建而成的
- 文本节点好像只能通过子节点进行查找

### 成员属性
- length：获取一个文本节点的字符个数
- data： 表示文本节点的文本内容字符串

### 成员方法
- spiltText：分开文本节点
    - 格式：文本节点.spiltText(切割位置)
    - 分开文本节点的意义是为了添加其他节点
- appendDate 在文本节点的最后插入字符串
    - 格式：文本节点.appendData(插入字符串)
- insertData 在文本节点的指定位置插入字符串
    - 格式：文本节点.insertData(位置， 要插入的字符串)
- replaceData 在文本节点中替换指定的字符串
    - 格式：文本节点.replaceDate(开始位置，要替换的字符串长度， 替换的字符串)
- substringData  在文本节点中截取指定长度的字符串
    - 格式：文本节点.substringData(指定位置, 截取长度）
- deleteData：在文本节点中删除指定的字符串
    - 格式：文本节点.deleteData(指定位置，指定长度)
    
- 提示：
    - 所有文本节点的操作方法都可以使用字符串代替，文本节点提供的操作文本内容的方法比较短，直接使用文本节点.文本节点方法，文本节点使用字符串操作相对麻烦，需要使用文本节点.nodeValue.字符串方法()。文本节点的修改大部分直接映射到节点当中，而字符串的修改就必须重新赋值为节点的nodeValue，才可以映射到节点当中
- 很少用文本节点的方法来对文本节点进行操作


## 7.文档节点document
- 在任何的文档中，文档节点都是不能创建的，是浏览器代为创建的节点类型
- 文档节点不需要创建，也不需要获取，固定名称就是document

### 成员属性
- body：提供在文档对象中直接访问body元素节点的快捷属性
    - document.body
- head:直接访问head元素节点
- title直接获取页面的title值
    - 注意：该属性获取或者设置title标签的内容而非title节点
- cookie：使用js获取当前浏览器中存储的cookie中的信息字符串
    - cookie一般是由服务器进行添加的，js中可以读取当然也可以设置
    - 注意：访问cookie必须使用相同的网址
    - cookie因为可以通过js获取，为确保安全，服务器端在设置cookie时，应坚持使用httpOnly
- domain：获取当前文档的域名地址，本地返回localhost
- URL：获取当前文档的完整地址信息
- lastModified:获取当前文档的最后一次修改时间
- documentElement:表示当前页面的HTML元素节点
    - 注意：低版本有些浏览器不支持
- doctype：获取当前页面的dtd声明节点，即最开始的那一句话

### 成员方法
#### 节点获取类
- getElementById()通过元素id获取指定id的元素节点
    - document.getElement("id值")
- getElementByName()通过name属性获取指定的节点集合
    - 返回的是数组
- getElementByTagName()通过标签名获取指定名称的元素节点集合
- nextElementSibling:获取下一个兄弟元素
- previousElementSibling:上一个兄弟元素
- 以下是新版本中的方法
- querySelector():通过css选择器进行元素节点的选取，该方法永远返回第一个符合条件的元素
    - 格式1：document.querySelector(css选择器)
        - 在文档中查找所有符合css选择器的元素，并返回第一个符合的元素
    - 格式2：元素节点.querySelector(css选择器)
        - 在指定的元素中查找....
- querySelectorAll()
- getElementsByClassName():通过class名字查找符合的元素节点，返回数组
    - 也是有两种格式，同上面一样

#### 文档流操作方法
- open()打开一个新的文档流
    - document.open():
    - 注意：打开一个新的文档相当于清空当前页面，准备重新添加内容
- write()在当前文档中输出指定的内容
    - document.write("字符串")
    - write可以直接写入一个文件，而这个文件可以放在服务器端，这样就可以防止被爬虫
- writeln()和write一样，只是在输出内容的结尾添加\n
- close()关闭打开的文档流，就算不用页面文档流也会自动关闭

#### 文档对象中节点创建方法
- createElement()创建指定名称的元素节点
    - document.createElement(标签名)
- createAttribute()创建指定类型的属性节点
- createTextNode()创建指定内容的文本节点
- 以上所有操作均是DOM1标准中的操作，在所有的浏览器中都可以使用
- 以下为HTML5扩展的
- innerHTML:获取或者设置元素节点汇总的所有内容
    - 该属性可以解析HTML标签
- innerText获取或者设置元素节点中的所有文本内容
    - 注意：该属性不可以解析HTML标签
- textContent，和innerText类似，区别是textContent返回隐藏元素的文本比如换行符号
- outerHTML获取或者设置元素节点内部的内容，包括当前标签本身,都是字符串
    - 效果和innerHTML类似，只是包含当前标签而已
    - 在设置操作时，outerHTML的内容会完全替换当前标签，包含当前标签本身
- outerText：获取或者设置元素节点内部的内容，包含当前标签本身（仅保留文字部分）
    - 效果和innerText效果类似，只是包含标签
    - 设置时，内容会直接替换当前标签
- childElementCount：获取当前元素节点的元素子节点个数
- children：获取当前节点的所有元素节点的集合
- classlist：获取当前节点额class所有的值的集合
- firstElementChild：返回元素节点中第一个元素子节点
- lastElementChild：返回元素节点中的最后一个元素子节点
- nextElementSibling：返回当前节点的下一个元素子节点
- previousElementSibling：获取当前节点的上一个元素子节点
- parentElement:获取当前节点的父元素节点，效果和parentNode一致
- localName：获取当前节点的名称，tagName获取大写之，localName获取小写值
