# css基本语法和使用
- css：层叠样式表，用来规定HTML文档表现的形式，说白了就是修饰HTML标签
- CSS页面引入方法
    - 外链式，在head标签中通过link标签引入CSS文件，推荐使用
        - 在CSS文件中写
                div{
                    width: 200px;
                    height: 200px;
                    background-color: red;}
    - 导入式：
            style
                @import url("css文件地址")
            /style

    - 嵌入式，通过在head中加入style标签来写CSS
            style
                div{width:200px;height:200px;background-color:yellow;}
            /style
    - 内联式：加上style属性来设置css，不建议使用内联式
    - 三者的优先级顺序，离设置对象最近的优先级最高
- 推荐使用方式：
    - 1.单独使用外链式时，更加有利于SEO的优化（个人小型站点适用）
    - 2.单独使用嵌入式，一般情况下常见大型新闻类的网站
    - 3.混合使用嵌入式与外链式，将页面的结构css样式写入嵌入式部分，细节样式部分用外链式写
    
## css的书写规范
         选择器{
            css属性名: css属性值;
            css属性名: css属性值;
        }
- 选择器：用于选择指定类型的标签
- css属性名：用于设置需要修改的样式名称
- css属性值：用于设置属性的结果类型

    
# css选择器
- 1.标签选择器
    - 此种选择器影响范围大，建议尽量应用在层级选择器中
- 2.id选择器(#id名）
    - 通过id来选择元素，元素的id名称不能重复，所以一个样式设置项只能对应于页面上一个元素，不能复用，id名一般给程序使用，所以不推荐使用id作为选择器
- 3.类选择器（.class名）
    - 通过类名来选择元素，一个类可应用于多个元素，一个元素也可以使用多个类，应用灵活，可复用，是css应用最多的一种选择器
- 优先级顺序：
    - 一般来说：id>class>标签
    - 若都写到了id层级，那么层数越多，优先级越高
    - 但是可以在写的时候末尾加上!important，表示这个比id选择器都重要，这个优先级最高

- 4.层级选择器
    - 主要应用在选择父元素的子元素，或者子元素下面的子元素，可与标签元素结合使用，减少命名，同时也可以通过层级，防止命名冲突。可以和前三种选择器配合使用
       使用举例：父元素+空格+子元素
           如： .father div{}
- 5.组选择器，可以给多个元素设置相同的格式
          .box1,.box2,.box3{width:200px;height:200px;} 将他们的高和宽设置的一样
- 6.伪类选择器
- 伪类选择器选中的不是元素本身，而是元素的某一个状态或者某一部分内容
    - a标签的设置：
        - a:link 设置正常连接的颜色
        - a:hover  设置鼠标经过时的样式
        - a:active  设置鼠标点击时的样式
        - a:visited  设置连接访问过后的样式
    - 其他标签的设置：
        - tr:hover设置单元格行标签鼠标经过的样式
        - 其他标签可以使用鼠标经过样式hover，但是不能使用a标签中其他的标签
        - 选择器:first-letter 选择元素内部第一个进行样式设置
        - 选择器:first-line  选择元素内部第一行进行样式设置
        - 选择器:after  选中元素内容的后面的部分进行样式或者内容的添加
            - 该元素不会在浏览器浏览时不会被选择
        - 选择器:before  选中元素的内容的前面部分进行样式或内容的添加
                .box:hover{width:200px;},.box:after{content:"aini";}
- 7.通用选择器
    - 选择所有，*表示所有标签
            *{
                margin:0;padding:0;list-style:none
            } 实现样式清零
