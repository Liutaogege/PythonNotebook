# JavaScript
- js的全称就是JavaScript，它和Java没有半毛钱关系
## 如何使用js代码
- script属性
    - src属性： 用于引入js文件的地址
    - language属性：设置脚本的语言类型（已经废弃）
    - type属性：用于引入文件的MIME类型
    - charset属性：用于设置引入js文件的字符集类型
    - defer属性：延时执行属性，可以使得js在HTML完全加载完毕之后才执行（对外部js文件有效）
    - async属性：异步加载属性，js下载中的VIP，注意，仅改变下载顺序不改变执行顺序
    
## 三种js的使用方式
- 外链式
        script src="文件地址" /script
- 嵌入式
        script
            书写js代码
        /script
- 内联式
        button 事件属性="js代码"> 按钮名称 /button
        
- 推荐格式：
    - 1.js的框架文件，一般使用外链式js
    - 2.局部性的js代码可以使用嵌入式js
    - 3.内联式尽量少用

# 数据类型的转换
- 分类：
    - 显式数据类型转换（强制数据类型转换）
    - 隐式数据类型转换（自动）
    
## 强制数据类型转换
### Number数据类型转换
- parseInt：强制将一个字符串类型转换为整型（仅针对字符串类型有效）
- 转换格式：新变量 = parseInt（原变量）
- 转换规则：
    - 1.如果是纯数字字符串，直接转换（取消小数点后面的数值）
    - 2.如果是以数值开头的字符串，将第一个非数值之前的数值部分保留，，删除所有第一个非数值字符串之后的内容，然后将保留的数值部分转换为整型（取消小数点后面的数值）
    - 3.如果不是数值开头的字符串，转换结果为NaN，undefined也转换为NaN
    - 4.boolean类型转换为NaN类型
#### parseFloat：
- 将一个字符串类型转换为浮点型
#### Number：将任意类型数据转换为number类型
- 转换规则：
    - 新变量= Number（变量）
    - undefined变成NaN
    - string，纯数字就转换成number，非纯数字就NaN
    - Boolean：true1 false0
    - function变成NaN
    - Object:null变成0，其余为NaN
### Boolean类型转换
- 格式：新变量= Boolean（变量）
- 转化规则：
    - 1.Number -> 0/0.0/NaN 都是false
    - 2.String -> 空字符串 ""为false
    - 3.Undefined -> 永远是false
    - 5.Function -> 永远为真
    - 6.Object -> null为false，其余为真
    
### String类型转换
- 转换规则：
    - 1.undefined，number，string， Boolean，function，object，类型转变，但内容不变

## 自执行函数
- 一次性函数，自动执行
    - 使用函数局部作用域的特征分割代码区域防止多段代码之间的变量重名和冲突问题
- 如果直接写出
        function(){};会报错
- 而自执行函数就是给他整体加个括号,它就会变成一个函数，再在后面加个括号就能自动执行
        (function(){};)()
- js中没有命名空间的概念，可以使用自执行函数来代替（命名空间就是防止命名冲突）

## 递归函数的改进

# object对象
- object对象是js中所有对象的原型，任何对象都是该对象创建的。在js中没有区分类和实例的概念。所有的对象都是实例，继承关系无非就是把一个对象的原型指向另一个对象。可以认为对象原型就是父类的意思。而object就是所有对象的父亲
- 属性：
    - prototype：返回当前对象的对象原型
        - 访问对象原型：构造方法名.prototype
        - 访问对象原型：对象实例.__proto__  (有双下划线)
    - constructor:返回当前对象的构造函数方法
        - 访问对象构造方法：对象实例.constructor
- 创建对象：
    - 第一种：obj = new Object()
    - 第二种：使用对象字面量创建对象： obj = {};
    - 示例
            var xiaoming = new Object();
            //小明是对象实例（即类实例），Object就是生产对象实例的对象原型（即类）
            xiaoming.name = "xiaoming";
            xiaoming.age = "18";
            xiaoming.running = function(){
                console.log("小明跑的很快")
            }
            //实例对象通过constructor查看构造方法
            console.log(xiaoming.constuctor)
            // 通过prototype查看对象原型
            console.log(Object.prototype);
            console.log(xiaoming.__proto__) //这个得到的结果和第一个一样
        
- 方法：以下方法不用主动调用，会自动调用
    - toString()：返回当前对象的字符串表示格式
    - toLocaleString()：返回当前对象的本地字符串表示格式
    - toString和toLocaleString的区别
        - 1.大多数情况下一样
        - 2.针对特定对象不同
            - 比如：Date()对象，
                    var time = new Date();
                    console.log(time.toString());
                    console.log(time.toLocaleString());
                    consoel.log(time.valueOf()):得到是时间戳
        
    - valueOf()： 返回当前对象的本身值
    
### 查找对象的对象原型方法（即父类）
- 第一种：
    - 1.先查找该对象的构造方法
        - con = obj.constructor
    - 2.再通过构造方法查找对象原型
        - con.prototype
- 第二种：
    - 直接访问对象原型
        - obj.__proto__(该方法只对部分浏览器有效）

# Array对象原型
- 数组就是一组数据
- 在js中没有数组这种数据类型，数组是数组对象创建得到
- index不像python，它不可以为负数
- 数组创建方法：
    - 方法一：数组字面量：
        - var arr = [];若括号中啥也没有，表示空数组，一个纯数字表示数组长度，放入内容就是指定内容了
    - 方法二：使用Array对象原型创建
        - var arr = new Array()
        - 若Array中只是一个纯数字，表示该数组的长度，若放入内容，就是内容了，若是空的，就是空数组
- 添加数组元素
    - 数组变量[index] = 值
- 删除数组元素
    - delete 数组变量[index];删除指定元素，但是删除不了键，所以数组长度不变。若想彻底删除某个元素，用splice这个方法
- 遍历数组元素：
    - 遍历数组操作就是逐个获取使用数组中每一个元素的操作
    - 可以使用计数循环for进行数组的遍历操作
- 二维数组元素
    - 嵌套一维数组
- 方法：
    - 拼接：concat():old_arr.concat(new_arr);和用加法效果是一样的
    - pop():删除最后一个元素，直接改变原有数组，返回值是被删除的元素
    - join():使用字符串将数组元素连接成一个字符串，返回值是字符串
    - push():在数组的结尾处添加元素，直接改变原有数组，返回值是新数组的长度
    - shift():在数组开头移除一个元素，直接改变原有数组
    - unshift():在数组开头添加一个元素，直接改变原有元素
    - reverse():倒转数组
    - sort(回调函数):数组排序，和python一样
    - slice():切割数组，slice(start, end),若是ole_arr.slice(),相当于是复制原来这个数组，对新数组的操作并不会影响原数组
    - splice():数组万能操作方法，直接改变原有数组
        - 在指定位置添加元素
            - 结果变量 = 数组变量.splice(指定位置，要删除元素的个数，新增元素...)
        - 在指定位置删除元素
            - 结果变量 = 数组变量.splice(指定位置， 删除个数)
        - 在指定位置替换元素
            - 结果变量 = 数组变量.splice(指定位置，被替换元素，替换元素...)
        - 注意：该方法非常适用于js游戏，需要保证数组的连续性

    - replace(要替换的字符串，替换字符串)：字符串替换函数
    - split()：使用指定的字符切割字符串

    - toLowerCase()

        - 若match(/pattern/)只返回第一个结果数组
        - 若match(/pattern/g)返回所有结果的数组
    - search():进行正则查找，返回的结果是该字符的index
    - substr(开始位置,指定的长度):返回指定位置的指定长度字符串
        - 开始的参数可以为正数，可以为负数。长度可以不填
    - substring(start, end)返回指定位置的字符串
        - start,end必须为自然数，如果是负数表示0
        - 和splice效果类似，但是substring更强，因为不用在意start，end的大小关系

    - source:表示当前正则表达式的正则字符串
    - lastindex:表示进行下一次正则匹配的开始位置。这个是搭配exec使用，对match来说永远为0
        - 注意：仅在g模式下会发生变化，没有g模式，该属性永远为0
    - exec():进行一次正则匹配
        - 在g模式下，每一次匹配都会在上一次匹配的位置之后开始继续匹配，在非g模式下，每次都会从头开始从新匹配
        - 格式：正则对象.exec(字符串);
        - 注：exec和match的区别：
            - 使用：pattern.exec(str), str.match(pattern)
            - 如果在非g模式下，exec和match操作格式不一样，结果是一样的
            - 在g模式下，exec依然进行一次匹配，只不过在上次匹配的内容之后开始一次新的匹配，match是一次性返回字符串中的所有符合正则内容组成的数组
        - 示例：

                console.log(pattern.source);  使用的正则规则
                console.log(pattern.lastIndex);
                var result = pattern.exec(buying);
                console.log(pattern.lastIndex);

        - 使用：pattern.test(str);
        - 该方法是测试正则内容是否在字符串中出现，匹配成功则返回true，匹配失败返回false
    - compile（）：重新编译正则，回收再利用
        - pattern.compile("new_pattern", 正则修正符)

                var new_result = words.match(pattern);

    - getAttribute():
        - 作用：获取元素节点的属性值
        - 格式：元素节点.getAttribute(“属性名”)
    - setAttribute():
        - 作用：获取元素节点的属性值
        - 格式：元素节点.setAttribute("属性名"， vaule)
    - removeAttribute():
        - 作用：移除指定的属性
        - 格式：元素节点.removeAttribute("属性名")

        
## 4.HTML元素节点
- 是在元素节点的基础上针对HTML文档专门设计的一套操作HTML元素节点的方法和属性，更加快捷和方便
- HTMLDOM的相关转换规则
    - 1.HTML标签所具备的属性会变成js对象的成员属性直接使用
    - 2.class属性在转换之后会变成对象的成员属性className（因为class是保留字之一）
    - 3.style属性转换之后会得到一个所有css样式和值组成的集合对象（浏览器为用户操作提前解析了css）
    - 4.在style属性对象访问指定的css样式时，如果样式名称带有-，需要变更为：去掉-，将-后面的第一个字母大写
    

## 5.属性节点
- 页面中任何元素的属性都是使用属性节点创建
- 属性节点一般都是在元素内部出现，单独脱离了元素无法生效
- 成员方法：
    - 以下三个方法都是操作元素节点的属性节点
        - getNamedItem()获取指定名称的属性节点
            - 格式：元素节点.attributes.getNamedItem(属性名)
        - setNamedItem()设置或者添加指定的属性节点
            - 格式：元素节点.attributes.setNamedItem(属性节点)
        - removeNamedItem()移除指定名称的属性节点
            - 格式：元素节点.attributes.removeNamedItem(属性名)

```python
<!DOCTYPE html>
<html>
<head>
<meta charset="utf-8">
<meta http-equiv="X-UA-Compatible" content="IE=edge,chrome=1">
<title>Examples</title>
<meta name="description" content="">
<meta name="keywords" content="">
<link href="" rel="stylesheet">
</head>
<body>
    <div id="mydiv" style="width:500px; height:500px; background:pink">
    	<span>我真的爱你</span>
    	<span>我不知道</span>
    	<span>好吧，其实我并没有想象中那么爱你</span>
    	<div id="kidding">哈哈哈哈哈</div>
    </div>
</body>
</html>

<script>
	var mydiv = document.getElementById("mydiv");
	// 获取元素节点
	// var mystyle = mydiv.getAttribute("style"); 这是获得元素节点的值，是一个字符串对象
	var mystyle = mydiv.attributes.getNamedItem("style");
	// 这是获取属性节点，是一个object对象，可以对该节点再
	console.log(mystyle, typeof mystyle);
	console.log("节点名：", mystyle.nodeName);
	console.log("节点的值：", mystyle.nodeValue);
	console.log("节点的类型:", mystyle.nodeType);

	// 设置属性节点
	var title = document.createAttribute("title");
	console.log(title);
	title.nodeValue = "thisLove";
	mydiv.attributes.setNamedItem(title);
	console.log(mydiv);

	// 修改属性节点
	var style = document.createAttribute("style");
	console.log(style);
	style.nodeValue="background:orange";
	mydiv.attributes.setNamedItem(style);

	// 移除节点
	mydiv.attributes.removeNamedItem("id")
	console.log(mydiv);
</script>
```

## 6.文本节点textNode
- 在HTML中一个页面中的文字都是由浏览器中的文本节点创建而成的
- 文本节点好像只能通过子节点进行查找

### 成员属性
- length：获取一个文本节点的字符个数
- data： 表示文本节点的文本内容字符串

### 成员方法
- spiltText：分开文本节点
    - 格式：文本节点.spiltText(切割位置)
    - 分开文本节点的意义是为了添加其他节点
- appendDate 在文本节点的最后插入字符串
    - 格式：文本节点.appendData(插入字符串)
- insertData 在文本节点的指定位置插入字符串
    - 格式：文本节点.insertData(位置， 要插入的字符串)
- replaceData 在文本节点中替换指定的字符串
    - 格式：文本节点.replaceDate(开始位置，要替换的字符串长度， 替换的字符串)
- substringData  在文本节点中截取指定长度的字符串
    - 格式：文本节点.substringData(指定位置, 截取长度）
- deleteData：在文本节点中删除指定的字符串
    - 格式：文本节点.deleteData(指定位置，指定长度)
    
- 提示：
    - 所有文本节点的操作方法都可以使用字符串代替，文本节点提供的操作文本内容的方法比较短，直接使用文本节点.文本节点方法，文本节点使用字符串操作相对麻烦，需要使用文本节点.nodeValue.字符串方法()。文本节点的修改大部分直接映射到节点当中，而字符串的修改就必须重新赋值为节点的nodeValue，才可以映射到节点当中
- 很少用文本节点的方法来对文本节点进行操作


## 7.文档节点
- 在任何的文档中，文档节点都是不能创建的，是浏览器代为创建的节点类型
- 文档节点不需要创建，也不需要获取，固定名称就是document

### 成员属性
- body：提供在文档对象中直接访问body元素节点的快捷属性
    - document.body
- head:直接访问head元素节点
- title直接获取或色红孩子title值
    - 注意：该属性获取或者设置title标签的内容而非title节点
- cookie：使用js获取当前浏览器中存储的cookie中的信息字符串
    - cookie一般是由服务器进行添加的，js中可以读取当然也可以设置
    - 注意：访问cookie必须使用相同的网址
- domain：获取当前文档的域名地址，本地返回localhost
- URL：获取当前文档的完整地址信息
- lastModified:获取当前文档的最后一次修改时间
- documentElement:表示当前页面的HTML元素节点
    - 注意：低版本有些浏览器不支持
- doctype：获取当前页面的dtd声明节点，即最开始的那一句话

### 成员方法
#### 节点获取类
- getElementByID()通过元素id获取指定id的元素节点
    - document.getElement("id值")
- getElementByName()通过name属性获取指定的节点集合
    - 返回的是数组
- getElementBJyTagName()通过标签名获取指定名称的元素节点集合
- 以下是新版本中的方法
- querySelector():通过css选择器进行元素节点的选取，该方法永远返回第一个符合条件的元素
    - 格式1：document.querySelector(css选择器)
        - 在文档中查找所有符合css选择器的元素，并返回第一个符合的元素
    - 格式2：元素节点.querySelector(css选择器)
        - 在指定的元素中查找....
- querySelectorAll()
- getElementsByClassName():通过class名字查找符合的元素节点，返回数组
    - 也是有两种格式，同上面一样

#### 文档流操作方法
- open()打开一个新的文档流
    - document.open():
    - 注意：打开一个新的文档相当于清空当前页面，准备重新添加内容
- write()在当前文档中输出指定的内容
    - document.write("字符串")
    - write可以直接写入一个文件，而这个文件可以放在服务器端，这样就可以防止被爬虫
- writeln()和write一样，只是在输出内容的结尾添加\n
- close()关闭打开的文档流，就算不用页面文档流也会自动关闭

#### 文档对象中节点创建方法
- createElement()创建指定名称的元素节点
    - document.createElement(标签名)
- createAttribute()创建指定类型的属性节点
- createTextNode()创建指定内容的文本节点
- 以上所有操作均是DOM1标准中的操作，在所有的浏览器中都可以使用
- 以下为HTML5扩展的
- innerHTML:获取或者设置元素节点汇总的所有内容
    - 该属性可以解析HTML标签
- innerText获取或者设置元素节点中的所有文本内容
    - 注意：该属性不可以解析HTML标签
- outerHTML获取或者设置元素节点内部的内容，包括当前标签本身,都是字符串
    - 效果和innerHTML类似，只是包含当前标签而已
    - 在设置操作时，outerHTML的内容会完全替换当前标签，包含当前标签本身
- outerText：获取或者设置元素节点内部的内容，包含当前标签本身（仅保留文字部分）
    - 效果和innerText效果类似，只是包含标签
    - 设置时，内容会直接替换当前标签
- childElementCount：获取当前元素节点的元素子节点个数
- children：获取当前节点的所有元素节点的集合
- classlist：获取当前节点额class所有的值的集合
- firstElementChild：返回元素节点中第一个元素子节点
- lastElementChild：返回元素节点中的最后一个元素子节点
- nextElementSibling：返回当前节点的下一个元素子节点
- previousElementSibling：获取当前节点的上一个元素子节点
- parentElement:获取当前节点的父元素节点，效果和parentNode一致
- localName：获取当前节点的名称，tagName获取大写之，localName获取小写值

# 事件
- 对可以识别的控件的操作就是事件，在特定时间发生的事情

## 事件流的概率
- 事件执行的流程就是事件流，事件流实际上就是在元素之间可以传递的事件的过程

## 事件流的分类
### 捕获流事件
- 事件触发机制是从最外层向最内层元素查找事件的来源

### 冒泡流事件
- 事件由触发事件的第一个元素向最外层元素逐渐扩散

### 标准事件流程
- 在DOM2中W3C提出来标准的事件流程，将事件的流程一共划分为三个阶段：
    - 1.事件的捕获阶段
    - 2.处于事件的阶段
    - 3.事件的冒泡阶段
    
- 绝大多数浏览器都是支持W3C制定的标准
