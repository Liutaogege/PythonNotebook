# 函数
- 结构：
        function 函数名(参数，参数，...){
        函数体
        }
- js函数特性：
    - 如果函数有参数，你没有传递参数进去，那么该参数默认值是undefined
    - 传递过多参数也不会影响，因为多的参数并不会影响函数的运行
    - 传递过少的参数，没有传递的就默认为undefined
- js老版中，不允许在设置参数时给他传递默认参数，这样会报错。如果要为参数添加默认值，是在函数体中通过判断传递的参数是否是为undefined类型，如果是，就给该参数赋值，最后返回该参数即可。


### arguments参数对象
- 它是所有函数中自带的一个特殊对象，只在函数内部起作用，并且永远指向当前函数的调用者传入的所有参数。它类似array，但不是
- 1.arguments对象中包含所有传入的实参组成的对象以及函数本身 
    - 访问时可以使用arguments[标识]访问对应的实参
- 2.arguments.callee在函数中表示当前的函数本身
- 属性和方法：
    - arguments.length:获取当前函数传入的所有实参的个数
    - arguments[index]:访问某个实参 
- 无论是在哪种函数中，都可以使用arguments对象

- 关于函数形参和函数的感应问题
    - 判断谁调用了这个函数
    - 使用：函数名.caller
            var hello = function(){
                if (arguments.callee.caller===zhangsan){
                    console.log("你吃饭了吗");
                }
                if (arguments.callee.caller===lisi){
                    console.log("你好，你天天吃猪草");
                }
            }

            function zhangsan(){
                hello();
            }
            function lisi(){
                hello();
            }
            zhangsan();
            lisi();
### 新增的rest参数对象
- rest参数来获取多余的哪些参数
- rest参数只能写在最后，前面用...标识，从运行结果可知，传入的参数先绑定a，多余的参数以数组形式交给变量rest，
- 示例：
        function saying(name, ...rest) {
            console.log(name, arguments.length);
            console.log(rest);
            for (var i=0; i<arguments.length;i++) {
                console.log("该函数的参数有:", arguments[i]);
            }
        }
        saying("taotao", 23, "成都", "贵州");


## 表达式函数/匿名函数
- var func = function(){函数体};
- 它没有函数名func相当于它的别名
- 注意：函数必须要按照顺序结构来执行 

## 自执行函数
- 一次性函数，自动执行
    - 使用函数局部作用域的特征分割代码区域防止多段代码之间的变量重名和冲突问题
- 如果直接写出
        function(){};会报错
- 而自执行函数就是给他整体加个括号,它就会变成一个函数，再在后面加个括号就能自动执行
        (function(){函数体})() 或者(funcion(){函数体}() )或者+/!/~/- function(){} 
- js中没有命名空间的概念，可以使用自执行函数来代替（命名空间就是防止命名冲突）



## 递归函数的改进
- 在递归函数中如果调用当前函数，禁止使用当前函数的名称，因为函数名是变量，可能随时被修改，所以要求递归函数中使用arguments.callee表示当前函数，arguments.callee()表示当前函数


# 变量提升现象
- JavaScript的函数定义有个特点，它会先扫描整个函数体的语句，把所以申明的变量提升到函数顶部，但是并不会提升变量的赋值

# 局部作用域
- JavaScript的变量作用域实际上是函数内部，所以在for循环等语句块中是无法定义具有局部作用域的变量的，即for语句块中的变量i仍可以被for语句块外调用访问
- 为了解决块级作用域，引入了新的关键字let，用let代替var可以申明一个块级作用域的变量
        for (var j=0; j<5; j++) {
            console.log("haha")
        }
        console.log(j);  //外界仍可访问j

        for (let k=0; k<3;k++) {
            console.log("haha");
        }
        console.log(k) //外界不可访问k，会报k不存在的错误

# 解构赋值
- 新增解构赋值的使用方法
- var name, age, school = ["taotao", [23, "四川大学"]]
- 还可以从对象中迅速取出指定的属性
        var new_person = {
            new_name: "xiaoming",
            age: 14,
            shcool: "六里小学"
        };
        var {new_name, age} = new_person; //必须要加大括号
        console.log(new_name, age);
- 如果取出的属性是赋值给另外一个名字的变量方法如下：
        var person = {
            name: '小明',
            age: 20,
            gender: 'male',
            passport: 'G-12345678',
            school: 'No.4 middle school'
        };

        // 把passport属性赋值给变量id:
        let {name, passport:id} = person;
        console.log(id);  //id已经得到了passport的值
        var {name, single=true} = person;  //当值不存在时传递默认值true
- 和python的解析赋值类似，本质就是结构层次要对应，比如是对象，取属性的时候就要加个大括号


# this变量
- 当处在不同的对象中，就会指向不同的对象。当函数嵌套过多，this指向就会混乱
- 虽然在一个独立的函数调用中，根据是否是strict模式，this指向undefined或window，不过，我们还是可以控制this的指向的！
## 使用apply方法
- apply可以指定函数的this指向哪个对象，是函数本身的方法
- 使用方法： 函数名.apply(需要绑定this的变量， array)，array表示函数本身的参数
- 与apply类似的方法是call()，唯一区别是apply把参数打包成array再传入，call把参数按顺序传入
        Math.max.apply(null, [3, 5, 4]); // 5
        Math.max.call(null, 3, 5, 4); // 5

# 高阶函数
### map函数：原理和python中的一样，只是用法不太相同
        var array = [1, 2, 3, 4, 5];
        function pow(x) {
            return x*x;
        }
        function add(x, y) {
            return x+y;
        }
        var new_array = array.map(pow);
        var new_array1 = array.reduce(add);
        console.log(new_array1);
        console.log(new_array);
### reduce函数，原理和python的一样，用法同上

### filter函数
        var arr = [1, 2, 4, 5, 6, 9, 10, 15];
        var r = arr.filter(function (x) {
            return x % 2 !== 0;
        });
- filter()接收的回调函数，其实可以有多个参数。通常我们仅使用第一个参数，表示Array的某个元素。回调函数还可以接收另外两个参数，表示元素的位置和数组本身：
        var arr = ['A', 'B', 'C'];
        var r = arr.filter(function (element, index, self) {
            console.log(element); // 依次打印'A', 'B', 'C'
            console.log(index); // 依次打印0, 1, 2
            console.log(self); // self就是变量arr
            return true;
        });

### sort函数
- 注意array的sort方法默认是把所有元素先转换为String再排序，所以比较的是ASCII码
- 要实现数字从小到大或者从大到小排序见下面：
        var arr = [10, 20, 1, 2];
        arr.sort(function (x, y) {
            if (x < y) {
                return 1;
            }
            if (x > y) {
                return -1;
            }
            return 0;
        }); // [20, 10, 2, 1]

# 函数的闭包
- 函数里面的变量是静态方式存储，创建之后不会被销毁，这种特征被称为闭包
- 闭包的本质：利用内层函数使得闭包外层函数的参数可以在内存中进行保留
- js里的闭包
    - 是为了让外界可以访问到函数内部的值
    - 使用私有变量，可以避免命名重复的问题，保护全局作用域污染
    - 延长变量生命周期，函数使用完成之后内部的变量不会立即释放，同全局变量一样不会被清理，下次再调用函数继续使用之前保存的局部变量的值
