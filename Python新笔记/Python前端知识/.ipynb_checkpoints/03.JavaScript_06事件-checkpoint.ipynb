# 事件
- 对可以识别的控件的操作就是事件，在特定时间发生的事情

## 事件流的概率
- 事件执行的流程就是事件流，事件流实际上就是在元素之间可以传递的事件的过程

## 事件流的分类
### 捕获流事件
- 事件触发机制是从最外层向最内层元素查找事件的来源

### 冒泡流事件
- 事件由触发事件的第一个元素向最外层元素逐渐扩散

### 标准事件流程
- 在DOM2中W3C提出来标准的事件流程，将事件的流程一共划分为三个阶段：
    - 1.事件的捕获阶段
    - 2.处于事件的阶段
    - 3.事件的冒泡阶段
    
- 绝大多数浏览器都是支持W3C制定的标准

- 缺点：1.无事件标准  2.事件的效率低， 优点：所有浏览器都兼容

## DOM2阶段
### IE版本
- 仅支持冒泡事件
- attachEvent()注册事件
    - 格式：元素节点.attachEvent("事件类型",事件函数)
    - 事件类型：使用带有on的事件类型，例如onclick
    - 事件函数：推荐你使用声明函数，因为删除还需要使用这个函数
- detachEvent()移除事件
    - 格式：元素节点.detachEvent("事件类型", 事件函数)
    - 事件类型：使用带有on的事件类型
    - 事件函数：也推荐使用声明函数
    
### 非IE版本
- addEventListener()注册事件
    - 格式：元素节点.addEventListener("事件类型", 事件函数，事件阶段)
    - 事件类型：使用事件名称，不可以使用on，例如click或者mousemover
    - 事件函数：不推荐使用匿名函数，推荐使用声明函数，因为移除操作需要同一个的函数
    - 事件阶段：true捕获流，false冒泡流
- removeEventListener()取消事件

# 事件类型的分类
### 一般事件
- 用于元素标签
- onclick单击事件
- ondblclick 双击事件
- onmousemove 鼠标移动事件

- onmouseover 鼠标经过事件，会冒泡
- onmouseleave  鼠标离开事件

- onmouseenter 鼠标进入事件，不会冒泡（即向外层元素传递）
- onmouseout  鼠标移出事件

- onmouseup  鼠标抬起事件
- onmousedown  鼠标按下事件

- onkeyup  键盘抬起事件
- onkeydown 键盘按下事件，任何键有效

- onkeypress 键盘释放事件，只能数字字母键有效
- mouseleave和mouseout的区别：
    - 1.不论鼠标指针离开被选元素还是任何子元素，都会触发mouseout事件
    - 2.只有在鼠标指针离开被选元素时，才会触发mouseleave事件
    - 常用mouseleave
    
### 页面事件
- 用于window
- onabort：页面终止事件，在网页没有完全加载的情况下关闭
- onunload：页面退出完毕事件，xp系统，ie才可以用
- onload：页面加载完毕才执行的事件，新版浏览器把这个问题都解决了
- onbeforeunload：页面即将退出事件，新版本不太好用
- oncopy：内容复制事件
- oncut：内容剪切事件
- onscroll：页面滚动事件
- onresize：页面大小调整事件
- onselect：页面内容选中事件（表单元素）
- onselectstart 页面内容选中开始事件（任意元素），开始选中的时候就执行
    - 为文档对象添加禁止选中操作
        - document.onselectstart=function(){
        return false
        }

### 表单事件
- onblur：失去焦点事件
- onfocus：获取焦点事件
- onchange：表单值改变事件
- onsubmit：提交表单事件
- onreset 重置表单事件



# 事件对象
- 在触发事件的时候，与事件相关的信息存储对象就是事件对象
- 事件对象在不同的浏览器中体现方式不相同
- IE浏览器
    - 事件对象是window的子对象之一，可以通过window，event或者event直接访问到事件对象
- 火狐浏览器
    - 事件对象是以参数的形式传入了事件函数的形参中，接收之后才可以使用
- 事件对象兼容性处理（函数中使用）
    - var e = window.event?window.event:evt;
    - var e = window.event || evt;

## 事件对象的成员
- 事件对象可以划分为键盘事件对象和鼠标事件对象

### 成员属性
- altkey：检测是否按下alt键
    - 按下为true，没按false
- shiftkey：检测是否按下shift键
    - 按下为true
- ctrlkey：
- keyCode：返回键盘按键的ASCII码
- bubbles:事件是否可以冒泡传播，true冒泡，false不冒泡
- ============================
#### 阻止冒泡传播的方法
- window.event.cancelBubble:(早期ie版本,不过现在都可以兼容了)
    - 是否取消冒泡传播，true是，false不阻止
- window.event.stopPropagation():w3c早期，现在也都支持了

- =============================


- window.event.currentTarget：返回触发当前事件的元素
- window.event.clinetX：事件属性返回当事件被触发时鼠标指针指向对于浏览器页面（或客户区）的水平坐标
- window.event.clientY:事件属性返回当事件被触发时鼠标指针指向对于浏览器页面（或客户区）的垂直坐标
- window.event.screenX:返回鼠标相对于屏幕左侧的水平距离
- window.event.screenY:返回鼠标相对于屏幕顶部的垂直距离
- window.event.pageX；相当于clinetX 非IE专用
- window.event.pageY
- window.event.offsetX：返回鼠标相对于有触发事件的元素的水平坐标
- window.event.offsetY：返回鼠标相对于。。垂直坐标
- window.event.layer：系列早期ie版本，offset系列，W3C版本
- window.event.SrcElement返回触发事件的元素(IE),target属性也表示相同的意义（标准）
- window.event.toElement:仅针对于mouseover和mouseout事件有效的属性，返回移入鼠标的元素
- window.event.type：返回发生事件的类型名称，没有on的类型字符串
- 取消默认行为
    - window.event.returnValue：把这个属性设置为false，可以取消发生事件的源元素的默认动作（IE）
    - window.event.preventDefault()和returnValue一样取消浏览器的默认操作的方法（W3C）
