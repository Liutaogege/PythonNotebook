# css属性介绍
## 背景属性
- background-color: yellow;背景颜色，可简写为background: yellow;
- background-image: url("地址"); 背景图片
    - 地址分为相对地址和绝对地址，相对地址是指处在同一文件夹下，直接使用名字即可，.表示上一级目录，..表示上二级目录
    - background-repeat: no-repeat; 背景是否重复直至填满整个背景
    - background-repeat: repeat-x; x轴方向重复，还有y轴
- background-position: 0 0; 偏移量。同时偏移量还可以用left bottom right以及百分比表示
    - background-position:50% 50%;
    - background-position:left bottom;
    - background-position-x:
    - 在Chrome中调试时，按住alt偏移量为0.1，shift偏移量为10，Ctrl键偏移量为100
- background-attachment:设置背景图片的绑定方式
    - scroll：背景图片随页面内容滚动（默认）
    - fixed：背景图片不随内容滚动
- background-size: 100% 100%; 图片大小设置,也可以用像素
    - 若是：background-size: cover; 就是裁剪来拉满
- background: url() yellow center center;组合写法

## 尺寸属性
- width：设置元素的宽度
- height：设置元素的高度
- min-width:设置元素的最小宽度
    - 最小宽度设置之后，如果外层元素的宽度大于最小宽度设置，则当前元素使用外层元素的最大宽度，如果外层元素的宽度小于当前元素的最小宽度设置，那么当前元素的宽度为设置的最小宽度值
    - 在页面布局中使用较少
- min-height:最小高度,论坛回复中使用很多
    - 如果元素中的内容没有超过最小高度值，那么当前元素高度为最小高度值，如果元素中的内容高度超过最小高度值，那么当前元素高度会增高，使得内容完全包含在元素中
- max-width
- max-height


## 字体属性
- color：设置文字的颜色，如：color：red
    - rgb（x,y,z）颜色模式：每个值代表一种颜色
        - color: rgb(0,0,255);
    - rgba(x,y,z,a)颜色模式，最后一个值代表透明度，取值范围0-1,0是完全透明，1是完全不透明
        - color: rgba(220, 200, 200, 0.7)
    - 16进制模式：
        - 由六位数字组成前两个代表红色，中间两位代表绿色，最后两位代表蓝色，比如：#ff0000表示红色
- font-size：字体大小，
    - chrome中字体最小为12px
    - 字体大小一般为偶数
- font-family：文字字体，
- font-style：设置字体是否倾斜，italic倾斜,normal不倾斜
- font-weight:文字是否加粗，normal不加粗，bold加粗，lighter比外层元素更加细，bolder比外层元素更加粗，100-500正常字体，600-900粗体，800常用的粗体代词。常用400表示正常，700
- font-variant：设置字体是否小写变为大写字母。与真正的大写字母的字体大小有区别
    - normal正常字体，small-caps小写字母大写。该属性仅针对英文有效
- 组合写法：
    - font: 倾斜 加粗 文字大小/行间距 serif;
        - italic bold 12px serif;
    - serif:衬线字体，就是有笔锋，sans-serif无衬线字体，无笔锋
    
## 文本属性
- text-indent：文字的首行缩进，单位：px，em（相对于文字大小缩进几个），rem（相对于根元素缩进几个），%
- text-align:文本的水平对齐方式，有center，left，right
    - 注意：vertical-align:垂直方向居中，对文本无效，只对内联元素和table有效
- line-height:行间距,设置大小单位有像素px，em相对大小（1em=当前值），rem相对于body的大小，%百分比。推荐使用em 
    - 可以利用line-height设置单行文本的垂直居中，值给父级元素的高度的值。比如单行文本嵌套在div中，将line-height的值设置为div高度的值即可
- word-spacing:单词之间的间距
- letter-spacing:字间距
- white-space：设置元素对于空白的处理方式
    - 值pre：显示源代码中的样式，遇到边界不换行
    - 值nowrap：取消自动换行，除非遇到br标签
    - 值pre-wrap:显示源代码中的格式，遇到边界换行
    - 值pre-line：保持换行字符显示，但是空白字符不显示
- text-transform:设置英文的大小写转换方式
    - none：不做修改
    - capitalize:首字母大写
    - uppercase：全部大写
    - lowercase：全部小写

## 定位属性
- position属性：设置元素的定位方式
    - 除了static都有top，bottom，left，right四个属性
    - static：静态定位方式（默认）
        - 元素默认就是静态方式定位
    - absolute：绝对定位方式，灵活性最高
        - 绝对定位元素会使得当前元素脱离文档流，可以在任何范围内移动，而且不会因为移动影响页面中的其他元素。已经不占据原来的位置了，而且飘的比浮动元素还高，会把浮动元素盖在下面，且不像浮动元素一样从左开始排列，所以会出现重叠现象
            - 1.如果当前绝对定位元素的外层元素都是静态定位元素，只能以HTML标签作为定位参考点
            - 2.距离当前绝对定位元素外层最近的一个非静态定位元素作为参考点
        - 绝对定位的特征：
            - 会从原来的位置飘起来，和浮动一样，它原来的位置会空出来，会被别的元素占据
            - 只有设置top，bottom，left，right四个属性来改变它飘起来的位置，否则它就会在它原来处的那个位置的上方，且后面的元素会移动到它原来空出来的那个地方。这四个属性的参考点如上所讲
            - 飘的位置比浮动的高，可以把浮动元素给遮住，其余的基本和relative一样
    - relative：相对定位方式
        - 相对应元素没有脱离文档流，在页面中还占据着原来的位置中，
        - 移动的参考对象就是原来的位置(static时的位置）。如果要移动，必须要设置设置top，bottom，left。right值才能发生移动
        - 相对定位的特征：
            - 元素会从原来的位置飘起来，不过同比浮动来说，它有双脚占据着原来的那个位置，所以空位置不会被别的元素占据。
            - 只有设置top，bottom，left，right四个属性来改变它飘起来的位置，否则看起来好像没什么改变（不过这点可以用来和absolute结合使用，父级用这个但不设置四个属性值，子级用absolute，再改变这四个属性值，那么就可以实现子元素相对父级任意位置的排列）。这四个属性的参考点是它原来占据的那个位置。
            - 他飘的位置比浮动的高，所以它可以把浮动元素给遮住
            - 如果两个元素同时设置了relative，且分别设置left和right属性出现了重叠现象，那么后面的那个元素会盖住前面的那个元素
    - fixed：绑定定位方式
        - 绑定定位的参考点是浏览器的视窗，和任何标签无关

- left属性：设置定位元素距离定位参考点左侧的距离
- right属性：
- top属性
- bottom属性
- 注意，left，right，top，bottom属性只有当position属性设置为relative,absolute,fixed才生效。
- z-index属性：（和PPT中上移一层类似）
    - 设置定位元素的层级关系，所有元素的默认z-index的值是0
    - 取值：整数值，没有上限
    - 在相同定位的情况下，谁的z-index值越大，显示越靠上
    - 注意：静态定位元素（即static）无法使用该属性
    
## 布局属性
### display布局
- display属性：设置元素的显示方式
    - none  不显示元素也不占用物理空间，即隐藏状态
    - inline  显示行内元素
    - block  显示为块状元素
    - inline-block  显示为行内块状元素
    - （前四个必须记住）
    - list-item  显示li列表标签的样式
    - table  显示为table标签的样式，作为块元素显示
    - inline-table  显示为table标签的样式，但是作为内联元素显示
    - table-row  显示为tr标签的样式
    - table-cell  显示Wietd标签的样式
    - 注意：display仅改变元素的显示效果，只用于的视觉改变，不会改变标签的意义
    
- visibility属性：设置元素显示与否
    - inherit  继承（默认值）显示与否由父元素的显示决定
    - visible 显示
    - hidden  隐藏元素，占用物理空间，需要注意和display: none的区别
    - collapse  专门用于表格行和单元格的隐藏设置属性
    
- overflow-y属性：设置元素垂直方向溢出的处理方式
    - auto  如果内容超过元素，则出现滚动条，不超过就不出现
    - scroll  一直存在滚动条，但是有没有滑块取决于内容的高度
    - hidden  对于超出高度的内容进行隐藏处理
    - visible  对于超出的内容做显示处理（默认值）
    
- overflow-x属性：设置元素水平方向溢出的处理方式
    - auto  如果内容超过元素，则出现滚动条，不超过就不出现
    - scroll  一直存在滚动条，但是有没有滑块取决于内容的宽度
    - hidden  对于超出宽度的内容进行隐藏处理
    - visible  对于超出的内容做显示处理（默认值）
    - 注意：正常情况下内容超过元素水平宽度会自动换行到下一行
        - 不换行的情况：
            - 1.如果内层包含的不是文字，而是其他图片，元素等内容，而且这些内容宽度超过外层元素
            - 2.连续的英文字符，没有空格和标点符号，会被浏览器当做一个完整的单词显示，不会强制换行
            - 3.设置内容显示效果为white-space:nowrap ,强制不换行属性
            
### float属性： 设置元素的浮动方式，用来排版页面
- float的值：
    - none  元素不浮动（默认值）
    - left 左浮动
    - right 右浮动
- 浮动的特征：
    - 从元素原有的位置漂浮起来，可以进行左右浮动，但是不能设置移动位置
    - 浮动元素不会脱离文档流，位置处于静态定位元素和绝对定位元素之间。
    - 文字是不会被设置了浮动属性的元素盖住的，它只能盖住元素
    - 设置了浮动属性的相邻元素会从左到右依次排列
    - 浮动元素相当于是从原来的位置升了起来，那么原来的位置就会空了。所以后面如果有元素的话就会去占据这个空位，就导致元素被浮动的元素遮住。为了避免元素去占据那个空的位置，即避免被遮住，就需要清除浮动（不是真的清除浮动，而是让浮动元素相当于长脚了把那个空位置守住）
- 清除浮动的方法：
    - 第一种方法：设置clear属性，
        - clear的值可以为：
            - none  不清除浮动（默认）
            - left  清除左浮动
            - right  清除右浮动
            - both  清除左右的浮动（常用）
        - clear属性是给想去占据空位的那个元素设置的，即让后面的元素清除浮动。而不是浮动元素本身清除
    - 第二种方法：
        - 可以把设置了浮动的元素包裹在一个父级元素中，然后在父级元素中设置overflow:hidden;也可以达到清除浮动的效果。或者在父级元素中最后添加一个div元素，然后设置clear属性也可以
- 注意：
    - 浮动属性和定位属性是不同的属性。
    - 如果想实现同级元素浮动在同一排，必须都设置float属性
- 备注：虽然position中的inline-block也可以用来排版页面，但是会很麻烦，无法控制元素的间距。一旦设置float属性，就会把position属性的效果给清除

### flex布局
#### 简介：
- 1.概念：css3弹性盒子（Flexible或者Flexbox），是一种用于在页面上布置元素的布局模式，使得当前页面布局必须适应不同的屏幕尺寸和不同的显示设备时，元素也可预测的运行。对于许多应用程序，弹性盒子模型提供了对块模型的改进，因为它不使用浮动，flex容器的边缘也不会与其它内容的边缘折叠
- 2.兼容性：参考：https://developer.mozilla.org/zh-CN/docs/Web/CSS/CSS_Flexible_Box_Layout/Using_CSS_flexible_boxes

#### 相关概念
- 1.弹性容器：包含弹性醒目的父元素，通过给这个元素设置display属性的值为flex或者inline-flex来定义弹性容器。
- 2.弹性项目：弹性容器的每个子元素都称为弹性项目
- 3.轴：分为主轴和纵轴，默认横向的轴是主轴，纵向的轴为纵轴。也可以通过flex-direction:column来将主轴设置为纵向的，主轴其实更恰当的解释为：元素按顺序排列的那根线
#### 相关css属性
- 1.flex-direction：确定轴的方向，默认横向
    - row：横向布局
    - row-reverse：横向反转布局
    - column：纵向布局
    - column-reverse：纵向反转布局
- 2.justify-content：属性定义了项目在主轴上的对齐方式
    - flex-start：主轴的起始点对齐（默认）
    - flex-end：主轴的结束点对齐
    - center：居中排列
    - space-between：两端对齐，项目之间的间隔相等
    - space-around：每个项目两侧的间隔相等，所以，项目之间的间隔比项目与边框的间距大一倍
- 3.align-items:确定项目在纵轴上如何对齐
    - flex-start：纵轴的起始点对齐
    - flex-end：纵轴的结束点对齐
    - center：纵轴的中点对齐
    - stretch：默认值：如果没有设置高度，弹性元素被在侧轴方向被拉伸到与容器相同的高度或宽度
- 4.flex-wrap：指定子元素在一行排列不下的时候，该如何表现
    - nowrap：不换行，被挤到一行
    - wrap:被打断到多行中
- 5，align-content：确定纵轴的轴线如何对齐，只有在多行的情况下才有效
    - flex-start：纵轴的起始点对齐
    - flex-end：纵轴的结束点对齐
    - center：纵轴的中点对齐
    - stretch：默认值：如果没有设置高度，弹性元素被在侧轴方向被拉伸到与容器相同的高度或宽度

# 盒子模型
- 页面中的任何一个元素在显示和处理时都是以一个盒子的方式进行处理的
- 把元素叫做盒子，设置对应的样式分别为：盒子的边框（border），盒子内的内容和边框之间的边距（padding），盒子与盒子之间的间距（margin）
    - 盒子真实尺寸：
        - 盒子宽度=width+padding左右
        - 盒子高度=height+padding上下
    - 在布局中，如果想增大内容和边距的距离，又不想改变盒子显示的尺寸
        - box-sizing: content-box|border-box
        

        
## 边框属性
- border:width style color;
    - 设置该值的时候，三个值尽量都写上
- border-color:边框颜色
- border-style:边框样式,solid实线，dotted点状线，dashed虚线
- border-width:边框宽度
- border-left:
- border-left-color:
- border-left-style:
- border-left-width
- border-radius:圆角处理
    - border-radius: 10px;单独给一个值是设置四个角的属性
    - border-radius: 4px, 6px, 8px, 10px;顺序是顺时针走。设置四个值是单独设置每一个角
    - 给两个值是对角线两两相同
- box-shadow:x轴偏移 y轴偏移 扩散程度 颜色 inset内阴影 设置或检索对象阴影;
    - 参数较多，不设置内阴影可以不写
    
## 内边距属性padding
- 指的是元素内容和边框之间的距离，值不能为负数
- padding:组合写法
    - 一个数值时：表示距离四边的距离都是这个值
    - 两个数值时：表示上下值相同，左右值相同
    - 三个数值时：表示上 左右 下三条线，即左右值相等
    - 四个数值时：表示上 右 下 左
- 分开写法：padding-top,padding-bottom,padding-right,padding-left
- 注意：padding属性的增加会使得元素整体宽高的增加，如果希望宽高高度不变，将padding属性的值从元素的宽度和高度的设置中减去。padding的值设置要在元素中设置而不是元素内容中设置

## 外间距属性margin
- margin属性：同时设置四个方向的外间距，值可以为负数
- margin:组合写法
    - 一个数值时：表示与四边的元素之间的间距都是这个值
    - 两个数值时：表示上下值相同，左右值相同
    - 三个数值时：表示上 左右 下三条线，即左右值相等
    - 四个数值时：表示上 右 下 左
- 分开写法：margin-top,margin-bottom,margin-right,margin-left
- 而非块级元素没有margin-top和bottom的值设置
- margin-left: auto;  margin-right: auto;设置了宽度的元素才能像这样实现自动间距（可以认为没有设置宽度的元素占一行，此时就是居中），比如inline-block元素，auto只针对外边距才有
- 注意：无论元素是包含嵌套关系，还是正常的并列关系都不会影响margin属性的设置
- 边距重合：
    - 一个元素的下外边距和另外一个元素的上外边距碰在一起发生了重合，外边距会顶着实际内容边缘。
    - 比如上元素设置外边距20px，下元素设置外边距30px，按道理应该是50px，但实际上只有30px
    - 解决办法；
        - 1.添加一个父元素，然后使用border：1px solid transparent;但是会长出1px
        - 2.在父元素使用overflow：hidden;

## 内容属性设置
- content属性 设置元素内容属性
    - 一般用于配合:before和:after设置内容使用，可以设置文字也可以设置图片
            格式
            元素:before{
                content:'字符内容'
             }
             元素:before{
                 content:url(地址)
             }
     - 这种before或after写法所增加的内容，并不是真正的内容，而是属于元素的。但是审查元素的时候不会查看到它，浏览器是看不到他的，是给人看的
