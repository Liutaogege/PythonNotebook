### 一、获取临时token

```apl
@api {get} /external_use/construction_dispatch/get_token
@apiParamExample {Array} 请求参数
@apiParam     	{Int}        uid													OA系统用户ID

@apiSuccessExample {Array} 返回数据
@apiSuccess     {Str}        access_token									临时token，有效期1H
返回值类型: JSON
{
		"message": "OK",
		"access_token": ""
}
```
