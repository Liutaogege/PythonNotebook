# 项目初始化
- 1.vue create meituan
- 2.sass配置：
        npm install node-sass --save-dev
        npm install sass-loader --save-dev
        style就可以通过lang="scss"来写scss语法了
- 3.rem配置
        npm install postcss-pxtorem --save-dev
        npm install lib-flexible --save-dev
        配置的是对应总项目目录下的那个package.json文件
        配置package.json文件
                "postcss": {
                  "plugins": {
                    "autoprefixer": {},
                    "postcss-pxtorem": {
                      "rootValue": 37.5,
                      "propList": [
                        "*"
                      ],
                      "selectorBlackList": [
                        "van-*"
                      ]
                    }
                  }
                }
        配置main.js
        import  "lib-flexible"
        
- 4.vant
        npm install babel-plugin-import --save-dev
        npm install vant --save
        在babel.config.js中配置
                module.exports = {
                    plugins: [
                        ['import', {
                            libraryName: 'vant',
                            libraryDirectory: 'es',
                            style: true
                            }, 'vant']
                    ]
                };
- 4.styles/init.css中做浏览器的初始化操作
    - 在App.vue中的style中
            <style lang="scss">
            // 注意路径，如果路径报错，则在前面加上~
            @import "~./assets/styles/init.css";
            </style>

- 其次：如果仍然不行，就就在项目目录下执行：npm i babel-plugin-import -D

### css中当文字溢出时设置显示为...
- 如果该元素的父元素使用了flex且没有设置宽度，那么记得也给父元素设置overflow:hidden(这是单行文本溢出的设置，多行溢出的自行百度)
        .notice{
              text-overflow: ellipsis; //设置当文字超出的时候显示为...
              white-space: nowrap;
              overflow: hidden;
            }
        其中，white-space：nowrap；表示文本不会换行，在同一行继续，知道遇到标签为止；
        overflow：hidden;不显示超过对象尺寸的内容，就是把超出的部分隐藏了；
        text-overflow：ellipsis;当文本对象溢出是显示...，当然也可是设置属性为clip不显示点点点；
        -o-text-overflow:为了兼容opera浏览器;
