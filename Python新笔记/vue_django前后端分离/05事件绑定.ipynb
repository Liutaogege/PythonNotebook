# 事件绑定
- 事件绑定就是在html元素中，通过v-on绑定事件的，事件代码可以直接放到v-on后面，也可以写成一个函数。可以把v-on:省略写成@
          <div id='app'>
            <p>{{ count }}</p>
            <button v-on:click="count-=1">点我减1</button>
            <button v-on:click="countAdd(100)">点我加</button>
            <button @click="count*=2">点我乘2</button>
          </div>
          <script>
            new Vue({
              el: '#app',
              data: {
                count: 10,
              },
              methods: {
                countAdd(add_value) {
                  this.count += add_value;
                }
              }
            })
          </script>
          
## 传入event参数
- 如果在事件处理函数中，想要获取原生的dom事件，那么会在html代码中，调用的时候，可以传递一个$event参数


## 中间修饰符
- 有时候事件发生，需要做一些操作，比如针对这个事件要它的默认行为，那么可以通过修饰符来实现
    - 常见的修饰符如下：
        - .stop:阻止事件冒泡
        - .capture:让事件成为捕获事件
        - .once:这个事件只执行一次
        - .self:代表当前这个被点击的元素自身
        - .passive:在页面滚动的时候告诉浏览器不会阻止默认行为，从而滚动的更加流畅
        
           

# 详见代码如下：
        <div id='app'>
        <p>{{ count }}</p>
        <button v-on:click="count-=1">点我减1</button>
        <button v-on:click="countAdd(100)">点我加</button>
        <button @click="count*=2">点我乘2</button>

        <!-- 这个event必须要加上$符号，否则无法识别 -->
        <a href="https://www.baidu.com" @click="goto360($event)">点我跳转</a>

        <!-- 比较简单的可以使用修饰符修饰 -->
        <a href="https://www.baidu.com" @click.prevent>点我跳转百度(会被阻止,且啥事也不干)</a>
        <a href="https://www.baidu.com" @click.prevent=gotoTaobao>点我跳转百度(会被阻止)</a>

        <!-- 阻止冒泡修饰符 -->
        <!-- 如果在这加上捕获，捕获就是由外向内，则先输出外部元素 -->
        <div class="outer" @click.capture="outerClick">
          <!-- 如果在这加上stop修饰符，则不会继续向外冒泡 -->
          <div class="mid" @click.stop="midClick">
            <div class="inner" @click="innerClick"></div>
          </div>
        </div>
      </div>
      <script>
        new Vue({
          el: '#app',
          data: {
            count: 10,
          },
          methods: {
            countAdd(add_value) {
              this.count += add_value;
            },
            // 这里的event不需要加上$符号，甚至无需为event名字，其他变量名也行
            goto360(event) {
              event.preventDefault();
              window.location = "https://www.360.cn";
            },
            gotoTaobao() {
              window.location = "https://www.taobao.com";
            },
            outerClick() {
              console.log("外部元素");
            },
            midClick() {
              console.log("中间元素");
            },
            innerClick() {
              console.log("内部元素");
            }
          }
        })
      </script>
