# Vue介绍
- Vue是一套用于构建前后盾分离的框架，刚开始时是由国内优秀选手尤雨溪开发出来的，目前是全球最流行的前端框架，使用vue开发网页很简单，并且生态技术完善，社区活跃，是前后端找工作必备之 技能

# Vue的安装和使用
- vue的安装大体上分成三种形式，第一种是通过script标签引用的，第二种是通过npm来安装，第三种是通过vue-cli命令行安装，建议直接通过第一种方式进行安装，简单快捷

## 基本使用
- 要使用vue，首先需要创建一个vue对象，并且在这个对象中传递el参数，el参数全称element，用来找到一个vue渲染的根元素，并且可以传递一个data参数，data中的所有值都可以直接在根元素下使用{{}}来使用，示例代码如下：
        <div id="app">
        <p>{{username}} is {{age}} years old</p>
        <button @click="username='娇娇'">改变我</button>
          </div>
          <script>
            new Vue({
              el: "#app",
              data: {
                "username": "liutao",
                "age": 23
              }
            })
- 另外也可以在vue对象中添加methods，这个属性中专门用来存储自己的函数，methods中的函数也可以在模板中使用，并且在methods中的函数来访问data中的值，只需要通过this属性名就可以访问到了，不需要添加this.data.属性名来访问，示例代码如下：
        <div id="app">
            <p>{{username}} is {{age}} years old</p>
            <button @click="username='娇娇'">改变我</button>
            <p>{{greet()}}</p>
          </div>
          <script>
            new Vue({
              el: "#app",
              data: {
                "username": "liutao",
                "age": 23
              },
              methods: {
                greet() {
                  return "I really loved you so much once"
                }
              }
            })
          </script>

## 文本
- 在HTML中通过{{}}（双大括号）中可以把vue对象中的数据插入到网页中，并且只有vue对象上对应的值发生改变了，那么HTML中双大括号中的值也会立马改变
- 代码如上第一个示例
- 如果在HTML的{{}}中，第一次渲染完成后，不想要跟着后期数据的更新的改变而改变，那么可以使用v-once指令来实现它的不被改变
    - <p v-once> {{username}} </p

## 显示原生的HTML
- 有时候在vue对象中，或者是后台传递给我们的一段原生的HTML代码，我们需要渲染出来，若直接通过{{}}渲染会把这个html代码当做字符串，这时候可以通过v-html指令来让其渲染。

## 属性绑定
- 若想要在HTML元素的属性上绑定我们vue对象中的变量，那么需要通过v-bind来实现（v-bind:name="username"）,可以直接省略v-bind留下:，比如img绑定src属性，直接使用{{}}是不行的
        <div id="app">
         <img src="{{logo}}" alt="">
         <img v-bind:src="logo" alt="">
            <hr>
            <p>可以简写</p>
            <img :src="logo" alt="">
          </div>
          <script>
            new Vue({
              el: "#app",
              data: {
                username: "liutao",
                code: "<a href='https://www.baidu.com'>百度一下你就知道</a>",
                logo: "https://hao1.qhimg.com/t01ea3a44e231e5d6f4.png"
              }

            })

# 属性绑定Class和Style
- 在绑定class或者style的时候，可以通过以下方式来实现

## 绑定Class
- 1.通过数组的方式来实现
        <style>
           .title {
             color: red;
           }
           .another_title {
             font-weight: 700;
           }
           .italic_title {
             font-style: italic;
           }
         </style>
        </head>
        <body>
          <div id='app'>
            <p v-bind:class="[pclass1, pclass2, pclass3]">许欣欣，你好</p>
          </div>
          <script>
            new Vue({
              el: '#app',
              data: {
                pclass1: "title",
                pclass2: "another_title",
                pclass3: "italic_title"
              }
            })
- 2.通过对象的方式来实现
        <body>
          <div id='app'>
            <!-- 使用对象的方式来实现 -->
            <p v-bind:class="{title:true, another_title:true, italic_title:italic}">许欣欣，你人挺不错的</p>
            <button @click="italic=true">点击我改变</button>
          </div>
          <script>
            new Vue({
              el: '#app',
              data: {
                italic: false
              }
            })
          </script>
        </body>

## 绑定style
- 同样也有两种方式
        <div id="app"> 
        <!-- 使用对象方式绑定 -->
            <p v-bind:style="{backgroundColor: bgColor}">说实话，你很漂亮，还很活泼阳光</p>

            <!-- 使用数组方式绑定 -->
            <p v-bind:style="[pStyle1, pStyle2]">如果我可以早生几年，如果可以早点碰见你</p>
          </div>
          <script>
            new Vue({
              el: '#app',
              data: {
                bgColor: 'red',
                pStyle1: {
                  'background': 'yellow',
                  'font-size': '24px',
                },
                pStyle2: {
                  'border-bottom': "1px solid red"
                }
              }
            })
          </script>
