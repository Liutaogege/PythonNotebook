```python
class Student():
    def __init__(self,name,age,yuwen,shuxue,English):
        self.name = name
        self.age = age
        self.yuwen = yuwen
        self.shuxue = shuxue
        self.English = English
    
    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    def get_course(self):
        return int(max(self.yuwen, self.shuxue, self.English))
    
a = Student("xiaoming", 13, 98, 89, 23)
max_grade = a.get_course()
name = a.get_name()
age = a.get_age()
print(max_grade, name, age)
```

```python
class DictClass():
    def __init__(self, dic):
        self.dic = dic
        
    def del_dict(self, key):
        try:
            self.dic.pop(key)
        except:
            print("没有该键值")
        finally:
            return self.dic
        
    def get_dict(self, key):
        if key in self.dic.keys():
            return self.dic[key]
        else:
            return "not found the %s"%key
        
    def get_key(self):
        return self.dic.keys()
    
    def update_dict(self, dic):
        self.dic.update(dic)
        return self.dic.values()
    
a = DictClass({"xiaoming":28, "xiaobai":78})
#print(a.del_dict("xiaoming"))
print(a.del_dict("xiaohong"))
print(a.get_dict("xiaobai"))
print(a.get_dict("xiaohong"))
a_list = a.get_key()
print(a_list)
print(a.update_dict({"xiaohong":89}))
```

```python
class ListInfo():
    def __init__(self, lst):
        self.lst = lst
        
    def add_key(self, ele):
        self.lst.append(ele)
        return self.lst
    
    def get_key(self, num):
        if num >= len(self.lst):
            return "该索引值超标"
        else:
            return self.lst[num]
    
    def update_list(self, lst):
        self.lst.extend(lst)
        return self.lst
    
    def del_key(self):
        return self.lst.pop()
    
lt = ListInfo(["huhu", 'hehe', 'xixi', 'henhen'])
print(lt.add_key("haha"))
print(lt.get_key(9))
print(lt.get_key(3))
print(lt.update_list(['xiuxiu', 'biubiu']))
print(lt.del_key())
```

```python
class SetInfo():
    def __init__(self, st):
        self.st = st
        
    def add_setinfo(self, ele):
        self.st.add(ele)
        return self.st
    
    def get_inter(self, new_set):
        return self.st & new_set
    
    def get_union(self, new_set):
        return self.st | new_set
    
    def del_diff(self, new_set):
        return self.st - new_set
    
st = SetInfo({"xiaxia", 'taotao', 'linlin'})
print(st.add_setinfo("xinxin"))
print(st.get_inter({"xiaxia", 'huihui', 'taotao'}))
print(st.get_union({"xiaxia", 'huihui', 'taotao'}))
print(st.del_diff({"xiaxia", 'huihui', 'taotao'}))
```

```python
course_lst = []

class School():
    def __init__(self, school_name):
        self.school_name = school_name
        self.stu_lst = []
        self.tea_lst = []
    global course_lst
    
    def hiring(self, obj):  # obj表示的是一个实例化对象
        self.tea_lst.append(obj.name)
        print("新老师{}入职成功".format(obj.name))
    
    def enrolling(self, obj):
        self.stu_lst.append(obj.name)
        print("新同学{}成功报名".format(obj.name))
    
class Grade(School):
    def __init__(self, school_name, grade_code, grade_course):
        super(Grade, self).__init__(school_name)
        self.code = grade_code
        self.course = grade_course
        self.members = []
        course_lst.append(self.course)
        print("我们开设了{}的{}年级的{}课程".format(self.school_name, self.code, self.course))
        print("课程{}的大纲是day01, day02, day03".format(self.course))
        


class School_members():
    def __init__(self, name, age, role):
        self.age = age
        self.name = name 
        self.role = role
        self.course_list = []
        
stu_num_id = 0
class Students(School_members):
    def __init__(self, name, age, role, course):
        super(Students, self).__init__(name, age, role)
        global stu_num_id
        stu_num_id += 1
        stu_id = course.school_name + "-S" + str(course.code) + str(stu_num_id).zfill(2)
        
        self.id = stu_id
        self.mark_dict = {}
        
    def introducing(self, course):
        print("hello，我叫{},是一名{},学习课程是{}，学号是{}".format(self.name, self.role, course.course, self.id))

        
    def pay(self, course):
        print("我为{}交了1000块钱的学费".format(course.course))
        self.course_list.append(course.course)
    
    def praise(self, obj):
        print("{} 觉得{}的上课方式非常有趣".format(self.name, obj.name))
    
    def mark_check(self):
        for item in self.mark_dict.items():
            print(item)
            
    def out(self):
        print("我退出系统了")
      
    
tea_num_id = 0
class Teachers(School_members):
    def __init__(self, name, age, role, course):
        super(Teachers, self).__init__(name, age, role)
        global tea_num_id
        tea_num_id += 1
        tea_id = course.school_name + "-T" + str(course.code) + str(tea_num_id).zfill(2)
        self.id = tea_id
        
    def introducing(self, course):
        print("大家好,我是你们{}课程的{} {}".format(course.course, self.name, self.role))
        
    def record_mark(self, Date,  obj, course, level):
        obj.mark_dict[course.course+'-Day'+Date] = level

# 整体流程如下：
# 首先宣布开课
Python = Grade("BJTL-Academy", '02', 'python')
Linux = Grade("CDYL-Academy", '01', 'linux')

# 招募老师并录入系统
zhang_Teh = Teachers("zhang", 37, 'teahcer', Python)
zhou_Teh = Teachers("zhou", 35, 'teacher', Linux)
Python.hiring(zhang_Teh)
Linux.hiring(zhou_Teh)
# 老师自我介绍
zhang_Teh.introducing(Python)
zhou_Teh.introducing(Linux)
# 招募学生
xiaoming = Students("xiaoming", 17, "student", Python)
xiaohong = Students("xiaohong", 24, "student", Linux)
Python.enrolling(xiaoming)
Linux.enrolling(xiaohong)

# 学生自我介绍
xiaoming.introducing(Python)
xiaohong.introducing(Linux)

# 老师打分
zhang_Teh.record_mark("04", xiaoming, Python, "A")
zhou_Teh.record_mark("03", xiaohong, Linux, "A+")
xiaoming.mark_check()
xiaohong.mark_check()
xiaoming.praise(zhang_Teh)
```

```python
class Ticket():
    def __init__(self, weekend=False, child=False):
        self.exp = 100
        if weekend:
            self.inc = 1.2
        else:
            self.inc = 1
        if child:
            self.discount = 0.5
        else:
            self.discount = 1
    
    def price(self, num):
        return self.exp * num * self.inc * self.discount
    
child = Ticket(child=True)
child.price(2)
adult = Ticket(weekend=True)
adult.price(2)
```

```python
import random 

class Turtle():
    def __init__(self):
        self.power = 100
        self.x = random.randint(0,10)
        self.y = random.randint(0,10)
        
    def move(self):
        new_x = self.x + random.choice([1, -1, 2, -2])
        new_y = self.y + random.choice([1, -1, 2, -2])
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else: 
            self.x = new_x
        
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        self.power -= 1
            
        return (self.x, self.y)
    
    def eat(self):
        self.power += 20
        if self.power > 100:
            self.power = 100
            
            
class Fishes():
    def __init__(self):
        self.x = random.randint(0,10)
        self.y = random.randint(0,10)
        
    def move(self):
        new_x = self.x + random.choice([1,-1])
        new_y = self.y + random.choice([1,-1])
        
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else: 
            self.x = new_x
        
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        return (self.x, self.y)
    
turtle = Turtle()
fishes = []
for i in range(10):
    fish = Fishes()
    fishes.append(fish)
    
while True:
    pos = turtle.move()
    
        
    if not turtle.power:
        print("乌龟力气被用光了，游戏结束")
        break
        
    
    # 在迭代器中删除列表的元素是非常危险的，容易出错，因为迭代器是直接引用列表的元素做操作
    # 所以这里把列表拷贝一份给迭代器，然后再对原列表做删除操作
    for fish in fishes[:]:
        if fish.move() == pos:
            turtle.eat()
            fishes.remove(fish)
            print("有一条鱼儿被吃掉了")
            if not len(fishes):
                print("鱼儿被吃光了，游戏结束")
                break
```
