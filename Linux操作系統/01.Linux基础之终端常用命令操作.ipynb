# 用户相关操作
- 添加一个用户：useradd(管理员)
    - 用法：useradd 用户名
- 设置用户的登录密码：passwd
    - 用法：passwd 用户名
- 查看当前的登录用户名：
    - whoami
- 切换用户：su
    - 用法： su 用户名，直接su的话是切换到root 
    - 特点：管理用户切换到普通用户不需要密码，但是反过来需要
- 如果是管理员登录状态则前面是#符号，如果是普通用户则是美元符号

- 删除用户：userdel
    - 用法1：在root目录下：userdel 用户名
    - 用法2：在其他用户目录下；sudo userdel -r 用户名
        - 前提是该用户已经经过权限提升
- 创建组：groupadd
    - groups:输出当前用户所属组的所有组员
    - gpasswd -a:为组添加用户，
        - 用法：gpasswd -a 用户 组名
    - gpasswd -d：为组删除用户
    - 组的默认信息：/etc/login.defs

# Linux常用命令
## 查看目录文件信息
- pwd:查看文件和目录
    - 命令格式： 命令 -选项 -参数（目录）
- ls：查看当前路径下的文件
    - d:只看当前目录的信息
    - l：看详细信息
    - a:显示所有文件(包括隐藏文件)，任何一个文件前面加上“.”就表示该文件是隐藏文件，意味着该文件很重要，用户不要轻易修改
    - h:显示文件大小
    - 以上命令可以任意组合：如ls -lh, ls al

## 路径切换
- cd：路径的切换
    - -:返回到之前的目录
    - ..:返回上一级目录
    - ~:到用户所在的家目录
    - /:回到根目录
- 绝对路径和相对路径：绝对路径指完整路径
- 创建一个文件：touch

# Linux文件处理
## 判断是文件还是目录
- drwxr-xr-x：关注最前面的那个字母
    - d：表示该文件是目录
    - -：表示文件
    - l：表示连接文件：软连接，即快捷方式
    - b：设备文件：提供存储的接口设备：U盘等
    - c：设备文件：提供串行的接口设备——键盘，鼠标

## 文件处理
- 创建文件：
    - touch 文件名及后缀
- 复制文件：cp 
    - 用法：复制一个文件：cp old_file new_file 
    - 如果目标文件已存在，复制时会将目标文件替换，为避免误操作，可以添加-i
    - 复制多个文件,必须要先建立一个目录：cp file1 file2 dir
- 移动/重命名文件：mv
    - 用法：重命名：mv file1 file2，若file2不存在，意味着将file1重命名为file2
    - 移动： mv file1 file2，若file2存在，意味着将file1移动至file2中
- 删除文件：rm
    - 用法：rm file1 file2 file3
    - 若在管理员下删除，会询问是否删除，为了避免询问，可使用强制删除：rm -f file 
- 创建目录：mkdir
    - 用法1： mkdir dir
    - 用法2：mkdir old_dir/new_dir:在已经存在的目录下递进下一层创建新的目录
    - 用法3：mkdir -p new_dir1/new_dir2:在新的目录下递进创建新的目录
- 删除目录：rm 
    - 删除一个文件：rm 文件名
    - 正常删除一个文件夹：rm -r 文件夹名
    - 若目录中含有文件或子目录，则无法正常删除，需要强制删除：rm -fr dir/*:表示删除该目录下的所有文件（不包含隐藏文件）和目录
    - rm -fr dir/.* 删除该目录下的所有文件和目录，包括隐藏文件
- 复制目录：cp
    - 用法：cp dir
    - 用法2：cp -r dir:递归复制，若复制的目录中有文件或目录，则需加r
- 移动目录：mv
    - 用法1：重命名：mv dir1 dir2:若要移动至的目录不存在，则给该目录重命名为dir2
    - 用发2：移动：mv dir1 dir2，若要移动至的目录存在，则将dir1移动至dir2中


# 查看文件
- cat（内容全部显示）：查看文件
    - -A：整合命令，相当于bET三个命令的叠加
    - -b：列出行号，但是空白行不显示行号
    - -E：将结尾的断行字符（美元符号）显示出来
    - -n：列出行号，空白行会显示行号
    - -T：将Tab键以^t显示
    = -v：列出一些看不出来的特殊字符
- less（一页一页翻动）：查看文件
    - 空格：向下翻动一页
    - pagedown：向下翻动一页
    - pageup：向上翻动一页
    - 按n：重复前一个搜索
    - 按N：反方向重复前一个搜索
    - 按q：退出
    - less -S：行过长时，将舍弃过长部分
- more（内容一页一页显示）：查看文件
    - 空格：向下翻一页
    - 回车：向下翻一行
    - /字符串：搜索想找的字符串
    - :f ：立刻显示文件名和行数
    - 按b:翻到第一页
    - 按q: :离开
- nl：显示的时候，顺便出现行号
    - -ba：无论是否有空行，都列出行号
    - -bt：如果有空行，则不列出行号
    - -nln：行号在屏幕最左方显示，前面不加0
    - -nrn：行号在屏幕最右方显示，前面加0
    - -w num：缩进num位
- tac：从最后一行开始显示，与cat刚好相反，从尾到头显示
- head：显示文件前几行 -n：设置行数
    - 常用字看日志文件
- tail：显示文件的最后几行， -n：设置行数
- wc：显示文件的行数，数字，字节
    - c:只显示字节
    - w：只显示字数，字是指空白，跳格或换行字符分隔的字符
    - l:只显示行

# 文件处理-软链接
- 软连接相当于Windows下的快捷方式
- 特点：
    - 1.可以对不存在的文件创建软链接
    - 2.可以对文件或者目录创建软链接
    - 3.删除了软链接，不会影响他的指向文件
    - 4.删除了指向文件，该软链接就会变成死链接
    
- 创建 方法
    - 创建软链接： ln -s oldfile newfile
    - 创建硬链接： link oldfile newfile / ln oldfile newfile
    
# 文件创建—硬链接
- 文件有一个inode编号：该编号是给机器看的
- 文件名是给人看的
- 注意：
    - 硬链接：多个文件对应同一个inode编号，当对多个文件中的某一个文件进行修改时，其他文件也会被修改，多用户协同一致工作
- 特点：
    - 1.不能对目录进行硬链接的创建
    - 2.只能对已经存在的文件进行硬链接的创建
    - 3.删除一个硬链接，不影响其他相同inode编号的文件
- ls -l -i dir：查看目录下文件的硬链接有多少个

# 关机
- 注意事项
    - 观察系统的使用状态
    - 通知在线使用者关机的时间
    - 正确使用关机指令
    
- 关机的指令：
    - 数据同步写入磁盘命令：sync
    - 常用的关机指令：shutdown
    = 重启，关机：reboot， halt， poweroff
    
## Shutdown
- t：添加秒数，几秒后关机
- k:不是真关机，而是发出警告信息
- r:在系统服务都停止后，重启
- h：在系统服务都停止后，关机
- f：关闭并且开机以后，强行略过磁盘检查
- F：重启后，强制进行磁盘检查
- c:取消已经进行的shutdown指令

- 服务等级
    - -init0：关机
    - -init6 重启
    - -init3 纯文本模式
    - -init5 含有图形接口
