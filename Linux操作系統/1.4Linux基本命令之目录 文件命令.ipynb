# Linux文件

## 判断是文件还是目录
- drwxr-xr-x：关注最前面的那个字母
    - d：目录文件
    - -：普通文件
    - l：表示链接文件：软连接，即快捷方式
    - b：块设备文件：提供存储的接口设备：U盘等
    - c：字符设备文件：提供串行的接口设备——键盘，鼠标
    - p：管道文件
    - s：套接字文件
    
## 路径切换
- cd：路径的切换
    - -:返回到之前的目录
    - ..:返回上一级目录
    - ~:到用户所在的家目录
    - /:回到根目录
- 绝对路径和相对路径：绝对路径指完整路径


## 查看目录下文件信息
- pwd:查看文件和目录
    - 命令格式： 命令 -选项 -参数（目录）
- ls：查看当前路径下的文件
    - d:只看当前目录的信息
    - l：看详细信息
    - a:显示所有文件(包括隐藏文件)，任何一个文件前面加上“.”就表示该文件是隐藏文件，意味着该文件很重要，用户不要轻易修改
    - h:显示文件大小
    - 以上命令可以任意组合：如ls -lh, ls al
    - 更多命令查看:ls --help



## 操作文件命令(创建，移动，删除等）
- 创建文件：
    - touch 文件名及后缀
- 复制文件：cp 
    - 用法：复制一个文件：cp old_file new_file 
    - 如果目标文件已存在，复制时会将目标文件替换，为避免误操作，可以添加-i
    - 复制多个文件,必须要先建立一个目录：cp file1 file2 dir
- 移动/重命名文件：mv
    - 用法：重命名：mv file1 file2，若file2不存在，意味着将file1重命名为file2
    - 移动： mv file1 file2，若file2存在，意味着将file1移动至file2中
- 删除文件：rm
    - 用法：rm file1 file2 file3
    - 若在管理员下删除，会询问是否删除，为了避免询问，可使用强制删除：rm -f file 
- 创建目录：mkdir
    - 用法1： mkdir dir
    - 用法2：mkdir old_dir/new_dir:在已经存在的目录下递进下一层创建新的目录
    - 用法3：mkdir -p new_dir1/new_dir2:在新的目录下递进创建新的目录
- 删除目录：rm 
    - 删除一个文件：rm 文件名
    - 正常删除一个文件夹：rm -r 文件夹名
    - 若目录中含有文件或子目录，则无法正常删除，需要强制删除：rm -fr dir/*:表示删除该目录下的所有文件（不包含隐藏文件）和目录
    - rm -fr dir/.* 删除该目录下的所有文件和目录，包括隐藏文件
- 复制目录：cp
    - 用法：cp dir
    - 用法2：cp -r dir:递归复制，若复制的目录中有文件或目录，则需加r
- 移动目录：mv
    - 用法1：重命名：mv dir1 dir2:若要移动至的目录不存在，则给该目录重命名为dir2
    - 用发2：移动：mv dir1 dir2，若要移动至的目录存在，则将dir1移动至dir2中


# 查看文件命令
- cat（内容全部显示）：查看文件
    - -A：整合命令，相当于bET三个命令的叠加
    - -b：列出行号，但是空白行不显示行号
    - -E：将结尾的断行字符（美元符号）显示出来
    - -n：列出行号，空白行会显示行号
    - -T：将Tab键以^t显示
    = -v：列出一些看不出来的特殊字符
- less（一页一页翻动）：查看文件
    - 空格：向下翻动一页
    - pagedown：向下翻动一页
    - pageup：向上翻动一页
    - 按n：重复前一个搜索
    - 按N：反方向重复前一个搜索
    - 按q：退出
    - less -S：行过长时，将舍弃过长部分
    - 在查看时，/表示向下查询，?表示向上查询相应的字符串，n表示向下继续查找下一个，N表示向上继续查找下一个
        - 比如/ set,表示向下在文件中查找set
- more（内容一页一页显示）：查看文件
    - 空格：向下翻一页
    - 回车：向下翻一行
    - /字符串：搜索想找的字符串
    - :f ：立刻显示文件名和行数
    - 按b:翻到第一页
    - 按q: :离开
- nl：显示的时候，顺便出现行号
    - -ba：无论是否有空行，都列出行号
    - -bt：如果有空行，则不列出行号
    - -nln：行号在屏幕最左方显示，前面不加0
    - -nrn：行号在屏幕最右方显示，前面加0
    - -w num：缩进num位
- tac：从最后一行开始显示，与cat刚好相反，从尾到头显示
- head：显示文件前几行 -n：设置行数
    - 常用字看日志文件
- tail：显示文件的最后几行， -n：设置行数
- wc：显示文件的行数，数字，字节
    - c:只显示字节
    - w：只显示字数，字是指空白，跳格或换行字符分隔的字符
    - l:只显示行
    
## 查找文件
- locate：根据文件名特征查找文件最快捷的命令是locate，是在数据库中搜索文件名关键字
- whereis:查找的是可执行程序或者手册页，查找范围小，但很快
- find：查找范围更广，可以使用选项进行过滤。不知道命令如何使用，使用man find即可。
