# Linux用户
- 如果是管理员登录状态则前面是#符号，如果是普通用户则是美元符号
- 账户和组名：给人看的
-  UID和GID:linux看的
- 命令id：查看当前登入用户的UID和GID
- 当我们使用ls -l 的时候，系统会根据/etc/passwd和/etc/group文件的内容，找到UID和GID对应的名称，进行显示
- 如果随意修改/etc/passwd的用户的UID会发生什么情况
    - 如果系统没找到对应的UID，那么就直接会显示那个UID/或GID数字
- /etc/passwd:记录了Linux上所有的账号
- /etc/shadow:记录了账户对应的密码
- /etc/group:记录了所有的组

## /etc/passwd
- 格式：
    - 用户名:密码:用户id:组id:用户信息:用户主目录:用户环境
- liutao:x:10001:1001:,,,:/home/liutao:/bin/bash
- 解释
    - liutao：用户名
    - x:早期这个部分放的是用户登录密码，现在密码放入了/etc/shadow中
    - UID:0表示系统管理员，1-999保留给系统使用的ID，1000以上给一般使用者
    - GID:0表示系统管理员，1-999保留给系统使用的ID,1000以上给一般使用者
    - ,,,:使用者信息说明
    - /home/liutao:用户的家目录，用户登录时，所在的目录
    - /bin/bash：用户在登录的时候，是否可以使用shell，如果不能使用shell，则会显示/sbin/nologin
    
## /etc/shadow
- 样式：
    - liutao:一串很长的字符:18666:0:99999:7:::
- 解释
    - 1.liutao：用户名
    - 2.一串字符：经过加密的密码
    - 3.最近更改过密码的日期：Linux中的日期，是通过1970年1月1号开始累加的日期
    - 4.密码不能修改的天数：0表示可以随时修改
    - 5.密码需要重新被修改的天数：通过修改值，可以强制用户修改密码
    - 6.密码需要变更的告警天数：78天内系统会向用户发出告警
    - 7.密码到期后，账号还可以使用的时间
    - 8.账号失效日期：通过1970年1月1日开始累加的日期，到了时间后，无论密码是否过期，该账号就不能使用了
    - 9.保留
  
## /etc/group
- 格式：
    - 组名:密码:组id:组员列表（用逗号分隔)
- 组密码也放在/etc/gshadow中
- liutao:x:1001:
    - 1.liutao:组名
    - 2.x：组密码，一般不需要
    - 3.1000：GID
    - 4.附属组

- 以上这些文件不建议直接修改，建议使用命令进行修改

# Linux用户操作命令
- 添加，删除，修改，切换用户

## useradd添加用户
- useradd命令
    - -u：设置UID
    - -g:设置主要组
    - -G：设置附属组
    - -c：设置用户说明
    - -d：指定用户家目录
    - -s：指定用户shell：比如useradd -s /sbin/nologin 用户名
    - -e：账号失效日期，格式为YYYYJ-MM-DD
    - -f：指定密码是否失效，0表示立刻失效，1表示永不失效
    - -m: 自动设置用户的家目录到/home目录下，比如 useradd -m liutao，则home/liutao
    
#### 使用了useradd后，会默认做以下几个操作
- 1.在/etc/passwd下建立相关的资料
- 2.在/etc/shadow下写入密码的相关的参数
- 3.在/etc/group中写入和账号名一样 的组
- 4.在/home下创建用户的家目录


## userdel删除用户
-  要彻底的删除一个用户，比较麻烦
- 1.删除/etc/passwd;/etc/shadow文件中的 内容
- 2.删除/etc/group;/etc/gshadow
- 3.删除/home/username;、/var/spool/mail/username
- 4.删除该用户曾经创建的文件
    - find进行搜索，先搜索，再使用userdel
- -r:连同家目录和邮箱一起删除


## usermod修改用户
- usermod可以对用户账户的信息进行细微的修改
    - -c：账号说明
    - -d：修改家目录
    - -g:修改主要组
    -  -G：修改附属组
    - -a:与-G一起使用，增加附属组
    - -l：修改用户名
    - -u：修改UID
    - -L:冻结账号
    - -U：解冻
    
## su切换用户
- su：直接切换到root
- su username

# 用户密码
- passwd 用户名
    - 所有人都可以通过该命令来修改自己的密码
    - -l:锁住该账号,在/etc/shadow中放密码的位置加个!,锁住之后就不能登录
    - -u:解锁
    - -S：显示账号的密码参数
    - -n 天数：设置多久可以不修改密码
    - -x 天数：设置多久内必须修改密码
    - -w 天数： 设置密码过期前警告天数
    - -i 天数：设置密码失效天数
    - -d：没有密码也不能登录
    
# 密码失效设置
- chage
- chage可以修改和密码有关的时间参数 
- -l：查看一个账户和密码有关的时间参数
- -d YYYY-MM-DD：修改最近一次更改密码的时间
    - 如果使用的是0，表示用户登入后必须修改密码，然后强制使用新的密码登录
    - 即chage -d 0 用户名
- -E YYYY-MM-DD：修改账号的失效时间
- -l 天数：设置密码几天后失效
- -m 天数：设置密码至少保留几天
- -M 天数：设置密码多久后，需要更新
- -W 天数：设置密码过期前警告时间

# 用户组group
- groups:输出当前用户所属组的所有组员
- 创建组groupadd
    - groupadd 组名
    - 实际上就是/etc/group的改写
    - -g：修改组id
- 删除组：groupdel
    - groupdel 组名
- 切换组:newgrp
    - newgrp 组名
- 修改组groupmod
    - groupmod
- 修改组成员
    - gpasswd -a:为组添加用户，
        - 用法：gpasswd -a 用户 组名
    - gpasswd -d：为组删除用户
    - 组的默认信息：/etc/login.defs
