# shell脚本
- shell既是命令行的交互界面，也是一种编程语言，用shell编写的程序叫做脚本程序。shell在linux系统中承担着不可或缺的作用。
- Linux最常使用的shell工具是bash
- 与bash相关的配置文件是个人用户目录下的.bash_profile，.bash_login, .bash_logout和.bashrc。
- .bash_profile仅仅在登录bash shell时候执行，使用bash打开一个窗口，则会执行.bashrc，在.bashrc中使用export导出的变量会在子进程中被使用，
- .bash_logout用来清除用户生成的临时文件，抹去用户的痕迹

# shell变量
- 命名变量
    - 变量无需特别声明，变量赋值直接在命令行操作，等号两边不用加空格，变量允许多行输入，使用>。字符串中的多个空格会被当做一个处理
    - 变量的值都是字符串！！！
        - 如果想让变量进行数值运算，要使用$(($var))
    - 使用变量前面需要加上$符号，为了醒目，还可以给变量加上大括号
    - 
    - 变量支持导出功能（即使用export），导出的变量可在bash的子进程中使用
- 声明变量：declare       $
    - -a 声明一个数组

    - -i 声明一个整型

    - -f 打印所有函数定义

    - -F 仅打印函数名字

    - -r 声明一个readonly变量，该变量的值无法改变，并且不能为unset

    - -x 和export的作用类似
- 还可以声明数组变量

# shell脚本文件
- 以.sh后缀名结尾，则是脚本文件
- 脚本文件的运行有以下3种方式：
    - source命令：
        - source 脚本文件
    - .+空格+脚本文件
    - bash 脚本文件（也可用sh）
