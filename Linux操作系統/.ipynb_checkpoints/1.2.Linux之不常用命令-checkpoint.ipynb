# 关机
- 注意事项
    - 观察系统的使用状态
    - 通知在线使用者关机的时间
    - 正确使用关机指令
    
- 关机的指令：
    - 数据同步写入磁盘命令：sync
    - 常用的关机指令：shutdown
    = 重启，关机：reboot， halt， poweroff
    
## Shutdown
- t：添加秒数，几秒后关机
- k:不是真关机，而是发出警告信息
- r:在系统服务都停止后，重启
- h：在系统服务都停止后，关机
- f：关闭并且开机以后，强行略过磁盘检查
- F：重启后，强制进行磁盘检查
- c:取消已经进行的shutdown指令

- 服务等级
    - -init0：关机
    - -init6 重启
    - -init3 纯文本模式
    - -init5 含有图形接口

### 环境变量PATH
- shell中有很多预定义的环境变量，在shell中执行命令时，有些命令会将这些环境变量作为自己的默认参数，利用命令env可以显示所有的环境变量
```
Jared@Jared-MacBook-Pro ~ % env
TMPDIR=/var/folders/z1/t_4y23sx7nb9381bkb867py00000gp/T/
__CFBundleIdentifier=com.apple.Terminal
XPC_FLAGS=0x0
TERM=xterm-256color
SSH_AUTH_SOCK=/private/tmp/com.apple.launchd.XpVJq5xxhO/Listeners
XPC_SERVICE_NAME=0
TERM_PROGRAM=Apple_Terminal
TERM_PROGRAM_VERSION=440
TERM_SESSION_ID=00AA5C22-8A09-4208-AB27-49FF051060BB
SHELL=/bin/zsh
HOME=/Users/jared
LOGNAME=Jared
USER=Jared
PATH=/Users/jared/.nvm/versions/node/v8.11.3/bin:/usr/local/bin:/usr/bin:/bin:/usr/sbin:/sbin:/Applications/VMware Fusion.app/Contents/Public:/Library/Apple/usr/bin:/Applications/Postgres.app/Contents/Versions/11/bin
SHLVL=1
PWD=/Users/jared
OLDPWD=/Users/jared
NVM_DIR=/Users/jared/.nvm
NVM_CD_FLAGS=-q
NVM_BIN=/Users/jared/.nvm/versions/node/v8.11.3/bin
PKG_CONFIG_PATH=/usr/local/opt/python@3.7/lib/pkgconfig
LANG=zh_CN.UTF-8
_=/usr/bin/env
Jared@Jared-MacBook-Pro ~ % 
```
- PATH是一个很重要的环境变量，是一组目录的列表，目录之间用冒号“:”分隔，只有在这些列表目录下的程序才可以直接用程序名执行，否则必须明确指定程序的目录
- 如果想把有些目录设置为环境变量可使用命令：
    ```
    PATH=目录:$PATH
    PATH=.:$PATH  这表示将当前目录设置为环境变量
    ```

- 设置别名的命令：
    ```
    alias 
    unlias   取消别名
    alias ls='ls --color=auto'
    ```
    
- 特殊符号
    - 分号(;):分隔多条命令：uptime;sleep 3;uptime
    - 反斜线(\):可表示续行符，当一行很长放不下；同时他也是转义符
    - 反引号(`):它之间的字符串会被当做命令执行，虽然可以使用，但不建议，使用()代替
    - 括号():实现命令的嵌套
    - 大括号{}:花括号明确变量名，和Python中的html模板中变量的使用很像
        ```
        var='hello'
        echo ${var}world                       $
        ```
- stat命令：记录了文件的三个时间信息，最后访问时间，最后修改时间，最后状态时间
