# 进程
- 定义：已经启动的可执行程序的运行实力
- 1.PID:进程的ID(每一个新进程都有一个唯一的PID)
- 2.PPID:父进程的ID
- 3.任何已个进程都可以创建一个子进程
- 4.在Redhat 7以上，所有进程的父进程：systemd
    - 在Redhat 5，6中，所有进程的父进程 ：init

# 查看进程
>最常用的组合命令
- **ps -aux | grep 名称**
    - 查看所有的进程
    -a：显示所有终端机下执行的程序
    -u: 用户指定，默认是当前用户
    -x: 显示所有程序，不以终端机来区分
- **ps -ef | grep 名称**
    - 和ps -aux基本一样，会多显示一个父进程id
    -e:显示所有程序
    -f:显示UID,PPIP,C与STIME栏位
- **pstree:为了更好的展示进程之间的父子关系，可采用树结构来查看**
    - pstree -pu
    -p:显示父id
    -u：指定用户，默认当前用户


- 命令
    - ps：用于显示当前进程状态
    - man ps：查看ps相关的所有参数
    - 常用选项：
        - -aux：列出所有进程
        - -ef：列出所有进程，包含了父进程的id
        - -l：列出和当前用户有关的进程
        - -u 用户： 查看某一 用户的进程装态
        
    - top：可以查看实时的进程状态
    - ps aux|grep redis:查看某一任务的进程是否启动

### ps-aux参数 的解释
- 1.%CPU:占用的CPU使用率
- 2.%MEME:占用的记忆体使用率
- 3.VSZ:占用的虚拟记忆体大小
- 4.RSS:占用记忆体大小
- 5.TTY:终端的次要装置号码（minor device number of ttv），与远程登录有关
- **6.STAT:该进程的状态
- 7.START:进程开始时间
- 8.TIME:执行的时间
- 9.COMMAND:所执行的指令


## 进程的状态由两部分组成

### 进程的运行状态
- 运行中：
    - R：该进程正在运行或等待运行
- 睡眠：
    - S：正在休眠但是可以被唤醒
    - D：正在休眠 且不可以被唤醒，如果该进程被中断，可能会导致设备的异常·状态
    - K：正在休眠，而且不可以被唤醒，该进程可以被中断
- 已停止
    - T：进程被停止，但是 可以通过其他进程来恢复 
    - T：正在被调试的进程
- 僵停
    - Z：子进程在退出时向父进程发出信号，除PID外，所有资源全部释放
    - X：父进程获取了子进程的结构，子进程可以完全释放，该状态进程是看不到的
    
### 进程的优先级
- <:高优先级
- n： 低优先级
- s：包含子进程
- +：位于后台的进程组
