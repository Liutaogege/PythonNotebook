# 五、MySQL数据库中的数据的使用 
- 如何操作：编程语言
    - 1.SQL概述
        - 结构化查询语言，可以用于查询和修改数据库的语言
        - SQL语言，是数据库的核心语言，被ISO采纳为数据库语言的国际标准几乎所有的数据库都支持SQL
    - 2.组成
        - DDL:（数据定义语言，Data Definition Language）:用户管理和定义数据对象，包括数据库，数据表等
            - create:创建数据库和表
            - drop:删除表对象
            - alter:修改表达结果
        - DML:（数据操作语言，Data Manipulation Language）:::用于从操作数据库对象中所包含的数据
            - insert：添加数据
            - delete：删除数据
            - update：修改数据
        - DQL:(数据查询语言， Data Query Language): 用用户查询数据库对象中所包含的数据
            - select 查询数据
        - TCL:(事务控制语言， Transaction Control Language）：包括提交事务，事务回滚
            - commit：提交事务
            - rollback：事务回滚
            - savepoint：事务节点
        - DCL:(事务控制语言， Data Crontrol Language):用于管理数据库的，包括管理权限以及数据更改
            - grant：授权
            - revoke：收回权限
            
### 1.数据定义语言（DDL）:
#### 数据库的操作
- （1）：创建数据库：
    - 语法格式：create database [if not exists] 数据库名称 [character set 编码名称]
    - 例如：
            create database test_db;
            create database if not exists test_db;  #忽视警告，显示警告
            show warnings;  #查看警告
            create database test_db character set gbk;
- (2):查询数据库
    - 语法格式：show databases;  #查看所有的数据库
        - show create database 数据库名称;  #查询某一个已经创建的数据库
- （3）：修改数据库
    - 语法格式：alter database [if not exists] 数据库名称 [character set 新的编码名称]
    - 例如：alter database test_db character set utf8;

- （4）：删除数据库
    - 语法格式：drop database 数据库名称;
    
#### 数据表的操作
- （1）：创建数据表
    - 使用一个数据库 ：use 数据库名称
    - 查看当前正在使用的数据库：select database();
    - 创建数据表语法：
            create table 表名(
                字段名 字段类型  约束,
                字段名 字段类型  约束,
                .....
            );
    - 例如：
            create table user(
                uid int,
                username varchar(32),
                password varchar(32),
                birthday date,
                sex   enum('男', '女')
            );
- （2）：查询数据表
    - show tables;  显示当前数据库中所有的数据表
    - show columns from 表名;  查询某张表表结构
    - desc 表名;   查询某张表表
- （3）：修改数据表
    - 修改表名
        - 方式一：alter table 表名 rename [as|to] 新表名;
            - 例如：alter table user rename new_user;
        - 方式二：rename table 旧表名 to 新表名
            - 例如: alter table user to new_user;
    - 添加列
        - alter table 表名 add [column] 列名 列类型定义 [first|after 列名];
        - 例如：alter table user add age int after username;
        - alter table user add age int;  默认是添加到最后
        - after table user add phone varchar(11) first;
    - 修改列定义
        - 语法格式：alter table 表名 modify 列名 列定义 [first|after 列名];
        - 例如：alter table user modify phone varchar(32);
    - 修改列名
        - 语法格式：alter table 表名 change 旧列名 新的列名 新的列定义 [first|after 列名];
        - 例如：alter table user change phone telephone varchar(32);
    - 删除列
        - 语法格式：alter table 表名 drop [column] 列名;
        - 例如：alter table user drop age;
- （4）：删除数据表
    - 语法格式：drop table 表名;
    - 例如：drop table user;

### 2.数据操作语言（DML）：
- 用于操作数据库对象中所包含的数据
- 基本查询数据的语句：select * from 表名;
- 1.插入操作
    - 语法格式：
        - insert [into] 表名 [(列名1, 列名2, ...)]  values|value (数值1|表达式, 数值2|表达式);
    - 例如：  
        - insert into user (telephone, uid, username, password, sex) values ("0852", 3345, "liutao", "123456", '男');
- 2.删除操作 
    - 语法格式： delete from 表名 [where 条件];
    - 例如：
        - delete from user where uid=3345;  删除满足指定条件的
        - delete from user;  不带条件的删除，会删除所有的记录
    - 扩展：
        - truncate table 表名;  ------清空表，会清空表中所有的数据
    - 面试题：
        - delete删除和truncate有什么区别？
            - delete是一条一条的删除，不清空，auto_increment自动计数不会重新开始，truncate直接将表中所有的数据进行清空，auto_increment会自动重置，从零开始
- 3.修改操作
    - 语法格式：update 表名 set 列名1=数值,列名2=数值,....where 条件;
    - 例如：
        - update user set telephone="1398" where uid=87923;
        - update employer set address="四川成都" where address is null;

### 3.数据查询语言（DQL）(重中之重)
- 完整语法格式：
        select 表达式1|字段,..
        [from 表名 where 条件]
        [group by 列名]
        [having 条件]
        [order by 列名 [asc|desc]]
        [limit 位置,数量]
- 1.普通查询：
    - select 查询表达式;   最简单的SQL语句，是一个函数
    - select database(); select version(); select now();
- 2.条件查询
    - where 条件表达式，支持运算符和函数
    - mysql支持的运算符有：
        - =, !=, <>（也是不等于）, >, <, <=, >=
        - and, or, not
        - is null, is not null
        - between...and ...  区间查询，多少到多少之间
        - in(set);
        - like 和通配符和占位符一起使用: % _
            - %:表示0个或者多个字符
            - _:表示占位一个
            - \：转义符
            - 比如：select * from user where name like "_涛"
            - select * from user where name like "%\_%";
- 3.分组查询
    - [group by 列名] [having 条件]
    - 一般情况下结合聚合函数一起使用
    - (max(), min(), sum(), avg(), count())
    - 例如：select sex, avg(salary) from user group by sex having avg(salary)>9000;
    - 记住：分组的正确使用方式：group by 后面没有出现的列名不能出现在select和from的中间，虽然不报错，但不是分组的正确使用方式。不过聚合函数中出现的列名无所谓
- 4.排序查询
    -  语法格式：order by 列名 asc|desc;     默认升序（asc）
    - 例如：select * from user order by salary desc;
- 5.限制结果及数量的查询（分页）
    - 语法格式：
        - limit n条数;  从第一条开始取n条数据（了解即可）
        - limit start开始下标索引, count条数;  从起始位置start取count条数据（起始位置从0开始）
        - 分页公式：
            - 开始下标索引（起始位置） = （当前页-1）*每页显示条数
            
- 扩展：别名
    - select name as "姓名", salary "薪资" from user;



### 4.事务控制语言（TCL）
- MySQL事务默认是开启的
    - 事务语法：
            set autocommit = false|true;  设置事务的提交方式
            rollback;   事务回滚
            commit;    事务提交
            savepoint 节点名称;  设置回滚的节点
            rollback to 节点名称;  回滚到具体的某个节点
- 在mysql数据库中只有使用InnoDB数据库引擎的数据表或库才支持事务
- 通过事务来管理insert, update, delete语句
- 事务必须满足4个条件（ACID）:
    - 原子性：要么全部完成，要么全部不完成，不会结束在某个中间的某个环节，在执行的过程中一旦出现错误/异常，会被回调（Rollback）到事务开始前的状态，就像这个事务从来没有执行过一样
    - 一致性：事务处理前后数据保持一致
    - 隔离型：事务处理必须是独立的彼此隔离    
    - 持久性：事务对数据的修改被永久保存

- （1）为什么使用事务
    - 银行转账案例
    - 事务广泛应用：订单系统，银行系统
- （2）MySQL事务控制
    - commit（提交）， rollback（回滚）， savepoint（事务节点）
