# 安装
- MongoDB官网： https://www.mongodb.com/
- 下载页面网址：https://www.mongodb.com/try/download/community

### Linux版本的安装
- 下载tgz压缩包文件
    - https://fastdl.mongodb.org/linux/mongodb-linux-x86_64-ubuntu2004-4.4.2.tgz
- 使用 tar命令解压
    - tar -xvf mongodb-linux-x86_64-ubuntu2004-4.4.2
- 将上述解压后的文件移动到/usr/local/mongodb下面
    - sudo mv mongodb-linux-x86_64-ubuntu2004-4.4.2 /usr/local/mongodb
- 建立MongoDB下的bin目录的mongod及mongo两个程序软链接到/usr/local/bin目录下（或者/usr/bin）,以便在命令行模式下在任意目录下使用mongo命令
    - sudo ln -s /usr/local/mongodb/bin/mongod /usr/local/bin/mongod
    - sudo ln -s /usr/local/mongodb/bin/mongo /usr/local/bin/mongo
    - 这一步也可以将mongodb/bin添加为环境变量
    

### linux下MongoDB的配置文件
- Linux下没有MongoDB配置文件，需要自己新建，新建配置文件的具体步骤如下：
    - 1.将以上的安装步骤完成后，cd /usr/loca/mongodb,然后新建数据库的服务文件夹，mkdir mongoDBserver
    - 2.cd mongoDBserver， 到该目录下后，
        - 创建数据库文件夹：mkdir data
        - 创建日志文件夹：mkdir log
        - 创建配置文件夹与配置文件:mkdir etc, cd etc, vim mongodb.conf
    - 3.然后在配置文件中录入以下内容：
            # 如果只需要本机连接，bind_ip可以只绑定127.0.0.1
            - 如果可以接收其他的机器进行连接，则绑定0.0.0.0
            bind_ip=0.0.0.0
            #启用日志文件，默认启用
            journal=true
            #这个选项可以过滤掉一些无用的日志信息，若需要调试使用请设置为false
            quiet=false
            # 日志文件位置
            logpath=/usr/local/mongodb/mongoDBserver/logs/mongodb.log
            # 以追加方式写入日志,如果是false，每次启动mongodb服务后 会新建日志文件
            logappend=true
            # 是否以守护进程方式运行，即后台进程
            fork = true
            # 默认27017
            port = 27017
            # 数据库文件位置
            dbpath=/usr/local/mongodb/mongoDBserver/data
            # 启用定期记录CPU利用率和 I/O 等待
            #cpu = true
            # 是否以安全认证方式运行，默认是不认证的非安全方式
            #auth = true
            #noauth = true
    - 4.然后执行
        - mongod --config /usr/local/mongodb/mongoDBserver/etc/mongodb.conf

# Linux系统下MongoDB的启动
- 服务端：
    - mongod -f <配置文件>
    - 比如：mongod --config /usr/local/mongodb/mongoDBserver/etc/mongodb.conf
    - -f是--config的缩写
- 客户端：
    - mongo -host ip -port 端口号
    - ip默认为127.0.0.1,端口号默认为27017

- 注：可以通过mongod --help获取帮助。要先启动服务端，才能使用客户端

### windows版本的安装
- 1.下载.msi文件，下载后双击该文件，按操作提示即可
- 2.安装过程中，可以通过点击custom（自定义）按钮来设置MongoDB数据库程序安装目录，也可以使用默认目录
- 3.下一步安装：“install MongoDB compass”（勾选需要联网下载（Compass软件）），勾选了也挺好的
    - MongoDB Compass是一个图形界面管理工具，可以自己到官网下载安装 
    - 在初学者阶段，不建议安装MongoDB Compass，使用命令行在操作MongoDB数据库，以此提高编程能力
- 4.在安装过程中选择install MongoDB as service 是将mongo注册为系统服务（由此可以开机自启动及后台运行）

### windows下MongoDB的配置文件
- 在windows下有一个文件C:\Program Files\MongoDB\Server\4.2\bin\mongod.cfg
- Mongodb常用配置参数
    - bindIp: 127.0.0.1
    - dbpath:C:\Program Files\MongoDB\Server\4.2\data
    - port: 27017
    - path:  C:\Program Files\MongoDB\Server\4.2\log\mongod.log(这是日志的存储路径)
    - logAppend: true

# Windows下MongoDB数据库的启动与连接
## 配置好mongodb文件的启动步骤
- 以管理员身份运行命令提示符，然后输入net start mongodb
- 再打开新的命令提示符，输入mongo即可

## 刚刚安装时的步骤
### 启动服务
- 1.创建数据目录
    - 启动MongoDB之前，首先要给它创建一个数据存储的路径，任意位置都行。存为data文件夹
- 2.把MongoDB的bin目录加入到环境变量中，方便后期调用
- 3.cmd中执行命令  mongod --dbpath 数据存储地址  。即可启动MongoDB。或者通过cmd进入bin目录输入这语句
    - 注意：该终端启动后，不能关闭，否则服务就会断开

### 连接MongoDB
- 在环境变量设置好的前提下，使用以下命令mongo就可以进入到mongo的操作终端了
- 第一种连接方法；
    - 使用compass软件连接
- 第二种方法：
    - 按照上面的方法启动MongoDB以后，cmd输入mongo即可连接
- 简化第二种方法：
    - 因为启动MongoDB后，终端不能关闭，而且每次运行的时候还需要指定data的路径，所以进行简化
    #### 将MongoDB制作成命令提示符可调用的命令
            1、在数据库data文件夹下新建一个log文件夹，用于存放日志文件，在log文件夹下新建文件mongodb.log（log文件夹放在哪都不影响，尽量和存储数据放在一起或者放到MongoDB的安装目录下）
            2、在MongoDB的安装目录下bin的目录下找到文件mongo.cfg，在mongo.cfg并修改以下内容:
            dbpath: 存储数据库data的路径 
            path: mongodb.log的路径
            比如：dbPath: D:\MongoDBSaveData\data， path:  C:\Program Files\MongoDB\Server\4.2\log\mongod.log
            3、以管理员身份打开cmd命令框
            4、命令提示符中输入mongod --config "C:\Program Files\MongoDB\Server\4.2\bin\mongod.cfg" --install --serviceName "MongoDB" --journal或者（mongod --config "输入mongo.config的地址" --install） 
            5、在管理员cmd中输入 net start mongoDB  开启服务
            6、输入mongo可以对数据库进行操作

    - 以后要启动MongoDB在管理员cmd中输入：net start mongodb
    - 结束：net stop mongodb
    - 移除MongoDB："D:\mongodb\bin\mongodb.ext" --remove
