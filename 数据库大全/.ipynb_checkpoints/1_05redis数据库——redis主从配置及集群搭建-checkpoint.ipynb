# redis主从配置及集群搭建
- 主从复制（Master-Replica replication）
    - replicaof masterip masterport
    - 例如：replicaof 192.168.0.105 6379
- 具体步骤：
    - 首先设置 主机中redis.conf配置文件，修改bind：bind 内网ip地址
    - 在从机中设置：recplicaof 主机内网ip地址 主机使用的端口号（默认为6379）- 记得注释掉protected-mode为no


# 集群搭建
- 像单例模式的话客户端想要连接任一个服务端都需要和每个服务端建立连接
- 而集群模式的话每个服务器都有关联，只要连接其中任一个服务器就相当于连接了所有的服务器

# redis集群的搭建
- 在同一台机器搭建集群需要各个redis服务器的端口号 port也不同
    - 因为每台机器每个端口号只能被一个程序占用
- 1.copy3份以上redis配置文件，分别修改文件内的port ，例如 port6000, port 6001,port6002,并都设置cluster-enabled为yes
- 2.分别启动上面三个配置好的redis服务，启动方式也是 redis-server 配置文件， 例如redis-server redis-6000.conf
- 3.在终端直接使用redis-cli命令搭建集群
    - redis-cli --cluster create 127.0.0.1 6000 127.0.0.1 6001 127.0.0.1 6002
    
## 集群搭建注意事项
- redis集群至少需要三个节点
- 在没有从节点的情况下，只要有一个主节点挂掉，整个集群就会挂掉
- 要保证集群 的高可用，需要每个节点都有从节点，但不是必须，集群和主从配置没有必然关系
- 在同一台机器（电脑）搭建 的是伪分布式集群，即一台服务器虚拟运行6个redis服务（每个服务采用不同的端口号），在实际生产环境中，redis集群搭建在不同机器上，和在同一台机器上搭建原理完全一样，无非端口号可以不用修改。
- 连接redis集群服务：redis-cli -c -h <任一节点ip> -p 端口号
