# 初识MySQL数据库
#### 数据库（Database， DB）：
- 存储数据的仓库，只不过这个仓库是在计算机存储设备上并将数据按一定的格式存放，是长期存储在计算机内，有组织的，可共享的数据集合（文件 ）

#### 数据库管理系统（Database Management System， DBMS）
- 一种操纵和管理数据库的大型软件，用于建立、使用和维护数据库，简称DBMS，它对数据库进行统一的管理和控制，以保证数据库的安全性和完整性。用户通过DBMS访问数据库中的数据，数据库管理员也 通过DBMS进行数据库的维护工作

#### 数据库系统（Database System， DBS）
- 由数据库机器管理软件组成的系统，一般情况下，称数据库为使用的数据库软件

#### 数据库类型
- 关系数据库：二维表形式存储数据
- 非关系型数据库：NoSQL（not only SQL）
    - Redis：键值对存储，k-v形式存储
    - MongoDB：文档存储，将数据以文档的形式存储，json/XML
    - HBASE：列存储，，Hadoop Database

#### 主流的关系型数据库
- Oracle：甲骨文公司
- SQLServer：微软
    - 简单，界面友好，交互和操作性更强
    - 开放性差，性能低
- DB2；IBM
    - 伸缩性比较强，适用于所有主流的平台
    - 收费
- MySQL：MySQL AB

### mysql字段格式
- text与blob的区别以及bit,binary,varbinary
    - TEXT与BLOB的主要差别就是BLOB保存二进制数据，TEXT保存字符数据。bit,binay也是保存二进制
        - bit(M):位字段类型。M 表示每个值的位数，范围为 1～64。如果 M 被省略，默认值为 1。如果为 BIT(M) 列分配的值的长度小于 M 位，在值的左边用 0 填充。
        - binary(M):BINARY 类型的长度是固定的，指定长度后，不足最大长度的，将在它们右边填充 “\0” 补齐，以达到指定长度。
        - varbinary(M):VARBINARY 类型的长度是可变的，指定好长度之后，长度可以在 0 到最大值之间。
        - BLOB 是一个二进制的对象，用来存储可变数量的数据。
    - TINYBLOB、BLOB、MEDIUMBLOB和LONGBLOB。它们只是可容纳值的最大长度不同。
        - TinyBlob 最大 255
        - Blob 最大 65K
        - MediumBlob 最大 16M
        - LongBlob 最大 4G
        - BLOB列没有字符集，并且排序和比较基于列值字节的数值值。TEXT列有一个字符集，并且根据字符集的校对规则对值进行排序和比较
        - BLOB和TEXT列不能有默认值。

# 三、使用MySQL软件
- 1.启动服务
    - 查看MySQL是否启动：任务管理器界面查看服务界面，查找MySQL即可
    - 若没有启动，可点击其启动。或者以管理员身份运行cmd界面，输入net start mysql即可。net stop mysql关闭服务
- 2.登录数据库
    - 格式： mysql -参数
    - 参数：
        - -u 用户名
        - -p 密码
        - -P 端口号
        - -h 数据库的服务器地址（远程连接数据库）
        - --prompt 设置提示符
        - mysql -uroot -proot -P3306 -h
- 3.退出数据库
    - ctrl + c
    - exit
    - quit
    - \q
- 4.修改密码
    - 第一种方式：当你忘记密码的时候这样修改
        - （1）：停止MySQL服务
        - （2）：在dos下输入mysqld --skip-grant-tables
        - （3）：再新开一个dos登录MySQL数据库（不需要密码）：mysql -uroot -p然后回车就行
        - （4）：更改密码：
            - use mysql
            - update user set password=password("新密码") where user=用户名
            - 更改成功后关闭dos窗口并结束MySQL进程，在任务管理器结束mysqld.exe
            - 重新登录：dos界面重启MySQL服务net start mysql,然后新密码登录即可
    - 第二种方式：当你没有忘记时但想修改密码时，比较简单
        - 格式一：mysqladmin -u 用户名 -p password 新密码 (回车后输入旧密码)
        - 格式二：mysqladmin -u 用户名 -p 密码 password 新密码
- 5.修改MySQL的提示符
    - 方式一：登录成功后输入：prompt hehe
    - 方式二:mysql -u用户名 -p密码 --prompt 新的提示符
    
- 6.识别版本号
    - select version();
    
# 四、MySQL数据库数据库类型
- MySQL数据库中支持多种数据类型：数值型  ，字符型，日期型
- 常用的数据类型
    - 1.整型
        - int：整型：存储整数
        - int(M):M表示预期值，与存储大小和数值的范围无关
                id int          10000
                id int(3)  zerofill 10000
                zerofill 表示用零填充  会在左侧补0
    - 2.浮点型
        - float[(M, D)]:浮点型，存储小数
            - salary float(6,2）：表示最多6位，小数占2位，最大存储数值为9999.99
        - double：浮点数，小数
        
    - 3.字符型
        - char(M): 固定长度的字符串类型，占据固定的长度
        - varchar(M): 可变长度的字符串类型，如果长度不够，就只占用数据本身的长度
        - text：大的字符串类型，4G字符串
        - blob：字节类型
    - 4.日期
        - date：日期类型：年月日  YYYY-MM-DD;日期格式
        - time：时间类型，时分秒  HH:mm:ss  时间格式
        - year:年份     YYYY格式的年份值
        - timestamp: 时间戳
        - datetime：年月日 时分秒YY-MM-DD hh:mm:ss
    - 5整型
        - tinyint：1  -128-127
        - samllint:2  32767/65535
        - int：4  21亿
        - bigint：8   兆
        - mediumint： 3  八百多万/一千六百多万
- 选择合适的数据类型，是数据库优化的一部分，short it better原则
