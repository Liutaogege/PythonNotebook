- 5.码点和代码单元
    - 码点：是指与一个编码表中的某个字符对应的代码值，在Unicode标准中，码点采用16进制书写，并加上前缀U+
    - 代码单元：在基本的多语言级别中，每个字符用16位表示，通常被称为代码单元
    - 个人理解：码点是完整的不会改变的代码值，虽然规定码点用16进制书写，但是某些特殊字符用一个代码单元是表示不出来的，就采用连续的两个代码单元来表示，但他仍然只有一个码点
            public class Test {
            public static void main(String[] args) {
                String testing = "hi🆚";
                System.out.println(testing.length());
                System.out.println(testing.codePointCount(0, testing.length()));
                // 返回index所对应的代码单元对应的元素: charAt(index)
                System.out.println(testing.charAt(0));
                System.out.println(testing.charAt(3));
                // 返回每个index对应的码点
                int codeCount = testing.codePointCount(0, testing.length());
                for (int i=0; i < codeCount; i++) {
                    int index = testing.offsetByCodePoints(0, i);
                    int cp = testing.codePointAt(index);
                    System.out.println(cp);
                }
            }
        }

- Scanner的其他方法
    - nextLine():读取输入的下一行内容
    - next():读取输入的下一个单词（以空格作为分隔符)
    - nextInt(), nextDouble():下一个输入为int/double类型
    - hasNext():检测输入中是否还有其他单词
    - hasNextInt(), hasNextDouble():下一个字符序列是否是int/double
