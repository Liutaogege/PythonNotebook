```python
'''
【课程3.5】  蒙特卡罗模拟

蒙特卡罗（Monte Carlo）方法，又称随机抽样或统计试验方法，是以概率和统计理论方法为基础的一种计算方法
使用随机数（或更常见的伪随机数）来解决很多计算问题的方法。
将所求解的问题同一定的概率模型相联系，用电子计算机实现统计模拟或抽样，以获得问题的近似解。
 
① π的计算
② 计算积分 y = x**2
③ 排队上厕所问题

'''
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# pi 的计算
from matplotlib.patches import Circle


n = 10000\
#投点次数

r = 1
a,b = (0,0)  #圆心

xmin,xmax = a-r,a+r
ymin,ymax = b-r,b+r
#正方形

x = np.random.uniform(xmin,xmax,n)
y = np.random.uniform(ymin,ymax,n)
# 在正方形区域内随机投点
# numpy.random.uniform(low,high,size) → 从一个均匀分布[low,high)中随机采样，均匀分布

d = np.sqrt((x-a)**2 + (y-b)**2)
res = sum(np.where(d < r,1,0))
# 计算点到圆心的距离
# 统计落在圆内的点的数目

pi = 4*res /n
print('pi为：',pi)

fig = plt.figure(figsize = (8,8))
axes = fig.add_subplot(1,1,1)
plt.plot(x,y,'ro',markersize = 1)
plt.axis('equal')

circle = Circle(xy = (a,b),radius = r,alpha = 0.5,color = 'r')
axes.add_patch(circle)
plt.grid(True,linestyle = '--',linewidth = '0.8')
```

```python
#计算积分

n = 1000
xmin ,xmax = 0,1
ymin ,ymax = 0,1
#矩形区域

x = np.random.uniform(xmin,xmax,n)
y = np.random.uniform(ymin,ymax,n)

def f(x):
    return x**2

res = sum(np.where(y < f(x),1,0))
integral = res / n
print('intergral为：',integral)

fig = plt.figure(figsize = (8,8))
axes = fig.add_subplot(1,1,1)
axes.plot(x,y,'o',markersize = 1)
plt.axis('equal')

xi = np.linspace(0,1,100)
yi = xi**2
plt.plot(xi,yi,'--k')
plt.fill_between(xi,yi,0,color= 'gray',alpha = 0.5,label = 'area')
plt.grid()
```
