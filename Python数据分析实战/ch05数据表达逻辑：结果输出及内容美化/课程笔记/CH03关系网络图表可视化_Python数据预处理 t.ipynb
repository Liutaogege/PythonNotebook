```python
'''
【课程3.4】  Python数据预处理

将数据中导演与演员的关系整理出来，得到导演与演员的关系数据，并统计合作次数

'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings
```

```python
import os
df = pd.read_excel(r'C:\Users\acer\Desktop\豆瓣电影数据.xlsx',sheetname = 0,header = 0)
df.head()
```

```python
data = df[['name','导演','主演']]
data.dropna(inplace = True)
data_zy = data['主演'].str.split('/',expand = True) 
col_len1 = len(data_zy.columns)
data_zy.columns = ['zy'+str(i) for i in range(col_len1)]

data_dy = data['导演'].str.split('/',expand = True)
col_len2 = len(data_dy.columns)
data_dy.columns = ['dy'+str(i) for i in range(col_len2)]


data2 = data_dy.join(data_zy).join(data['name'])
data2.head()
```

```python
data_new = pd.DataFrame(columns = ['name','导演','演员'] )

col_dy = data_dy.columns
col_zy = data_zy.columns 

for dy in col_dy:
    for zy in col_zy:
        data_i = data2[['name',dy,zy]].dropna()
        data_i.columns = ['name','导演','演员'] 
        data_new = pd.concat([data_new,data_i])
print(data_new.head())
```

```python
result = data_new.groupby(['导演','演员']).count()
result.reset_index(inplace = True)
result.columns = ['导演','演员','合作次数']
result.head()

writer = pd.ExcelWriter(r'C:\Users\acer\Desktop\output.xlsx')
result.to_excel(writer,'sheet1')
writer.save()
print('finished')
```
