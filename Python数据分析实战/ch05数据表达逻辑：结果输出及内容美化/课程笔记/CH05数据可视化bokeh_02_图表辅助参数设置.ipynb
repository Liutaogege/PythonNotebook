```python
'''
【课程5.3】  图表辅助参数设置

辅助标注、注释、矢量箭头

参考官方文档：https://bokeh.pydata.org/en/latest/docs/user_guide/annotations.html#color-bars

'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings 
warnings.filterwarnings('ignore')

from bokeh.io import output_notebook
output_notebook()

from bokeh.plotting import figure,show
```

```python
# 辅助标注 - 线

from bokeh.models.annotations import Span
# 导入Span模块

x = np.linspace(0,20,200)
y = np.sin(x)
#创建x,y数据

p = figure(y_range = (-2,2))
p.line(x,y)
# 绘制曲线

upper = Span(location = 1,#设置位置，对应坐标值
            dimension = 'width',# 设置方向，width为横向，height为纵向
             line_color = 'red',line_width = 4#设置颜色线宽
            )
p.add_layout(upper)
#绘制辅助线1

lower = Span(location = -1,dimension = 'width',line_color = 'red',line_width = 4)
p.add_layout(lower)
#绘制辅助线2

show(p)
```

```python
# 辅助标注 - 矩形

from bokeh.models.annotations import BoxAnnotation

x = np.linspace(0,20,200)
y = np.cos(x)

p = figure(y_range=(-2,2))
p.line(x,y)

upper = BoxAnnotation(bottom=1,fill_alpha = 0.1,fill_color = 'yellow')
p.add_layout(upper)
# 绘制辅助矩形

lower = BoxAnnotation(top = -1,fill_alpha = 0.3,fill_color = 'firebrick')
p.add_layout(lower)
# 绘制辅助矩形2

center = BoxAnnotation(top = 0.6,bottom=-0.3,left = 7,right=12, # 设置矩形四边位置
                       fill_alpha=0.1, fill_color='blue' )       # 设置透明度、颜色
p.add_layout(center)

show(p)
```

```python
# 绘图注释

from bokeh.models.annotations import Label
 
p = figure(x_range=(0,10),y_range=(0,10))

p.circle([2,5,8],[3,5,8],color = 'yellow',size = 10)
# 绘制散点图

label = Label(x=5,y=5,  #标注注释位置
             x_offset = 12,  # x 偏移，同理y_offset
              y_offset = -10,
             text = '随便画的图',
             text_font_size = '12pt',
             border_line_color = 'red',background_fill_color = 'gray',background_fill_alpha = 0.4 # 背景线条颜色、背景颜色、透明度
             )
p.add_layout(label)
show(p)
```

```python
#注释箭头

from bokeh.models.annotations import Arrow
from bokeh.models.arrow_heads import OpenHead,NormalHead,VeeHead

p = figure(plot_width = 400,plot_height = 400)
p.circle(x = [0,1,0.5],y = [0,0,0.9],radius=0.2,color = ['blue','yellow','red'],fill_alpha = 0.5)

p.add_layout(Arrow(end=OpenHead(line_color='firebrick',line_width=4),#设置箭头类型
                  x_start = 0,y_start = 0,x_end=1,y_end=0))
#绘制箭头1

p.add_layout(Arrow(end=NormalHead(fill_color = 'yellow'),
                  x_start=1,y_start=0,x_end=0.5,y_end= 0.7))

p.add_layout(Arrow(end=VeeHead(size=35),line_color='red',
                  x_start=0.5,y_start=0.7,x_end=0,y_end=0))

show(p)
```
