```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
# 创建正弦函数及图表

def sinplot(flip = 1):
    x = np.linspace(0,14,10000)
    for i in range(1,7):
        plt.plot(x,np.sin(x + i *5) * (7 - i) * flip)
sinplot()
```

```python
# 1、set() 
# 简单切换为seaborn图表风格


sinplot()
```

```python
# 2、set_style()
# 切换seaborn图表风格
# 风格选择包括："white", "dark", "whitegrid", "darkgrid", "ticks"

fig = plt.figure(figsize = (8,8))
ax1 = fig.add_subplot(2,1,1)
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data = data)
plt.title('style - whitegrid')
ax2 = fig.add_subplot(2,1,2)
sns.set_style('darkgrid')
sinplot()
```

```python
# 3、despine()
# 设置图表坐标轴
# seaborn.despine(fig=None, ax=None, top=True, right=True, left=False, 
# bottom=False, offset=None, trim=False)

sns.set_style("white")
fig = plt.figure(figsize = (9,9))
plt.subplots_adjust(hspace= 0.3)

ax1 = fig.add_subplot(3,1,1)
sinplot()
sns.despine()
# 删除了上、右坐标轴

ax2 = fig.add_subplot(3,1,2)
sns.violinplot(data = data)
#sns.despine(offset=10, trim=True)
# offset：与坐标轴之间的偏移
# trim：为True时，将坐标轴限制在数据最大最小值

ax3 = fig.add_subplot(3,1,3)
sns.boxplot(data = data ,palette = 'colorblind')
#palette：调色板，控制图像的色调
#sns.despine(left=True, right = False)
# top, right, left, bottom：布尔型，为True时不显示
```

```python
# 4、axes_style()
# 设置局部图表风格，可学习和with配合的用法

with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
# 设置局部图表风格，用with做代码块区分

sns.set_style('whitegrid')
plt.subplot(212)
sinplot()
# 外部表格风格
```

```python
# 5、set_context()
# 设置显示比例尺度
# 选择包括：'paper', 'notebook', 'talk', 'poster'

sns.set_context('paper')
sinplot()
# 默认为notebook
```

```python
# 分组颜色设置 - "Paired"

sns.palplot(sns.color_palette('Paired',10))
```

```python
# 2、设置亮度，饱和度
# 可用方法：
# ① husl_palette([n_colors, h, s, l])
# ② hls_palette([n_colors, h, l, s])

sns.palplot(sns.hls_palette(9,l = 0.3,s = .9,h = .1))
# l → 亮度
# s → 饱和度
```

```python
# 3、cubehelix_palette()
# 按照线性增长计算，设置颜色

sns.palplot(sns.cubehelix_palette(8,gamma = 1))
sns.palplot(sns.cubehelix_palette(8,start = 0.1,rot = 1))
sns.palplot(sns.cubehelix_palette(8,start = 2,rot = 0,dark=0.1,light = 0.9,reverse = True))
# n_colors → 颜色个数
# start → 值区间在0-3，开始颜色
# rot → 颜色旋转角度
# gamma → 颜色伽马值，越大颜色越暗
# dark，light → 值区间0-1，颜色深浅
# reverse → 布尔值，默认为False，由浅到深
```

```python
# 4、dark_palette(color[, n_colors, reverse, ...]) / light_palette(color[, n_colors, reverse, ...])
# 颜色深浅

sns.palplot(sns.light_palette('green'))  # 按照green做浅色调色盘
sns.palplot(sns.light_palette('green',reverse = True)) 

sns.palplot(sns.color_palette('Greens'))  # cmap为Greens风格
sns.palplot(sns.color_palette('Greens'))

sns.palplot(sns.dark_palette('yellow'))# 按照yelloew做深色调色盘
sns.palplot(sns.dark_palette('yellow',reverse = True))
```

```python
# 5、diverging_palette()
# 创建分散颜色
# seaborn.diverging_palette(h_neg, h_pos, s=75, l=50, sep=10, n=6, 
# center='light', as_cmap=False)

sns.palplot(sns.diverging_palette(200,285,s = 85,l = 50,n = 8))
sns.palplot(sns.diverging_palette(200,285,s = 85,l = 50,n = 8,center = 'dark'))
# h_neg, h_pos → 起始/终止颜色值
# s → 值区间0-100，饱和度
# l → 值区间0-100，亮度
# n → 颜色个数
# center → 中心颜色为浅色还是深色“light”，“dark”,默认为light
#sep 为步长
```

```python
# 5、diverging_palette()
# 创建分散颜色
plt.figure(figsize = (9,9))
x = np.arange(25).reshape(5,5)
cmap = sns.diverging_palette(200,20,sep = 20,as_cmap=True)
sns.heatmap(x,cmap = cmap)  #热力图呈现
```

```python
# 设置调色板后，绘图创建图表

sns.set_style('whitegrid')
#设置风格

with sns.color_palette('Greens'):
    plt.subplot(211)
    sinplot()
    
sns.set_palette('husl')
plt.subplot(212)
sinplot()
```
