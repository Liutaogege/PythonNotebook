```python
# 在spyder等非notebook中创建绘图空间

from bokeh.plotting import figure,show,output_file
# 导入图表绘制、图标展示模块
# output_file → 非notebook中创建绘图空间

import os
os.chdir(r'C:\Users\acer\Desktop')
# 创建工作目录
output_file('线索.html')
# notebook绘图命令，创建html文件
# 运行后会弹出html窗口

p = figure(plot_width=400, plot_height=400)   # 创建图表，设置宽度、高度
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20, color="blue", alpha=0.5)
# 创建一个圆形散点图

show(p)
```

```python
# 创建图表工具 
# figure()

df = pd.DataFrame(np.random.randn(100,2),columns = ['A','B'])

p = figure(plot_width=600, plot_height=400,    # 图表宽度、高度
           tools = 'pan,wheel_zoom,box_zoom,save,reset,help',  # 设置工具栏，默认全部显示
           toolbar_location='above',     # 工具栏位置："above"，"below"，"left"，"right"
           x_axis_label = 'A', y_axis_label = 'B',    # X,Y轴label
           x_range = [-3,3], y_range = [-3,3],        # X,Y轴范围
           title="测试图表"                       # 设置图表title
          )
#figure创建图表
## tool参考文档：https://bokeh.pydata.org/en/latest/docs/user_guide/tools.html

p.title.text_color = "white"
p.title.text_font = "times"
p.title.text_font_style = "italic"
p.title.background_fill_color = "black"
# 设置标题：颜色、字体、风格、背景颜色

p.circle(df['A'],df['B'],size = 20,alpha = 0.5)
show(p)
```

```python
# 颜色设置

p = figure(plot_width = 400,plot_height = 400)

p.circle(df.index,df['A'],color = 'red',size = 10,alpha = 0.7)
p.circle(df.index,df['B'],color = '#800080',size = 11,alpha = 0.7)

show(p)
# 颜色设置
# ① 147个CSS颜色，参考网址：http://www.colors.commutercreative.com/grid/
# ② RGB颜色值，参考网址：https://coolors.co/87f1ff-c0f5fa-bd8b9c-af125a-582b11
```

```python
# 图表边框线参数设置

p = figure(plot_width = 400,plot_height = 400)
p.circle(df.index,df['A'],color = 'green',size =10,alpha = 0.5)
p.circle(df.index,df['B'],color = 'red',size = 11,alpha = 0.6)

p.outline_line_width = 7  #边框线宽
p.outline_line_alpha = 0.3 #边框线透明度
p.outline_line_color = 'gray'
#设置图表边框
show(p)
```

```python
#设置绘图空间背景

p = figure(plot_width = 600,plot_height = 400)
p.circle(df.index,df['A'],size = 10,alpha = 0.4)
p.circle(df.index,df['B'],size = 10,alpha = 0.5)
p.background_fill_color = 'beige'
p.background_fill_alpha = 0.5
#设置背景参数

show(p)
```

```python
#设置外边界背景

p = figure(plot_width = 400,plot_height = 400)

p.circle(df.index,df['A'],size = 10,alpha = 0.5)
p.circle(df.index,df['B'],size = 10,alpha = 0.5)

p.border_fill_color = 'black'  #外边界背景颜色
p.min_border_left = 50   #外边界背景 —— 左边宽度
p.min_border_right = 69
p.min_border_top = 10
p.min_border_bottom = 10

show(p)
```

```python
# axes - 轴线设置
# 轴线标签，轴线宽度，轴线颜色
# 字体颜色，字体角度

p = figure(plot_width = 400,plot_height = 400)
p.circle([1,2,3,4,5],[2,3,5,6,6],size = 10)

p.xaxis.axis_label = '数字顺序'
p.xaxis.axis_line_width = 5
p.xaxis.axis_line_color  = 'red'
#设置X轴线：标签，线宽，轴线颜色


p.yaxis.axis_label = '随机数字'
p.yaxis.major_label_text_color = 'orange'
p.yaxis.major_label_orientation = 'vertical'
#设置y轴线：标签，字体颜色，字体角度

p.axis.minor_tick_in = 20  #刻度往绘图区内延伸长度
p.axis.minor_tick_out = 15  #刻度往绘图区外延伸长度
#设置刻度

p.xaxis.bounds = (2,4)
#设置轴线范围

show(p)
```

```python
# axes - 轴线设置
# 标签设置

p = figure(plot_width = 400,plot_height = 400)
p.circle([1,2,3,4,5],[4,5,7,8,1],size = 10)

p.xaxis.axis_label = '顺序数字'
p.xaxis.axis_label_text_color = 'red'
p.xaxis.axis_label_text_font_size = '100'
p.xaxis.axis_label_standoff = 30
#设置标签名称，字体颜色，大小，偏移距离

p.yaxis.axis_label = '随机数字'
p.yaxis.axis_label_text_font_style = 'italic'
#设置标签名称，字体

show(p)
```

```python
# Grid - 格网设置
# 线型设置

p = figure(plot_width = 500,plot_height = 400)
p.circle(df.index,df['A'],size = 10,alpha = 0.5)
p.circle(df.index,df['B'],size = 10,alpha = 0.5)

p.xgrid.grid_line_color = 'red'
# 颜色设置，None时则不显示

p.ygrid.grid_line_alpha = 1
p.ygrid.grid_line_dash = [1,1]
# 设置透明度，虚线设置
# dash → 通过设置间隔来做虚线

p.xgrid.minor_grid_line_color = 'green'
p.xgrid.minor_grid_line_alpha = 0.1
# minor_line → 设置次轴线

show(p)
```

```python
# Grid - 格网设置
# 颜色填充

p = figure(plot_width = 500,plot_height = 400)
p.circle(df.index,df['A'],size = 10,alpha = 0.5)
p.circle(df.index,df['B'],size = 10,alpha = 0.5)

p.xgrid.grid_line_color = None
#设置颜色为空
p.ygrid.band_fill_alpha = 0.7
p.ygrid.band_fill_color = 'red'
#设置颜色填充及透明度
p.xgrid.band_fill_alpha = 0.7
p.xgrid.band_fill_color = 'green'


#p.grid.bounds = (-1, 1)
# 设置填充边界
show(p)
```

```python
# legend - 图例设置
# 设置方法 —— 在绘图时设置图例名称 + 设置图例位置

p = figure(plot_width = 500,plot_height = 400)
#创建图表

x = np.linspace(0,4*np.pi,100)
y = np.sin(x)

p.circle(x,y,legend = 'sin(x)')
p.line(x,y,legend = 'sin(x)')
# 设置line，设置图例名称

p.line(x,2*y,legend = '2*sin(x)',line_dash = [4,4],line_color = 'yellow',line_width = 5)
# 绘制line，设置图例名称

p.square(x,3*y,legend = "3*sin(x)",fill_color = None,line_color = 'green')
p.line(x,3*y,legend = '3*sin(x)',line_color = 'green')
# 绘制line，设置图例名称

p.legend.location = 'bottom_left'
#设置图例排列方向 ：vertical 或者 horizontal

p.legend.label_text_font = 'times'
p.legend.label_text_font_style = 'italic'
p.legend.label_text_color = 'red'
p.legend.label_text_font_size = '12pt'
#设置图例：字体，风格，颜色，字体大小

p.legend.border_line_width = 3
p.legend.border_line_color = 'blue'
p.legend.border_line_alpha = 0.5
#设置图例外边线：宽度，颜色，透明度
show(p)
```

```python
总结一下：
Line Properties → 线设置
Fill Properties → 填充设置
Text Properties → 字体设置

1、Line Properties → 线设置
（1）line_color，设置颜色
（2）line_width，设置宽度
（3）line_alpha，设置透明度
（4）line_join，设置连接点样式：'miter' miter_join，'round' round_join，'bevel' bevel_join
（5）line_cap，设置线端口样式，'butt' butt_cap，'round' round_cap，'square' square_cap
（6）line_dash，设置线条样式，'solid'，'dashed'，'dotted'，'dotdash'，'dashdot'，或者整型数组方式（例如[6,4]）

2、Fill Properties → 填充设置
（1）fill_color，设置填充颜色
（2）fill_alpha，设置填充透明度

3、Text Properties → 字体设置
（1）text_font，字体
（2）text_font_size，字体大小，单位为pt或者em（ '12pt', '1.5em'）
（3）text_font_style，字体风格，'normal' normal text，'italic' italic text，'bold' bold text
（4）text_color，字体颜色
（5）text_alpha，字体透明度
（6）text_align，字体水平方向位置，'left', 'right', 'center'
（7）text_baseline，字体垂直方向位置，'top'，'middle'，'bottom'，'alphabetic'，'hanging'

4、可见性
p.xaxis.visible = False
p.xgrid.visible = False
基本参数中都含有.visible参数，设置是否可见
```
