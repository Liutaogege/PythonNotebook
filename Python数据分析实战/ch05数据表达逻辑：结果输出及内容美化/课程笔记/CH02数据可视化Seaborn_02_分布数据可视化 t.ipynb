```python
'''
【课程2.4】  分布数据可视化 - 直方图与密度图

distplot() / kdeplot() / rugplot() 

'''
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

sns.set_style('darkgrid')
sns.set_context('talk')
#设置风格尺度

import warnings 
warnings.filterwarnings('ignore')
```

```python
# 1、直方图——distplot()

rs = np.random.RandomState(1)
s = pd.Series(rs.randn(100) * 100)
sns.distplot(s,bins = 10,hist = True,kde = True,norm_hist = True,rug = True,
        vertical = False,
        color = 'y',label = 'displot',axlabel = 'x')
plt.legend()
# bins → 箱数
# hist、ked → 是否显示箱/密度曲线
# norm_hist → 直方图是否按照密度来显示
# rug → 是否显示数据分布情况
# vertical → 是否水平显示
# color → 设置颜色
# label → 图例
# axlabel → x轴标注
```

```python
# 1、直方图 - distplot()
# 颜色详细设置

sns.distplot(s,rug = True,
            rug_kws = {'color':'r'},
            #设置数据频率分布颜色
             kde_kws = {'color':'k','lw':1,'label':'KDE','linestyle':'--'},
            hist_kws = {'histtype': 'step','lw':2,'alpha':0.8,'color':'gray'} )
```

```python
# 2、密度图 - kdeplot()
# 单个样本数据密度分布图
sns.kdeplot(s,shade = False, #是否填充
           color = 'r',
           vertical = False)

sns.kdeplot(s,bw=5,label='bw:0.2',
           linestyle = '-',linewidth = 1.2,alpha = 0.5)
sns.kdeplot(s,bw=20,label='bw:2',
           linestyle = '-',linewidth = 1.2,alpha = 0.5)
# bw → 控制拟合的程度，类似直方图的箱数


sns.rugplot(s,height = 0.2,color = 'k',alpha = 0.5)
# 数据频率分布图
```

```python
# 2、密度图 - kdeplot()
# 两个样本数据密度分布图

rs = np.random.RandomState(2)
df = pd.DataFrame(rs.randn(100,2),
                 columns = ['A','B'])
sns.kdeplot(df['A'],df['B'],
           cbar = True,  #是否显示颜色图例
           shade = True,  #是否填充颜色
           cmap = 'Greens',
           shade_lowest = False,# 最外围颜色是否显示
           n_levels = 10 # 曲线个数（如果非常多，则会越平滑）
           )
# 两个维度数据生成曲线密度图，以颜色作为密度衰减显示

sns.rugplot(df['A'],color = 'g',axis = 'x',alpha = 0.5)
sns.rugplot(df['B'],color = 'r',axis = 'y',alpha = 0.8)
# 注意设置x，y轴
```

```python
# 2、密度图 - kdeplot()
# 两个样本数据密度分布图
# 多个密度图

rs1 = np.random.RandomState(9)
rs2 = np.random.RandomState(8)
df1 = pd.DataFrame(rs1.randn(100,2)+2,columns = ['A','B'])
df2 = pd.DataFrame(rs2.randn(100,2)-2,columns = ['A','B'])

sns.kdeplot(df1['A'],df1['B'],cmap = 'Reds',
           shade = True,shade_lowest = False,cbar = True)
sns.kdeplot(df2['A'],df2['B'],cmap = 'Greens',
           shade = True,shade_lowest = False,cbar = True)

sns.rugplot(df1['A'],color = 'r',axis = 'x',alpha = 0.5)
sns.rugplot(df1['B'],color = 'r',axis = 'y',alpha = 0.8)
# 注意设置x，y轴

sns.rugplot(df2['A'],color = 'g',axis = 'x',alpha = 0.5)
sns.rugplot(df2['B'],color = 'g',axis = 'y',alpha = 0.8)
```

```python
# 1、综合散点图 - jointplot()
# 散点图 + 分布图

rs = np.random.RandomState(9)
df = pd.DataFrame(rs.randn(200,2),columns = ['A','B'])

sns.jointplot(x=df['A'], y=df['B'],  # 设置xy轴，显示columns名称
              data=df,   # 设置数据
              color = 'k',   # 设置颜色
              s = 50, edgecolor="w",linewidth=1,  # 设置散点大小、边缘线颜色及宽度(只针对scatter）
              kind = 'scatter',   # 设置类型：“scatter”、“reg”、“resid”、“kde”、“hex”
              space = 0.2,  # 设置散点图和布局图的间距
              size = 8,   # 图表大小（自动调整为正方形）
              ratio = 5,  # 散点图与布局图高度比，整型
              marginal_kws=dict(bins=15, rug=True)  # 设置柱状图箱数，是否设置rug
              )  
```

```python
# 1、综合散点图 - jointplot()
# 散点图 + 分布图
# 密度图

rs = np.random.RandomState(15)
df = pd.DataFrame(rs.randn(300,2),columns = ['A','B'])
# 创建数据

g = sns.jointplot(x = df['A'],y = df['B'],
                 data = df,kind = 'kde',color = 'r',
                 shade_lowest= False)
#创建密度图
#必须要先创建g，然后才能添加散点图
g.plot_joint(plt.scatter,c ='w',s = 30,linewidth = 1,marker = '+')
# 添加散点图
```

```python
# 1、综合散点图 - JointGrid()
# 可拆分绘制的散点图
# plot_joint() + ax_marg_x.hist() + ax_marg_y.hist()

sns.set_style('whitegrid')
tips = sns.load_dataset('tips')
print(tips.head())
# 导入数据

g = sns.JointGrid(x = 'total_bill',y = 'tip',data = tips)
# 创建一个绘图表格区域，设置好x、y对应数据
g.plot_joint(plt.scatter,color = 'r',edgecolor = 'y')  # 设置框内图表，scatter.
g.ax_marg_x.hist(tips['total_bill'],color = 'b',alpha = .8,
                bins = np.arange(0,60,3))
g.ax_marg_y.hist(tips['tip'],color = 'r',alpha =0.6,orientation = 'horizontal',
                bins = np.arange(1,12,1))

from scipy import stats
g.annotate(stats.pearsonr)
plt.grid(linestyle = '--')
```

```python
# 1、综合散点图 - JointGrid()
# 可拆分绘制的散点图
# plot_joint() + plot_marginals()

g = sns.JointGrid(x = 'total_bill',y = 'tip',data = tips)

g = g.plot_joint(plt.scatter,color ='g',s = 39,edgecolor = 'white')# 绘制散点图
plt.grid(linestyle = '--')

g.plot_marginals(sns.distplot,kde = True,color = 'r')  # 绘制x，y轴直方图
```

```python
# 1、综合散点图 - JointGrid()
# 可拆分绘制的散点图
# plot_joint() + plot_marginals()
# kde - 密度图

g = sns.JointGrid(x = 'total_bill',y = 'tip',data = tips)

g =g.plot_joint(sns.kdeplot,cmap = 'Reds_r')
plt.grid(linestyle = '--')

g.plot_marginals(sns.kdeplot,shade = True,color = 'y')
```

```python
# 2、矩阵散点图 - pairplot()

sns.set_style('white')

iris = sns.load_dataset('iris')
print(iris.head())

sns.pairplot(iris,
            kind = 'scatter',  # 散点图/回归分布图 {‘scatter’, ‘reg’}  
            diag_kind="hist",  # 直方图/密度图 {‘hist’, ‘kde’}
            hue="species",   # 按照某一字段进行分类
            palette="husl",  # 设置调色板
            markers=["o", "s", "D"],  # 设置不同系列的点样式（这里根据参考分类个数）
            size = 3,   # 图表大小
            )
```

```python
# 2、矩阵散点图 - pairplot()
# 只提取局部变量进行对比

sns.pairplot(iris,vars = ["sepal_width", "sepal_length"],kind = 'reg',diag_kind = 'kde',
            hue = 'species',palette = 'husl')
```

```python
# 2、矩阵散点图 - pairplot()
# 其他参数设置

sns.pairplot(iris,diag_kind='kde',markers = '+',
             plot_kws=dict(s = 50,edgecolor = 'b',linewidth=1),
             # 设置点样式
            diag_kws=dict(shade=True)
             # 设置密度图样式
            )
```

```python
# 2、矩阵散点图 - PairGrid()
# 可拆分绘制的散点图
# map_diag() + map_offdiag()

g = sns.PairGrid(iris,hue = 'species',palette = 'hls',
                vars = ['sepal_length','sepal_width','petal_length','petal_width'],# 可筛选
                )
# 创建一个绘图表格区域，设置好x、y对应数据，按照species分类

g.map_diag(plt.hist,
          histtype = 'barstacked',# 可选：'bar', 'barstacked', 'step', 'stepfilled'
           linewidth = 1,edgecolor = 'w'
          )
# 创建一个绘图表格区域，设置好x、y对应数据，按照species分类

g.map_offdiag(plt.scatter,edgecolor = 'w',s=40,linewidth = 1# 设置点颜色、大小、描边宽度
             )
# 其他图表，plt.scatter/plt.bar...

g.add_legend()
```

```python
# 2、矩阵散点图 - PairGrid()
# 可拆分绘制的散点图
# map_diag() + map_lower() + map_upper()

g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_upper(plt.scatter,color = 'y')
g.map_lower(sns.kdeplot,cmap= 'Blues_d')
```
