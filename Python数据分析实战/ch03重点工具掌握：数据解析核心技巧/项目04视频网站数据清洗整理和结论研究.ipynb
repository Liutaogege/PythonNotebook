```python
'''
【项目04】  视频网站数据清洗整理和结论研究

作业要求：
1、数据清洗 - 去除空值
要求：创建函数
提示：fillna方法填充缺失数据，注意inplace参数

2、数据清洗 - 时间标签转化
要求：
① 将时间字段改为时间标签
② 创建函数
提示：
需要将中文日期转化为非中文日期，例如 2016年5月24日 → 2016.5.24

3、问题1 分析出不同导演电影的好评率，并筛选出TOP20
要求：
① 计算统计出不同导演的好评率，不要求创建函数
② 通过多系列柱状图，做图表可视化
提示：
① 好评率 = 好评数 / 评分人数
② 可自己设定图表风格

4、问题2 统计分析2001-2016年每年评影人数总量
要求：
① 计算统计出2001-2016年每年评影人数总量，不要求创建函数
② 通过面积图，做图表可视化，分析每年人数总量变化规律
③ 验证是否有异常值（极度异常）
④ 创建函数分析出数据外限最大最小值）
⑤ 筛选查看异常值 → 是否异常值就是每年的热门电影？
提示：
① 通过箱型图验证异常值情况
② 通过quantile(q=0.5)方法，得到四分位数
③ IQR=Q3-Q1
④ 外限：最大值区间Q3+3IQR,最小值区间Q1-3IQR （IQR=Q3-Q1）
⑤ 可自己设定图表风格

'''
```

```python

import csv 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#问题1
aqydata = pd.read_csv(r'C:\Users\acer\Desktop\爱奇艺视频数据.csv',engine = 'python')
#print(aqydata)
```

```python
def data_clean(df):
    col = []
    cols = df.columns 
    for col in cols:
        if df[col].dtype == 'object':
            df[col].fillna('数据缺失',inplace = True)
        else:
            df[col].fillna(0,inplace = True)
    return(df)
#定义数据清洗函数

aqydata = data_clean(aqydata)
print(aqydata.head(2))
```

```python
#问题2   
def date_clean(df):
    rows = df.iloc[:,0]
    new_date = []
    for row in rows:
        row1 = row.replace('年','/').replace('月','/').replace('日','') #清洗掉年月日
        row2 = pd.datetime.date(pd.to_datetime(row1))
        new_date.append(row2)
    df.iloc[:,0] = new_date   #替换老日期
    return(df)
#定义日期清洗函数

aqydata = date_clean(aqydata)
print(aqydata.head())
```

```python
#问题3
evl = aqydata.groupby('导演').sum()
evl['好评率'] = evl['好评数'] / evl['评分人数']  
#计算好评率
evl_ = evl['好评率']
evl_20 = evl_.sort_values(ascending = False)[:20]
#选出好评率前20的导演
print(evl_20)


plt.figure(figsize = (5,5))
evl_20.plot(kind = 'bar',ylim = [0.98,1],color = 'y',width = 0.9,facecolor = 'black',edgecolor = 'r')
plt.grid(axis = 'y',color = 'y',linewidth = 0.8,linestyle = '--')
plt.legend(loc = 'best')
plt.title('不同电影的好评率',fontsize = 25)
```

```python
#问题4
evl_peo = aqydata[['上映年份','评分人数']].groupby('上映年份').sum()
_01_16 = evl_peo[evl_peo.index>2000]
#筛选出2001年到2016年的影评人数
plt.figure(figsize = (16,10))
_01_16.plot.area(color = 'black')
plt.xticks(range(2001,2017))
plt.title('2001年到2016年每年的影评总人数')
plt.grid(axis = 'both')
'''分析：从图表可以看出，从2001年到2009年，尽管每年影评人数有所波动，但总体来说是呈缓慢上升趋势。
     但2011年影评人数骤然上升，可能是和热门电影上映有所关联。到了2012年，影评人数又急剧下降，
     且12年以后，影评人数又有上升趋势，尽管也有下降趋势。可以看出，电影市场是正在缓慢升温，    
'''

evl_pn = aqydata[['上映年份','评分人数','好评数','导演']]
evl_pn = evl_pn[evl_pn['上映年份']>2000].sort_values('上映年份')
fig = plt.figure(figsize = (10,10))
startyear = 2001
for i in range(1,17):
    evl_pn1 = evl_pn[evl_pn['上映年份'] == startyear]
    evl_pn1[['评分人数','好评数']].boxplot(whis = 3,  # IQR为3
                                            return_type='dict',
                                            ax = fig.add_subplot(4,4,i))
    startyear +=1
print('看图可知，每组数据都有异常，且都是极度异常')

def out_value(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1   # IQR
    qmax = q3 + 3 * iqr 
    # 外限最大值
    qmin = q1 - 3 * iqr  
    # 外限最小值
    return(qmax,qmin)
for i in range(2001,2017):
    evl_pn2 = evl_pn[evl_pn['上映年份'] == i].groupby('导演').sum()
    evl_pn3 = evl_pn2[['评分人数','好评数']]
    evl_error = out_value(evl_pn3,'评分人数')
    print(evl_pn3[evl_pn3['评分人数']>evl_error[0]])
    
    print('===============')
    #输出异常值
```
