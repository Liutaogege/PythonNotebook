```python
'''
【项目03】  知乎数据清洗整理和结论研究

作业要求：
1、数据清洗 - 去除空值
要求：创建函数
提示：fillna方法填充缺失数据，注意inplace参数

2、问题1 知友全国地域分布情况，分析出TOP20
要求：
① 按照地域统计 知友数量、知友密度（知友数量/城市常住人口），不要求创建函数
② 知友数量，知友密度，标准化处理，取值0-100，要求创建函数
③ 通过多系列柱状图，做图表可视化
提示：
① 标准化计算方法 = (X - Xmin) / (Xmax - Xmin)
② 可自行设置图表风格

3、问题2 知友全国地域分布情况，分析出TOP20
要求：
① 按照学校（教育经历字段） 统计粉丝数（‘关注者’）、关注人数（‘关注’），并筛选出粉丝数TOP20的学校，不要求创建函数
② 通过散点图 → 横坐标为关注人数，纵坐标为粉丝数，做图表可视化
③ 散点图中，标记出平均关注人数（x参考线），平均粉丝数（y参考线）
提示：
① 可自行设置图表风格

'''
```

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
peopdata = pd.read_csv(r'C:\Users\acer\Desktop\六普常住人口数.csv',encoding = 'gb18030')
zhidata = pd.read_csv(r'C:\Users\acer\Desktop\知乎数据_201701.csv',encoding = 'gb18030')
#读取数据
def f(data):
    columns = data.columns
    for col in columns:
        if data[col].dtype ==  'object':
            data[col].fillna('缺失数据', inplace = True)
        else:
            data[col].fillna(0, inplace = True)
    return(data)
peopdata = f(peopdata)
zhidata = f(zhidata)
df_zhihu = zhidata[['居住地','教育经历','关注者','关注']]   
df_peop = peopdata[['常住人口','地区']]
#选取出所需数据
#print(df_zhihu.head(),'\n',df_peop.head())
df_zhihu_peop = df_zhihu['居住地'].value_counts(sort = True)
#算出知乎用户居住地数
df_peop = df_peop.groupby(['地区']).sum()
df_peop.index = df_peop.index.str[:2]  
#修改常住人口数的城市名
df_total = df_peop.join(df_zhihu_peop,how = 'inner')
#将常住人口数和知乎数量合并
df_total.columns = ['常住人口','知友数量']
#更改合并的列名
df_total['知友密度'] = df_total['知友数量']/df_total['常住人口']
#计算人口密度
df_total_1 = df_total.sort_values('知友数量',ascending = False)
#知友数量排序
df_total_2 = df_total.sort_values('知友密度',ascending = False)
#知友密度排序
#print(df_total_1)
#print(df_total_2)

def f(df):
    df1 = (df - df.min())/(df.max()- df.min()) * 100
    return(df1[:20])
#创建标准计算化函数

df_total_20_1_bz = f(df_total_1['知友数量'])
df_total_20_2_bz = f(df_total_2['知友密度'])
#标准化计算

plt.figure(figsize = (15,6))
df_total_20_1_bz.plot(kind = 'bar',color = 'y',width = 0.9,grid = True)
for i,j in zip(range(len(df_total_20_1_bz.index)),df_total_20_1_bz):
    plt.text(i - 0.45,j + 0.0001*i,'%.2f'%j ,color = 'red')
#标准化计算知友数量
plt.title('知友数量TOP20\n')

plt.figure(figsize = (12,6))
df_total_20_2_bz.plot(kind = 'bar',color = 'g',width = 0.9,grid = True)
for i,j in zip(range(len(df_total_20_2_bz.index)),df_total_20_2_bz):
    plt.text(i - 0.5,j + 0.0001*i,'%.2f'%j ,color = 'black',fontsize = 8.9)
#标准化计算知友密度
plt.title('知友密度TOP20\n')
```

```python

df_dist = zhidata[['教育经历','关注者','关注']] 
df_edu = df_dist.groupby(['教育经历']).sum().drop(['五道口男子职业技术学院','为往圣继绝学','我的老師，是山川和大地','重庆第一工程尸培养基地','缺失数据','医学','Phd Candidate','本科']).sort('关注者',ascending = False)[:20]
#筛选出前关注人数前20的学校
large = np.random.randint(15,35,len(df_edu))
x = df_edu['关注']
y = df_edu['关注者']  #设定坐标
attpeop_mean = df_edu['关注者'].mean()
att_mean = df_edu['关注'].mean()

plt.scatter(x,y,s = large,cmap = 'categorical',marker = '*',edgecolors = 'r',label = '学校')
plt.grid()
plt.axvline(att_mean,label="平均关注人数：%i人" % att_mean,color='r',linestyle='--',alpha=1) 
# 添加平均关注人数线
plt.axhline(attpeop_mean,label="平均粉丝数：%i人" % attpeop_mean,color='gold',linestyle='--',alpha=1)   
# 添加平均粉丝数线

plt.legend(loc = 'upper left')
m = 0
for i,j in zip(df_edu['关注'],df_edu['关注者']):
    m = m + 1
    plt.text(i - 0.5,j + 0.0001*i,df_edu.index[m - 1] ,color = 'black',fontsize = 8.9)
```
