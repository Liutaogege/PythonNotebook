```python
'''
【项目01】  商铺数据加载及存储

作业要求：
1、成功读取“商铺数据.csv”文件
2、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
3、数据清洗：
① comment，price两个字段清洗成数字
② 清除字段缺失的数据
③ commentlist拆分成三个字段，并且清洗成数字
4、结果存为.pkl文件

'''
```

```python
import csv
import pickle
path = r'C:\Users\acer\Desktop\商铺数据.csv'
f = csv.reader(open(path,'r',encoding = 'utf8'))
save =[]
for row in f:
    #print(row)
    classify = row[0]
    name = row[1]
    comment = row[2]
    star = row[3]
    price = row[4]
    address = row[5]
    commenlist = row[6]
    if comment == '我要点评' or comment == 'comment':
        continue
    for k in range(len(commenlist)):
        if commenlist[k] != '口':
            continue
    for j in range(len(comment)):
        if comment[j] == '条':
            comment1 = comment[:j].strip()
    lst = commenlist.split(' ')
    lst.sort(reverse = True)
    for m in range(len(lst)):
        if lst[m] == '':  
            break
        lst1 = lst[0:m+1]
    commenlist1 = [lst1[0][2:],lst1[1][2:],lst1[2][2:]]
    for i in range(len(price)):
        if price[i] != '￥':
            continue
        elif price[i] == '￥':
            data = [['classify',classify],['name',name],['comment',comment1],['star',star],['price',price[i + 1:]],['address',address],['commenlist',commenlist1]]
            save.append(dict(data))


print(save) 
pic = open('C:\\Users\\acer\\Desktop\\save.pkl','wb')
pickle.dump(save,pic)
pic.close()
print('finished')
```
