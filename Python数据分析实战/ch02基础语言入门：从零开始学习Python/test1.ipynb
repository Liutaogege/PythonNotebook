```python
import numpy as np
import scipy.stats as sta  
import matplotlib.pyplot as plt 
%matplotlib inline
from scipy.stats import binom as B  # 二项分布




rv = B(10,0.5)

x = np.arange(0,11,1)
y = rv.pmf(x)

#plt.bar(x,y)
plt.bar(x,y,color = 'red')
```

```python
from scipy.stats import geom as G  # 几何分布
rv1 = G(0.2)
x1 = np.arange(1,11,1)
y1 = rv1.pmf(x1)
plt.bar(x1,y1,color = 'red')
```

```python
from scipy.stats import poisson as Pie #泊松分布
rv2 = Pie(4.8)
x2 = np.arange(0,11,1)
y2 = rv2.pmf(x2)
plt.bar(x2,y2)
```

```python
from scipy.stats import norm as N  # 正态分布
import pandas as pd
x = np.linspace(-10,10,100)
rv1 = N(loc = 0,scale=1)
rv2 = N(loc = -5,scale=1)
rv3 = N(loc = 0,scale=3)
plt.plot(x,rv1.pdf(x))
plt.plot(x,rv2.pdf(x))
plt.plot(x,rv3.pdf(x))

df = pd.DataFrame(np.random.randn(10))
df.values.sort()
plt.plot(df.index,df.values.sort())
df
```
