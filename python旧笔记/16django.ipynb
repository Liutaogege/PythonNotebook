# 基于类的视图
- 和基于函数的视图的优势和区别
    - HTTP方法的methode可以有各自的方法，不需要使用条件分支来解决
    - 可以使用OOP技术（例如Mixin，把来自父类的行为和属性组合在一起，还有ListView）
- 概述
    - 核心是允许使用不同的实例方法来响应不同的HTTP请求方法，而避开条件分支实现.从本质上讲，基于类的视图允许您使用不同的类实例方法响应不同的HTTP请求方法，而不是在单个视图函数中使用条件分支代码。
    - 基于类的视图具有一个 as_view()类方法，该方法返回一个函数，当请求到达匹配关联模式的URL时，可以调用该函数。该函数创建类的实例，调用 setup()初始化其属性，然后调用其dispatch()方法。 dispatch查看请求以确定它是否为GET， POST等等，并将请求中继到匹配方法（如果已定义），若方法没有定义，则引发HttpResponseNotAllowed
- 类属性使用
    - 在类定义时直接覆盖
    - 在调用as_view的时候直接作为参数使用
            urlpatters = [
            path('about'/,GreetingView.as_view(greeting="G'day")),]
- 对基于类的视图的扩充大致有三种方法：Mixin，装饰as_view，装饰dispatch
- 使用Mixin
    - 多继承的一种形式，来自弗雷的行为和属性组合在一起
    - 解决多继承问题
    - view的子类只能单继承，多继承会导致不可期问题
    - 多继承带来的问题
         = 结构复杂
- 装饰基于类的视图
- 介绍
    - 基于类的视图的扩展不限于使用mixins。您也可以使用装饰器。由于基于类的视图不是函数，因此根据您使用as_view()或创建子类，装饰它们的工作方式也不同。
