# 数据结构的概念
- 数据元素是表示数据的基本单位，是数据这个集合中的一个个体。在数据结构中数据元素经常被称为结点
- 一个数据元素可以由若干个数据项组成
    - 数据项可分为：初等项和组合项
        - 初等项：具有独立含义的最小标识单位，比如买菜费用中，蔬菜费用中的土豆费用
        - 组合项：若干个初等项组成，比如蔬菜费
- 数据对象：是具有相同性质的数据元素的集合，是数据集合的一个子集
    - 例如：整数数据对象集合，英文字母数据对象集合，比如某公司研发部人眼的工资表集合，而每个人的工资情况就是一个数据元素
- 一般的数据结构都联系着三个方面：
    - 数据的逻辑结构：数据之间的逻辑关系
    - 数据存储结构：数据在计算机中存储方式
    - 数据运算集合：数据的操作
    
    
## 数据的逻辑结构
- 在不发生混淆的情况下，通常把数据的逻辑结构直接成为数据结构
- 数据的逻辑结构可分类两大类：
    - 线性结构：有且仅有一个开始节点和一个终端节点，并且每个节点至多有一个前驱和一个后继
    - 非线性结构：
        - 一个节点有一个前驱和多个后继：树形结构
        - 多个前驱和多个后继：图结构

## 数据的存储结构
- 数据的存储结构是数据的逻辑结构在计算机存储器中的实现，依赖于计算机。
- 计算机的存储器是由多个存储单元组成的，每个存储单元都一个一个唯一的地址，各个存储单元的地址是连续编码的，每个存储单元都有唯一的后继单元，他们称为相邻单元，一片连续的存储单元的整体称为存储区域
- 数据的存储结构有以下四种基本的存储方式
    - 1.顺序存储
        - 把逻辑上相邻的结点存储在物理位置上也相邻的存储单元里，结点之间的逻辑关系用存储单元的邻接关系来体现
        - 通常使用高级语言中的一维数组来实现的
    - 2.链接存储方式
        - 不要求存储单元的相邻，结点之间的逻辑关系是由附加的指针字段来表示的
    - 3.索引存储方式
        - 在存储结点数据的同时，还需要建立附加的索引表，索引表中的每一项称为索引项，一般由关键字和地址组成。关键字是能够唯一标示一个结点的一个或多个字段，若每个结点在索引表中都有一个索引项，则称该索引为稠密索引，若一组结点在索引表中只对应一个索引项，则称为稀疏索引。稀疏索引中索引项的地址则是一组结点的起始位置
    - 4.散列存储方式
        - 根据结点的关键字，通过反映结点关键字与存储地址之间的对应关系的函数直接计算出该结点的存储地址
- 如果结构所占用的存储区域都分配给了数据，则这样的数据称为紧凑结构，否则称为非紧凑结构，顺序结构就是紧凑结构

## 数据的运算
- 数据的运算是定义在数据的逻辑结构上的，每种逻辑结构都有一个运算的集合，这些运算实际上是定义在抽象的数据上所施加的一系列的操作(所谓抽象，是指知道这些操作是什么，但是不知道具体操作）

# 数据类型与抽象数据类型
- 数据类型是一组性质相同的集合以及在这些值上定义一组操作（运算）的总称。每种语言都有自己的数据类型定义，比如Python中的整数，列表，字典
- 数据类型已经反映出对数据的抽象，而抽象数据类型是更高一级的抽象
- 抽象数据类型：是指数据的组织和与之相关的操作，通常是由用户自己定义（感觉很像类）。（可以看作是是描述问题的模型，独立于具体实现）
